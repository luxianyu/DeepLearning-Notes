这种图像分类称为语义图像分割 (Semantic Image Segmentation)。它类似于目标检测（Object Detection），两者都在问：“这张图像中有什么物体，这些物体在图像中的位置在哪里？”，但区别在于：
- 目标检测使用边界框标注物体，边界框内可能包含非物体像素；
- 语义图像分割可以为图像中的每个物体生成精确的掩码，通过给图像中的每个像素分配对应类别标签实现精确标注。

这里的“语义”指的是像素所表示的具体内容。例如，下图中，“Car” 类用深蓝色掩码表示，而“Person” 类用红色掩码表示：

### 2.2 - 数据预处理


```python
import torch
from torch.utils.data import Dataset, DataLoader
from PIL import Image
import numpy as np
import torchvision.transforms.functional as TF
import matplotlib.pyplot as plt

# ===============================
# 1. 查看前 N 个文件路径
# ===============================
N = 2
for img_path, mask_path in zip(image_list[:N], mask_list[:N]):
    print(f"Image: {img_path}, Mask: {mask_path}")

# ===============================
# 2. process_path
# ===============================
def process_path(image_path, mask_path):
    """
    读取图像和 mask，并转换为 Tensor
    """
    # 读取图像
    img = Image.open(image_path).convert("RGB")
    img = TF.to_tensor(img)  # [C,H,W], float32, 0~1

    # 读取 mask
    mask = Image.open(mask_path).convert("RGB")
    mask = TF.to_tensor(mask)  # [C,H,W], 0~1
    mask, _ = torch.max(mask, dim=0, keepdim=True)  # reduce_max, [1,H,W]

    return img, mask

# ===============================
# 3. preprocess
# ===============================
def preprocess(image, mask, size=(96,128)):
    """
    对图像和 mask 进行 resize
    """
    image = TF.resize(image, size, interpolation=Image.NEAREST)
    mask = TF.resize(mask, size, interpolation=Image.NEAREST)
    return image, mask

# ===============================
# 4. PyTorch Dataset
# ===============================
class SegmentationDataset(Dataset):
    def __init__(self, image_files, mask_files, resize=None):
        self.image_files = image_files
        self.mask_files = mask_files
        self.resize = resize

    def __len__(self):
        return len(self.image_files)

    def __getitem__(self, idx):
        img_path = self.image_files[idx]
        mask_path = self.mask_files[idx]

        # 原始尺寸
        image_orig, mask_orig = process_path(img_path, mask_path)

        # resize（可选）
        if self.resize:
            image_resized, mask_resized = preprocess(image_orig, mask_orig, self.resize)
        else:
            image_resized, mask_resized = image_orig, mask_orig

        return image_resized, mask_resized, image_orig, mask_orig, img_path, mask_path

# ===============================
# 5. 创建 Dataset 和 DataLoader
# ===============================
image_ds = SegmentationDataset(image_list, mask_list, resize=None)       # 原始尺寸
processed_image_ds = SegmentationDataset(image_list, mask_list, resize=(96,128))  # resize

dataloader_orig = DataLoader(image_ds, batch_size=1, shuffle=False)
dataloader_resized = DataLoader(processed_image_ds, batch_size=1, shuffle=False)
```

## 3 - U-Net

U-Net，因其U形结构而得名，最初于2015年为肿瘤检测而创建，但近年来已成为其他语义分割任务中非常流行的选择。

U-Net 基于之前的架构，称为全卷积网络（Fully Convolutional Network，FCN），该网络用转置卷积层（transposed convolution layer）替换了典型 CNN 中的全连接层，以将特征图上采样回原始输入图像的大小，同时保留空间信息。这是必要的，因为全连接层会破坏空间信息（图像中的“位置”信息），而这在图像分割任务中至关重要。使用转置卷积的额外好处是，输入尺寸不再需要固定，就像使用全连接层时那样。

不幸的是，FCN 的最终特征层由于下采样过多而导致信息丢失。信息丢失太多后，再进行上采样会变得困难，从而产生输出看起来粗糙的结果。

U-Net 在 FCN 的基础上进行了改进，设计上有一定相似性，但在一些关键方面有所不同。网络末端不再使用单一的转置卷积，而是为下采样输入图像到特征图和上采样这些特征图回到原始输入图像大小的每一层使用匹配数量的卷积和转置卷积。它还增加了跳跃连接（skip connections），以保留在编码过程中可能丢失的信息。跳跃连接将信息从编码器中对应的下采样层传递到解码器中的每个上采样层，捕捉更细的信息，同时保持计算量较低。这有助于防止信息丢失以及模型过拟合。


<a name='3-1'></a>
### 3.1 - 模型细节

<img src="images/unet.png" style="width:700px;height:400;">
<caption><center> <u><b> 图2 </u></b>: U-Net 架构<br> </center></caption>

**收缩路径**（包含下采样步骤的编码器）：

图像首先经过多个卷积层，这些卷积层会减少图像的高度和宽度，同时增加通道数。

收缩路径遵循常规 CNN 架构，包括卷积层及其激活函数、以及用于下采样图像并提取特征的池化层。具体来说，它由重复应用两个 3 x 3 的无填充卷积组成，每个卷积后跟一个 ReLU 激活函数，再加一个 2 x 2 的最大池化操作（步幅为2）用于下采样。在每次下采样步骤中，特征通道数量加倍。

**裁剪函数（Crop function）**：这一步会裁剪收缩路径中的图像，并将其与当前扩展路径上的图像进行拼接，以创建跳跃连接（skip connection）。

**扩展路径**（包含上采样步骤的解码器）：

扩展路径执行与收缩路径相反的操作，将图像恢复到原始尺寸，同时逐步缩小通道数。

具体来说，扩展路径的每一步都会对特征图进行上采样，然后进行一个 2 x 2 卷积（转置卷积）。该转置卷积会将特征通道数量减半，同时增加图像的高度和宽度。

接下来，将其与收缩路径中对应裁剪的特征图进行拼接，然后执行两个 3 x 3 卷积，每个卷积后跟一个 ReLU 激活函数。每次卷积都会丢失边界像素，因此需要进行裁剪以处理这个问题。

**最终特征映射块**：在最后一层，使用 1 x 1 卷积将每个 64 维特征向量映射到所需的类别数量。上一层的通道维度对应使用的滤波器数量，因此通过选择适当数量的 1 x 1 卷积滤波器，可以改变通道维度。当这个思想应用到最后一层时，可以将通道维度减少到每个类别对应一层。

U-Net 网络总共有 23 层卷积层。


<a name='3-2'></a>
### 3.2 - 编码器（下采样模块）

<img src="images/encoder.png" style="width:500px;height:500;">
<caption><center> <u><b>图3</u></b>: U-Net 编码器详细视图 <br> </center></caption>

编码器是由多个 `conv_block` 堆叠而成：

每个 `conv_block()` 由 2 个 **Conv2D** 层组成，并使用 ReLU 激活函数。我们将对部分 `conv_block` 应用 **Dropout** 和 **MaxPooling2D**，如后续章节所示，尤其是在下采样的最后两个模块中。

该函数将返回两个张量：
- `next_layer`：将输入到下一个卷积块。
- `skip_connection`：将输入到对应的解码块。

**注意**：如果 `max_pooling=True`，`next_layer` 将是 MaxPooling2D 层的输出，而 `skip_connection` 将是先前应用的层（Conv2D 或 Dropout）的输出，否则两者结果相同。

<a name='ex-1'></a>
### 练习 1 - conv_block

实现 `conv_block(...)`。以下是 `conv_block`（收缩块）每步的指令：

* 添加 2 个 **Conv2D** 层，每层滤波器数量为 `n_filters`，`kernel_size` 设置为 3，`kernel_initializer` 使用 ['he_normal'](https://www.tensorflow.org/api_docs/python/tf/keras/initializers/HeNormal)，`padding` 设置为 'same'，激活函数为 'relu'。
* 如果 `dropout_prob` > 0，则添加一个 Dropout 层，参数为 `dropout_prob`。
* 如果 `max_pooling` 设置为 True，则添加一个 2x2 池化的 MaxPooling2D 层。


```python
import torch
import torch.nn as nn
import torch.nn.functional as F

# ===============================
# conv_block (PyTorch 版本)
# ===============================
class ConvBlock(nn.Module):
    def __init__(self, in_channels, n_filters=32, dropout_prob=0, max_pooling=True):
        """
        Convolutional downsampling block

        Arguments:
            in_channels -- 输入通道数
            n_filters -- 卷积层通道数
            dropout_prob -- Dropout 概率
            max_pooling -- 是否使用 MaxPool2d 下采样
        """
        super(ConvBlock, self).__init__()

        # 两个 3x3 卷积层 + ReLU
        self.conv1 = nn.Conv2d(in_channels, n_filters, kernel_size=3, padding=1)
        self.conv2 = nn.Conv2d(n_filters, n_filters, kernel_size=3, padding=1)

        # He 正态初始化 (对应 TensorFlow he_normal)
        nn.init.kaiming_normal_(self.conv1.weight)
        nn.init.kaiming_normal_(self.conv2.weight)

        self.relu = nn.ReLU(inplace=True)

        # Dropout 层
        self.use_dropout = dropout_prob > 0
        if self.use_dropout:
            self.dropout = nn.Dropout2d(dropout_prob)

        # MaxPooling
        self.use_maxpool = max_pooling
        if self.use_maxpool:
            self.maxpool = nn.MaxPool2d(kernel_size=2, stride=2)

    def forward(self, x):
        # 两个卷积层
        x = self.relu(self.conv1(x))
        x = self.relu(self.conv2(x))

        # Dropout
        if self.use_dropout:
            x = self.dropout(x)

        # skip connection 保存卷积后的特征
        skip_connection = x

        # 下采样
        if self.use_maxpool:
            next_layer = self.maxpool(x)
        else:
            next_layer = x

        return next_layer, skip_connection
```

```python
import torch
import torch.nn as nn
from torchsummary import summary

# ===============================
# ConvBlock 已经定义过，这里直接复用
# ===============================
# class ConvBlock(nn.Module):
#     ...

# ===============================
# 测试 Block 1
# ===============================
input_size = (3, 96, 128)  # PyTorch: [C,H,W]
n_filters = 32

# 构建 Block 1
block1 = ConvBlock(in_channels=input_size[0], n_filters=n_filters, dropout_prob=0, max_pooling=True)

# 模拟 Input
x = torch.randn(1, *input_size)  # batch_size=1

# 前向测试
next_layer, skip_connection = block1(x)

print("Block 1 Test:")
print("Next layer shape:", next_layer.shape)       # 下采样后
print("Skip connection shape:", skip_connection.shape)  # 卷积输出

# 使用 torchsummary 打印参数信息
print("\nBlock 1 Summary:")
summary(block1, input_size)

# ===============================
# 测试 Block 2
# ===============================
block2 = ConvBlock(in_channels=input_size[0], n_filters=n_filters*32, dropout_prob=0.1, max_pooling=True)

next_layer2, skip_connection2 = block2(x)

print("\nBlock 2 Test:")
print("Next layer shape:", next_layer2.shape)
print("Skip connection shape:", skip_connection2.shape)

print("\nBlock 2 Summary:")
summary(block2, input_size)
```

### 3.3 - 解码器（上采样模块）

解码器（或上采样模块）用于将特征图还原回与原始图像相同的尺寸。在每个上采样阶段，你需要将对应编码器块的输出进行拼接（skip connection），然后再输入到下一个解码块。

<img src="images/decoder.png" style="width:500px;height:500;">
<caption><center> <u><b>图4</u></b>: U-Net 解码器详细视图 <br> </center></caption>

解码器中包含两个新组件：`up` 和 `merge`。  
- `up` 是转置卷积层（Transpose Convolution），用于上采样。  
- `merge` 是跳跃连接（skip connection），将对应编码器层的特征与当前层拼接起来。  

此外，解码器还包含两个与编码器相同参数的卷积层。

在此部分，你会遇到 `Conv2DTranspose` 层，它的作用与 `Conv2D` 相反。你可以在 [TensorFlow 官方文档](https://www.tensorflow.org/api_docs/python/tf/keras/layers/Conv2DTranspose) 中了解更多。

### 练习 2 - upsampling_block

实现 `upsampling_block(...)`。

该函数的功能是构建上采样块，具体步骤如下：

* 接收两个输入：
  - `expansive_input`：来自上一层的输入张量；
  - `contractive_input`：来自对应跳跃连接（skip connection）的张量。

* 使用与之前下采样块相同数量的滤波器 `n_filters`。

* 添加一个 `Conv2DTranspose` 层：
  - 滤波器数量：`n_filters`
  - 卷积核大小：`(3,3)`
  - 步幅（stride）：`(2,2)`
  - 填充方式：`'same'`
  - 作用对象：`expansive_input`（上一层的输出）

* 将 `Conv2DTranspose` 层的输出与 `contractive_input` 拼接（skip connection），拼接轴为 `axis=3`。  
  👉 拼接顺序请保持为 `[up, contractive_input]`，以确保与评分器一致。

* 最后，添加两个 **Conv2D** 层：
  - 激活函数：`'relu'`
  - 初始化器：`'he_normal'`
  - 填充方式：`'same'`

这些参数与编码器部分的两层 Conv2D 保持一致。


```python
import torch
import torch.nn as nn
import torch.nn.functional as F

# ===============================
# upsampling_block (PyTorch 版本)
# ===============================
class UpsamplingBlock(nn.Module):
    def __init__(self, in_channels, skip_channels, n_filters=32):
        """
        Convolutional upsampling block

        Arguments:
            in_channels -- 输入通道数（expansive_input）
            skip_channels -- skip connection 通道数（contractive_input）
            n_filters -- 卷积层输出通道数
        """
        super(UpsamplingBlock, self).__init__()

        # 转置卷积上采样
        self.upconv = nn.ConvTranspose2d(in_channels, n_filters, kernel_size=3, stride=2, padding=1, output_padding=1)

        # 两个卷积层
        self.conv1 = nn.Conv2d(n_filters + skip_channels, n_filters, kernel_size=3, padding=1)
        self.conv2 = nn.Conv2d(n_filters, n_filters, kernel_size=3, padding=1)

        # He 初始化
        nn.init.kaiming_normal_(self.upconv.weight)
        nn.init.kaiming_normal_(self.conv1.weight)
        nn.init.kaiming_normal_(self.conv2.weight)

        self.relu = nn.ReLU(inplace=True)

    def forward(self, expansive_input, contractive_input):
        # 上采样
        up = self.upconv(expansive_input)  # [B, n_filters, H*2, W*2]

        # 拼接 skip connection
        merge = torch.cat([up, contractive_input], dim=1)  # PyTorch 通道维度 dim=1

        # 两个卷积
        conv = self.relu(self.conv1(merge))
        conv = self.relu(self.conv2(conv))

        return conv
```

```python
import torch
import torch.nn as nn
from torchsummary import summary

# ===============================
# 上采样块已定义 (复用之前的 UpsamplingBlock)
# ===============================
# class UpsamplingBlock(nn.Module):
#     ...

# ===============================
# 测试 Block 1
# ===============================
n_filters = 32

# PyTorch 输入尺寸: [C,H,W]
input_size1 = (256, 12, 16)  # expansive_input
input_size2 = (128, 24, 32)  # contractive_input

# 构建上采样块
block1 = UpsamplingBlock(in_channels=input_size1[0], skip_channels=input_size2[0], n_filters=n_filters)

# 模拟输入
expansive_inputs = torch.randn(1, *input_size1)
contractive_inputs = torch.randn(1, *input_size2)

# 前向传播测试
output = block1(expansive_inputs, contractive_inputs)
print("Output shape:", output.shape)  # [B, n_filters, H*2, W*2] = [1, 32, 24, 32]

# 使用 torchsummary 打印参数信息
print("\nUpsampling Block 1 Summary:")
summary(block1, [(input_size1[0], input_size1[1], input_size1[2]),
                  (input_size2[0], input_size2[1], input_size2[2])])
```

### 练习 3 - unet_model

实现函数 `unet_model(...)`，定义 U-Net 的整体结构，包括输入形状、卷积核数量以及类别数。

#### ✅ 编码器部分（Contracting Path）

1. 从输入层开始，传入第一个 `conv_block`；
2. 将每个卷积块（conv_block）的第一个输出连接到下一个卷积块的输入；
3. 每进入下一层，将滤波器数量 **翻倍**；
4. 从 `conv_block4` 开始，添加 `dropout=0.3`；
5. 在最后一个 `conv_block` 中，同样设置 `dropout=0.3`，但 **关闭 max pooling（max_pooling=False）**。

---

#### 🧠 瓶颈层（Bottleneck）

* 使用 `cblock5` 作为 `expansive_input`，`cblock4` 作为 `contractive_input`；
* 滤波器数量为 `n_filters * 8`；
* 这是模型中最深的层，负责提取最抽象的特征表示。

---

#### 🔁 解码器部分（Expanding Path）

1. 每一步上采样时：
   - `expansive_input` 为上一层解码输出；
   - `contractive_input` 为对应的编码器输出（即 max pooling 之前的结果）；
2. 每向上一层，上采样的滤波器数量减半；
3. 通过 `Conv2DTranspose` 进行上采样，然后与对应的 skip connection 拼接；
4. 使用两个 **Conv2D** 层（ReLU 激活，He normal 初始化，`same` 填充）进行特征恢复。

---

#### 🎯 输出层（Final Output Layer）

1. 添加一个卷积层 `conv9`：
   - 使用 ReLU 激活；
   - He normal 初始化；
   - `padding='same'`。
2. 添加最终的 `conv10`：
   - 卷积核数量：类别数（`num_classes`，此处为 23）；
   - 核大小：`(1,1)`；
   - 填充：`'same'`；
   - 输出结果即为整个模型的输出张量。

---

最终模型的结构如下：
输入图像
↓
[编码器：多层 conv_block + 下采样]
↓
[瓶颈层：特征压缩与抽象]
↓
[解码器：上采样 + skip connections]
↓
[1x1 卷积 → 输出 23 类预测]


这就是完整的 U-Net 模型框架。



```python
import torch
import torch.nn as nn
import torch.nn.functional as F

# ===============================
# 定义 U-Net 模型 (PyTorch)
# ===============================
class UNet(nn.Module):
    def __init__(self, input_channels=3, n_filters=32, n_classes=23):
        super(UNet, self).__init__()

        # --------------------
        # Contracting path (Encoder)
        # --------------------
        self.cblock1 = ConvBlock(input_channels, n_filters*1)
        self.cblock2 = ConvBlock(n_filters*1, n_filters*2)
        self.cblock3 = ConvBlock(n_filters*2, n_filters*4)
        self.cblock4 = ConvBlock(n_filters*4, n_filters*8, dropout_prob=0.3)
        self.cblock5 = ConvBlock(n_filters*8, n_filters*16, dropout_prob=0.3, max_pooling=False)

        # --------------------
        # Expanding path (Decoder)
        # --------------------
        self.ublock6 = UpsamplingBlock(n_filters*16, n_filters*8, n_filters*8)
        self.ublock7 = UpsamplingBlock(n_filters*8, n_filters*4, n_filters*4)
        self.ublock8 = UpsamplingBlock(n_filters*4, n_filters*2, n_filters*2)
        self.ublock9 = UpsamplingBlock(n_filters*2, n_filters*1, n_filters*1)

        # --------------------
        # 最后的卷积层
        # --------------------
        self.conv9 = nn.Conv2d(n_filters*1, n_filters, kernel_size=3, padding=1)
        self.relu = nn.ReLU(inplace=True)
        self.conv10 = nn.Conv2d(n_filters, n_classes, kernel_size=1)

        # He 初始化
        nn.init.kaiming_normal_(self.conv9.weight)
        nn.init.kaiming_normal_(self.conv10.weight)

    def forward(self, x):
        # Encoder
        x1, skip1 = self.cblock1(x)
        x2, skip2 = self.cblock2(x1)
        x3, skip3 = self.cblock3(x2)
        x4, skip4 = self.cblock4(x3)
        x5, skip5 = self.cblock5(x4)  # max_pooling=False, x5=skip5

        # Decoder
        x6 = self.ublock6(x5, skip4)
        x7 = self.ublock7(x6, skip3)
        x8 = self.ublock8(x7, skip2)
        x9 = self.ublock9(x8, skip1)

        x9 = self.relu(self.conv9(x9))
        out = self.conv10(x9)  # 输出类别数 n_classes

        return out
```

### 3.6 - 损失函数（Loss Function）

在语义分割（semantic segmentation）中，你需要的掩码数量与目标类别的数量相同。  
在你所使用的数据集中，每个掩码中的每个像素都被赋予了一个整数标签，用来表示该像素属于某个类别的概率，其取值范围从 0 到 `num_classes - 1`。  
正确的类别对应于具有最高概率的那一层。  

这与 **categorical crossentropy（分类交叉熵）** 不同，在分类交叉熵中，标签必须是 **one-hot 编码**（即仅包含 0 和 1 的向量）。  
在这里，你将使用 **sparse categorical crossentropy（稀疏分类交叉熵）** 作为损失函数，用于执行逐像素的多类别预测。  

当你需要处理大量类别时，**稀疏分类交叉熵** 比其他损失函数更加高效。


```python
import torch
import torch.nn as nn
import torch.optim as optim

# ===============================
# 假设 unet 是之前定义的 PyTorch 模型
# ===============================
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
unet.to(device)

# -------------------------------
# 优化器
# -------------------------------
optimizer = optim.Adam(unet.parameters(), lr=1e-3)

# -------------------------------
# 损失函数
# SparseCategoricalCrossentropy(from_logits=True)
# PyTorch 对应 nn.CrossEntropyLoss()
# 输入 shape: [B, n_classes, H, W]
# 目标 shape: [B, H, W], 值为类别索引 (0..n_classes-1)
# -------------------------------
criterion = nn.CrossEntropyLoss()

# -------------------------------
# 训练示例
# -------------------------------
# 假设 imgs, masks 是 batch 数据
# imgs: [B, 3, 96, 128], masks: [B, 96, 128] (整数类别索引)
imgs = torch.randn(4, 3, 96, 128).to(device)
masks = torch.randint(0, 23, (4, 96, 128)).to(device)

# 前向传播
outputs = unet(imgs)  # [B, n_classes, H, W]

# 计算损失
loss = criterion(outputs, masks)

# 反向传播
optimizer.zero_grad()
loss.backward()
optimizer.step()

print("Loss:", loss.item())
```

```python
# ===============================
# 6. display 函数
# ===============================
def display(display_list):
    plt.figure(figsize=(15, 15))
    title = ['Input Image', 'True Mask', 'Predicted Mask'][:len(display_list)]
    for i, item in enumerate(display_list):
        plt.subplot(1, len(display_list), i+1)
        plt.title(title[i])
        if isinstance(item, torch.Tensor):
            if item.dim() == 3:  # [C,H,W]
                img = item.permute(1,2,0).numpy()
            else:  # [H,W]
                img = item.numpy()
            plt.imshow(img)
        else:
            plt.imshow(item)
        plt.axis('off')
    plt.show()
```

### 4.1 - 创建预测掩码（Create Predicted Masks）

现在，定义一个函数，使用 `tf.argmax` 在类别数量的维度上返回最大值的索引，并将预测结果合并为单张图像：


```python
def create_mask(pred_mask):
    pred_mask = tf.argmax(pred_mask, axis=-1)
    pred_mask = pred_mask[..., tf.newaxis]
    return pred_mask[0]
```

# 图像分割与 U-Net

在本次作业中，你将构建自己的 U-Net，这是专门为快速、精确图像分割设计的一种卷积神经网络 (CNN)，并用它对图像中的每个像素进行预测——在本例中，是来自自动驾驶汽车数据集的图像。

这种图像分类称为语义图像分割 (Semantic Image Segmentation)。它类似于目标检测（Object Detection），两者都在问：“这张图像中有什么物体，这些物体在图像中的位置在哪里？”，但区别在于：
- 目标检测使用边界框标注物体，边界框内可能包含非物体像素；
- 语义图像分割可以为图像中的每个物体生成精确的掩码，通过给图像中的每个像素分配对应类别标签实现精确标注。

这里的“语义”指的是像素所表示的具体内容。例如，下图中，“Car” 类用深蓝色掩码表示，而“Person” 类用红色掩码表示：

### 2 - 将数据集拆分为未标注图像和标注图像及数据预处理


````python
# ===============================
# 1. 查看前 N 个文件路径
# ===============================
N = 2  # 设置查看前 N 个样本
for img_path, mask_path in zip(image_list[:N], mask_list[:N]):
    # zip 将两个列表打包成成对元素 (img_path, mask_path)
    print(f"Image: {img_path}, Mask: {mask_path}")  # 打印图像路径和对应的 mask 路径

# ===============================
# 2. process_path 函数
# ===============================
def process_path(image_path, mask_path):
    """
    读取单张图像和 mask，并转换为 PyTorch 张量

    参数:
        image_path (str): 图像文件路径
        mask_path  (str): mask 文件路径

    返回:
        img  (torch.Tensor): 图像张量, shape=[C,H,W], float32, 范围[0,1]
        mask (torch.Tensor): mask 张量, shape=[H,W], long 类型
    """
    # 打开图像，并确保为 RGB 三通道
    img = Image.open(image_path).convert("RGB")
    img = TF.to_tensor(img)  # 转为 [C,H,W] float32，范围 [0,1]

    # 打开 mask，并转换为灰度图（单通道）
    mask = Image.open(mask_path).convert("L")
    mask = torch.from_numpy(np.array(mask)).long()  # 转为 long 类型张量 [H,W]

    return img, mask

# ===============================
# 3. preprocess 函数
# ===============================
def preprocess(image, mask, size=(96,128)):
    """
    对图像和 mask 进行 resize

    参数:
        image (torch.Tensor): 原始图像张量 [C,H,W]
        mask  (torch.Tensor): 原始 mask 张量 [H,W]
        size  (tuple): 目标尺寸 (H,W)

    返回:
        image_resized (torch.Tensor): resize 后图像 [C,H,W], float32
        mask_resized  (torch.Tensor): resize 后 mask [H,W], long
    """
    # 图像 resize，使用最近邻插值 (Image.NEAREST) 保留像素值
    image = TF.resize(image, size, interpolation=Image.NEAREST)

    # mask 先加一个通道维度 [1,H,W] 再 resize
    mask = mask.unsqueeze(0).float()  # [1,H,W] 转 float，便于 resize
    mask = TF.resize(mask, size, interpolation=Image.NEAREST)
    mask = mask.squeeze(0).long()     # 去掉通道维度，恢复为 [H,W] long 类型

    return image, mask

# ===============================
# 4. 自定义 Dataset
# ===============================
class SegmentationDataset(Dataset):
    """
    PyTorch 自定义 Dataset，用于语义分割任务
    """
    def __init__(self, image_files, mask_files, resize=(96,128)):
        """
        参数:
            image_files (list): 图像文件路径列表
            mask_files  (list): mask 文件路径列表
            resize      (tuple 或 None): resize 尺寸 (H,W)，如果 None 不进行 resize
        """
        self.image_files = image_files
        self.mask_files = mask_files
        self.resize = resize

    def __len__(self):
        """
        返回数据集大小
        """
        return len(self.image_files)

    def __getitem__(self, idx):
        """
        获取单个样本

        参数:
            idx (int): 样本索引

        返回:
            image_resized  (torch.Tensor): resize 后图像
            mask_resized   (torch.Tensor): resize 后 mask
            image_orig     (torch.Tensor): 原始图像
            mask_orig      (torch.Tensor): 原始 mask
            img_path       (str): 图像路径
            mask_path      (str): mask 路径
        """
        # 获取对应路径
        img_path = self.image_files[idx]
        mask_path = self.mask_files[idx]

        # 读取原始图像和 mask
        image_orig, mask_orig = process_path(img_path, mask_path)

        # 如果设置了 resize，则进行 resize
        if self.resize:
            image_resized, mask_resized = preprocess(image_orig, mask_orig, self.resize)
        else:
            image_resized, mask_resized = image_orig, mask_orig

        return image_resized, mask_resized, image_orig, mask_orig, img_path, mask_path

# ===============================
# 5. DataLoader
# ===============================
# 原始大小 Dataset，不进行 resize
image_ds = SegmentationDataset(image_list, mask_list, resize=None)
# resize 后 Dataset，例如用于训练输入固定尺寸
processed_image_ds = SegmentationDataset(image_list, mask_list, resize=(96,128))

# DataLoader 用于批量加载
dataloader_orig = DataLoader(image_ds, batch_size=1, shuffle=False)       # 原始图像每批 1 张
dataloader_resized = DataLoader(processed_image_ds, batch_size=4, shuffle=True)  # resize 后每批 4 张，可打乱顺序
````

## 3 - U-Net

U-Net，因其U形结构而得名，最初于2015年为肿瘤检测而创建，但近年来已成为其他语义分割任务中非常流行的选择。

U-Net 基于之前的架构，称为全卷积网络（Fully Convolutional Network，FCN），该网络用转置卷积层（transposed convolution layer）替换了典型 CNN 中的全连接层，以将特征图上采样回原始输入图像的大小，同时保留空间信息。这是必要的，因为全连接层会破坏空间信息（图像中的“位置”信息），而这在图像分割任务中至关重要。使用转置卷积的额外好处是，输入尺寸不再需要固定，就像使用全连接层时那样。

不幸的是，FCN 的最终特征层由于下采样过多而导致信息丢失。信息丢失太多后，再进行上采样会变得困难，从而产生输出看起来粗糙的结果。

U-Net 在 FCN 的基础上进行了改进，设计上有一定相似性，但在一些关键方面有所不同。网络末端不再使用单一的转置卷积，而是为下采样输入图像到特征图和上采样这些特征图回到原始输入图像大小的每一层使用匹配数量的卷积和转置卷积。它还增加了跳跃连接（skip connections），以保留在编码过程中可能丢失的信息。跳跃连接将信息从编码器中对应的下采样层传递到解码器中的每个上采样层，捕捉更细的信息，同时保持计算量较低。这有助于防止信息丢失以及模型过拟合。


<a name='3-1'></a>
### 3.1 - 模型细节

<img src="images/unet.png" style="width:700px;height:400;">
<caption><center> <u><b> 图2 </u></b>: U-Net 架构<br> </center></caption>

**收缩路径（编码器 / 下采样）**：

U-Net 的收缩路径通过多个卷积块提取特征，同时逐步减小图像的空间尺寸并增加通道数。在 PyTorch 中，每个卷积块使用 `ConvBlock` 实现，其特性如下：

1. 两个 3x3 卷积层（`nn.Conv2d`）+ ReLU 激活  
2. 可选 Dropout（`nn.Dropout2d`）  
3. 可选最大池化（`nn.MaxPool2d`）进行下采样  
4. 返回两个张量：
   - `next_layer`：下一个卷积块的输入（经过 max pooling 时空间尺寸减半）  
   - `skip_connection`：用于 U-Net 跳跃连接的特征图  

收缩路径遵循标准 CNN 架构：卷积层提取特征，ReLU 激活增加非线性，MaxPooling 降低空间分辨率，同时通道数通常随下采样逐步增加。  

**裁剪与跳跃连接**：  
在上采样阶段，需要将收缩路径对应的特征图与扩展路径进行拼接。由于卷积会丢失边界像素，通常会裁剪收缩路径的特征图，使其与扩展路径尺寸匹配，然后再进行通道维拼接（skip connection）。

---

**扩展路径（解码器 / 上采样）**：

扩展路径的任务是将特征图恢复到原始图像尺寸，同时逐步减少通道数。每个上采样步骤包括：

1. 特征图上采样（`nn.ConvTranspose2d` 或 `F.interpolate`）  
2. 通道维减半，空间尺寸加倍  
3. 与收缩路径对应的 skip connection 拼接  
4. 两个 3x3 卷积 + ReLU，进一步提取特征  

**最终输出层**：  
使用 1x1 卷积（`nn.Conv2d`）将最后一层通道映射到类别数，实现语义分割或其他任务所需输出。

> PyTorch 版本的 U-Net 通常有 23 层卷积（包括 Conv2d + ConvTranspose2d + 输出卷积），并通过 skip connection 保留局部信息。

<a name='3-2'></a>
### 3.2 - 编码器（下采样模块）

<img src="images/encoder.png" style="width:500px;height:500;">
<caption><center> <u><b>图3</u></b>: U-Net 编码器详细视图 <br> </center></caption>

编码器由多个 `ConvBlock` 堆叠，每个块的执行流程如下：

1. 输入张量 `[B, C, H, W]`  
2. 经过两次 3x3 卷积 + ReLU 激活  
3. 根据需要应用 Dropout  
4. 保存当前输出为 skip connection  
5. 根据 `max_pooling` 决定是否进行 2x2 最大池化，输出 `next_layer`  

- `next_layer` 用于下一个卷积块  
- `skip_connection` 用于扩展路径拼接  

> 如果 `max_pooling=False`，`next_layer` 与 `skip_connection` 相同，常用于收缩路径最后一级或特征保留。

<a name='ex-1'></a>
### 练习 1 - ConvBlock

在 PyTorch 中实现收缩块（conv_block）的核心思想：

* 添加 2 个 `nn.Conv2d` 层，每层滤波器数量为 `n_filters`，卷积核大小为 3x3，使用 He 初始化（`nn.init.kaiming_normal_`），ReLU 激活。  
* 可选 Dropout 层，使用 `nn.Dropout2d(dropout_prob)`。  
* 可选最大池化层，使用 `nn.MaxPool2d(kernel_size=2, stride=2)` 下采样。  
* 返回 `(next_layer, skip_connection)`，用于构建完整 U-Net 的收缩路径。


````python
# ===============================
# 8. ConvBlock (卷积模块)
# ===============================
class ConvBlock(nn.Module):
    """
    卷积块，用于 U-Net 编码器部分或其他卷积网络
    特性：
    - 两个 3x3 卷积 + ReLU
    - 可选 Dropout
    - 可选 MaxPooling
    - 返回当前层输出 + skip connection（用于 U-Net 的跳跃连接）
    """
    def __init__(self, in_channels, n_filters=32, dropout_prob=0, max_pooling=True):
        """
        初始化卷积块

        参数:
            in_channels   (int): 输入通道数
            n_filters     (int): 卷积层输出通道数（即过滤器数量），默认 32
            dropout_prob  (float): dropout 概率，0 表示不使用 dropout
            max_pooling   (bool): 是否在输出后进行 2x2 最大池化
        """
        super(ConvBlock, self).__init__()

        # ------------------------
        # 卷积层 1：输入 in_channels -> n_filters，卷积核 3x3，padding=1 保持尺寸
        # ------------------------
        self.conv1 = nn.Conv2d(in_channels, n_filters, kernel_size=3, padding=1)

        # 卷积层 2：输入 n_filters -> n_filters，卷积核 3x3，padding=1 保持尺寸
        self.conv2 = nn.Conv2d(n_filters, n_filters, kernel_size=3, padding=1)

        # 使用 He 初始化（适用于 ReLU 激活）
        nn.init.kaiming_normal_(self.conv1.weight)
        nn.init.kaiming_normal_(self.conv2.weight)

        # ReLU 激活函数
        self.relu = nn.ReLU(inplace=True)

        # 判断是否使用 dropout
        self.use_dropout = dropout_prob > 0
        if self.use_dropout:
            # 二维 Dropout，作用于通道维
            self.dropout = nn.Dropout2d(dropout_prob)

        # 判断是否使用最大池化
        self.use_maxpool = max_pooling
        if self.use_maxpool:
            # 2x2 最大池化，stride=2，尺寸减半
            self.maxpool = nn.MaxPool2d(kernel_size=2, stride=2)

    def forward(self, x):
        """
        前向传播

        参数:
            x (torch.Tensor): 输入张量 [B, C, H, W]

        返回:
            next_layer (torch.Tensor): 下一个卷积层输入（可能经过 maxpool）
            skip       (torch.Tensor): skip connection 输出，用于 U-Net 上采样阶段
        """
        # 第一次卷积 + ReLU
        x = self.relu(self.conv1(x))
        # 第二次卷积 + ReLU
        x = self.relu(self.conv2(x))

        # 如果使用 dropout
        if self.use_dropout:
            x = self.dropout(x)

        # skip connection 保存当前输出
        skip = x

        # 如果使用 max pooling，则作为下一个层输入，否则直接输出
        if self.use_maxpool:
            next_layer = self.maxpool(x)
        else:
            next_layer = x

        return next_layer, skip
````

````python
import torch
import torch.nn as nn
from torchsummary import summary

# ===============================
# 说明：ConvBlock 已经定义过，这里直接复用
# ===============================
# class ConvBlock(nn.Module):
#     ...

# ===============================
# 1. 测试 Block 1
# ===============================

# 定义输入图像尺寸 (C,H,W)
input_size = (3, 96, 128)  # 3 通道 RGB 图像，96 高，128 宽
n_filters = 32              # 输出卷积通道数

# 构建 Block 1
# 输入通道数 = 3 (RGB)
# 输出卷积通道数 = 32
# dropout_prob=0 表示不使用 dropout
# max_pooling=True 表示输出后进行 2x2 最大池化
block1 = ConvBlock(in_channels=input_size[0], n_filters=n_filters, dropout_prob=0, max_pooling=True)

# 模拟输入张量
# batch_size=1，输入通道 3，高 96，宽 128
x = torch.randn(1, *input_size)

# 前向传播测试
next_layer, skip_connection = block1(x)

# 打印结果形状
print("Block 1 Test:")
print("Next layer shape:", next_layer.shape)       # next_layer: maxpool 后尺寸应为 [1,32,48,64]
print("Skip connection shape:", skip_connection.shape)  # skip_connection: 卷积输出，尺寸应为 [1,32,96,128]

# 使用 torchsummary 打印详细参数信息
print("\nBlock 1 Summary:")
summary(block1, input_size)

# ===============================
# 2. 测试 Block 2
# ===============================

# 构建 Block 2
# 输入通道 = 3
# 输出卷积通道 = n_filters*32 = 1024
# dropout_prob=0.1 表示使用 10% dropout
# max_pooling=True 表示下采样
block2 = ConvBlock(in_channels=input_size[0], n_filters=n_filters*32, dropout_prob=0.1, max_pooling=True)

# 前向传播
next_layer2, skip_connection2 = block2(x)

# 打印结果形状
print("\nBlock 2 Test:")
print("Next layer shape:", next_layer2.shape)          # 下采样后尺寸
print("Skip connection shape:", skip_connection2.shape) # 卷积输出尺寸

# 打印参数信息
print("\nBlock 2 Summary:")
summary(block2, input_size)
````

<a name='3-3'></a>
### 3.3 - 解码器（上采样模块，PyTorch 版本）

解码器用于将特征图恢复到原始图像尺寸。在每个上采样阶段，需要将对应编码器块的 skip connection 拼接进来，然后再输入到卷积模块以提取融合特征。

<img src="images/decoder.png" style="width:500px;height:500;">
<caption><center> <u><b>图4</u></b>: U-Net 解码器详细视图 <br> </center></caption>

PyTorch 中的 `UpsamplingBlock` 实现了这个思想：

1. **上采样**：
   - 使用 `nn.ConvTranspose2d` 对输入特征图进行上采样，将空间尺寸翻倍。
   - 输出通道数通常与卷积块的滤波器数相同（`n_filters`）。

2. **跳跃连接（skip connection）**：
   - 将上采样输出与对应编码器层的特征图在通道维度拼接（`torch.cat([up, skip], dim=1)`）。
   - 保留局部细节信息，有助于恢复空间精度。

3. **卷积融合**：
   - 拼接后的特征图再经过两个 3x3 卷积层 + ReLU 激活。
   - 卷积参数与编码器相同（He 初始化，padding=1），保证非线性表达能力。

> 通过重复上述步骤，解码器逐步将 Bottleneck 特征图恢复为原始尺寸，同时利用 skip connections 保留关键细节信息，使 U-Net 在分割或重建任务中表现出色。

<a name='ex-2'></a>
### 练习 2 - UpsamplingBlock

在 PyTorch 中实现上采样块（upsampling_block）时，需要：

* 输入：
  - `expansive_input`：上一层解码器输出  
  - `contractive_input`：对应编码器层的 skip connection

* 步骤：
  1. 对 `expansive_input` 使用 `nn.ConvTranspose2d` 上采样：
     - 输出通道数：`n_filters`
     - 卷积核大小：3x3
     - 步幅：2
     - padding 和 output_padding 保证空间尺寸翻倍
  2. 与 `contractive_input` 在通道维拼接：`torch.cat([up, skip], dim=1)`
  3. 经过两个 3x3 卷积 + ReLU 激活（卷积权重使用 He 初始化，padding=1）

> 完整的上采样模块实现了 U-Net 解码器的核心思想：上采样 + skip connection + 卷积融合，使特征图逐步恢复至原始图像尺寸。


````python
# ===============================
# 反卷积上采样模块 (UpsamplingBlock)
# ===============================
class UpsamplingBlock(nn.Module):
    """
    功能：
    1. 将输入特征图上采样（通过反卷积/转置卷积）
    2. 与 encoder 的 skip connection 拼接
    3. 再经过两个卷积+ReLU
    """

    def __init__(self, in_channels, skip_channels, n_filters=32):
        """
        参数：
        - in_channels: 输入特征图的通道数（来自上一层 Decoder 或 Bottleneck）
        - skip_channels: 来自 Encoder 的 skip connection 特征图通道数
        - n_filters: 卷积输出通道数
        """
        super(UpsamplingBlock, self).__init__()

        # --------------------
        # 1. 上采样 (转置卷积)
        # 输入: (B, in_channels, H, W)
        # 输出: (B, n_filters, 2*H, 2*W)
        self.upconv = nn.ConvTranspose2d(
            in_channels,  # 输入通道
            n_filters,    # 输出通道
            kernel_size=3, # 卷积核大小 3x3
            stride=2,      # 步幅 2，进行上采样
            padding=1,     # 边界填充
            output_padding=1 # 确保输出尺寸正好翻倍
        )

        # --------------------
        # 2. 拼接后卷积
        # 拼接后通道 = n_filters + skip_channels
        self.conv1 = nn.Conv2d(
            n_filters + skip_channels,  # 输入通道
            n_filters,                  # 输出通道
            kernel_size=3,
            padding=1
        )

        self.conv2 = nn.Conv2d(
            n_filters,  # 输入通道
            n_filters,  # 输出通道
            kernel_size=3,
            padding=1
        )

        # --------------------
        # 权重初始化
        nn.init.kaiming_normal_(self.upconv.weight)
        nn.init.kaiming_normal_(self.conv1.weight)
        nn.init.kaiming_normal_(self.conv2.weight)

        # 激活函数
        self.relu = nn.ReLU(inplace=True)

    def forward(self, x, skip):
        """
        前向传播：
        - x: 上一层 Decoder 输出的特征图 (B, in_channels, H, W)
        - skip: 对应 Encoder 的 skip connection (B, skip_channels, H*2, W*2)
        """
        # 1. 上采样
        x = self.upconv(x)  # 输出尺寸翻倍 (B, n_filters, H*2, W*2)

        # 2. 与 skip connection 拼接
        # dim=1 对通道进行拼接
        x = torch.cat([x, skip], dim=1)  # 通道数 = n_filters + skip_channels

        # 3. 卷积 + 激活
        x = self.relu(self.conv1(x))
        x = self.relu(self.conv2(x))

        return x  # 输出尺寸: (B, n_filters, H*2, W*2)
````

````python
import torch
import torch.nn as nn
from torchsummary import summary

# ===============================
# 上采样块已定义 (复用之前的 UpsamplingBlock)
# ===============================
# class UpsamplingBlock(nn.Module):
#     ...

# ===============================
# 1. 测试上采样块
# ===============================

n_filters = 32  # 输出通道数，即卷积后特征图的通道数

# -------------------------------
# 模拟输入尺寸 (PyTorch 格式: [C,H,W])
# -------------------------------
input_size1 = (256, 12, 16)  # expansive_input: 来自上一层 decoder 或 bottleneck
input_size2 = (128, 24, 32)  # contractive_input: 对应 encoder 的 skip connection

# -------------------------------
# 2. 构建上采样块
# -------------------------------
block1 = UpsamplingBlock(
    in_channels=input_size1[0],  # 输入通道数
    skip_channels=input_size2[0], # skip connection 的通道数
    n_filters=n_filters           # 输出卷积通道数
)

# -------------------------------
# 3. 模拟输入张量
# -------------------------------
expansive_inputs = torch.randn(1, *input_size1)  # batch_size=1, [B, C, H, W]
contractive_inputs = torch.randn(1, *input_size2)  # batch_size=1, skip connection

# -------------------------------
# 4. 前向传播测试
# -------------------------------
# 输入上采样块后输出
output = block1(expansive_inputs, contractive_inputs)

# 输出形状打印
# 输出尺寸应为 [B, n_filters, H*2, W*2] = [1, 32, 24, 32]
print("Output shape:", output.shape)

# -------------------------------
# 5. 打印参数信息
# -------------------------------
# torchsummary 可以显示每一层的输出尺寸和参数量
# 对于多输入情况，需要用 list/tuple 传入每个输入的尺寸
print("\nUpsampling Block 1 Summary:")
summary(
    block1,
    [(input_size1[0], input_size1[1], input_size1[2]),  # expansive_input 尺寸
     (input_size2[0], input_size2[1], input_size2[2])]  # skip connection 尺寸
)
````

### 3.4 - 构建完整模型（PyTorch 版本）

现在，我们将编码器（Encoder）、瓶颈层（Bottleneck）和解码器（Decoder）组合起来，构建完整的 U-Net 模型。在 PyTorch 中，这对应于 `UNet` 类的定义：

- **输入通道数**：RGB 图像为 3  
- **卷积基数**：`n_filters`，后续卷积块通道数按 2 的倍数增长  
- **输出类别数**：每个像素属于 23 个类别，因此 `n_classes=23`


<a name='ex-3'></a>
### 练习 3 - UNet

实现 `UNet(...)`，完成 U-Net 整体结构：

#### ✅ 编码器（Contracting Path）

- 使用多个 `ConvBlock` 堆叠，每个块包含两个 3x3 卷积 + ReLU；
- 每进入下一层，滤波器数量翻倍；
- 从第四个卷积块开始，加入 `dropout=0.3`；
- 最底层卷积块设置 `dropout=0.3` 且 `max_pooling=False`。

#### 🧠 瓶颈层（Bottleneck）

- 底层卷积块输出作为 Bottleneck 输入，负责最抽象特征提取；
- 滤波器数量维持最高值，捕获全局上下文信息。

#### 🔁 解码器（Expanding Path）

- 每一步上采样：
  - 对上一层输出使用 `ConvTranspose2d` 进行上采样，空间尺寸翻倍；
  - 与对应编码器输出的 skip connection 在通道维拼接；
  - 拼接后再通过两个 3x3 卷积 + ReLU 激活，恢复特征表示；
- 每向上一层，卷积滤波器数量减半，逐步还原图像空间维度。

#### 🎯 输出层（Final Output Layer）

- 上采样完成后，添加一个 3x3 卷积 + ReLU，进一步融合特征；
- 最终使用 1x1 卷积映射到类别数（23），得到每个像素的类别预测。

最终模型结构示意：

输入图像 [B, 3, H, W]

▼

[编码器：多层 ConvBlock + 下采样]

▼

[瓶颈层：特征压缩与抽象]

▼

[解码器：上采样 + skip connections]

▼

[3x3 卷积 + ReLU → 1x1 卷积 → 输出 23 类预测]


通过这种设计，U-Net 能够同时捕捉全局上下文信息（编码器与瓶颈层）和局部空间细节（skip connection），在语义分割或像素级预测任务中表现优异。


````python
# ===============================
# U-Net 网络定义
# ===============================
class UNet(nn.Module):
    def __init__(self, input_channels=3, n_filters=32, n_classes=23):
        """
        UNet 网络初始化
        
        参数：
        - input_channels: 输入图像通道数，RGB 图像为 3
        - n_filters: 卷积基数，后续卷积层通道数按 2 的倍数增长
        - n_classes: 输出分割类别数
        """
        super(UNet, self).__init__()

        # -------------------------------
        # 编码器（下采样）
        # -------------------------------
        # ConvBlock(in_channels, n_filters, dropout_prob=0, max_pooling=True)
        self.c1 = ConvBlock(input_channels, n_filters)              # Block 1
        self.c2 = ConvBlock(n_filters, n_filters*2)                # Block 2
        self.c3 = ConvBlock(n_filters*2, n_filters*4)              # Block 3
        self.c4 = ConvBlock(n_filters*4, n_filters*8, dropout_prob=0.3)  # Block 4 加 dropout
        self.c5 = ConvBlock(n_filters*8, n_filters*16, dropout_prob=0.3, max_pooling=False)  # 底层，不下采样

        # -------------------------------
        # 解码器（上采样 + skip connection）
        # UpsamplingBlock(in_channels, skip_channels, n_filters)
        # in_channels: 输入通道（来自上一层 decoder）
        # skip_channels: 对应 encoder 的 skip connection 通道
        # n_filters: 输出通道
        # -------------------------------
        self.u6 = UpsamplingBlock(n_filters*16, n_filters*8, n_filters*8)
        self.u7 = UpsamplingBlock(n_filters*8, n_filters*4, n_filters*4)
        self.u8 = UpsamplingBlock(n_filters*4, n_filters*2, n_filters*2)
        self.u9 = UpsamplingBlock(n_filters*2, n_filters, n_filters)

        # -------------------------------
        # 最终卷积层：输出预测分割图
        # -------------------------------
        self.conv_final1 = nn.Conv2d(n_filters, n_filters, 3, padding=1)  # 卷积 + ReLU
        self.conv_final2 = nn.Conv2d(n_filters, n_classes, 1)             # 1x1 卷积映射到类别数

        # 初始化权重
        nn.init.kaiming_normal_(self.conv_final1.weight)
        nn.init.kaiming_normal_(self.conv_final2.weight)

        # ReLU 激活
        self.relu = nn.ReLU(inplace=True)

    def forward(self, x):
        """
        前向传播
        
        参数：
        - x: 输入图像张量，尺寸 [B, C, H, W]
        
        返回：
        - out: 输出分割图，尺寸 [B, n_classes, H, W]
        """
        # -------------------------------
        # 编码器（下采样，每个 block 返回 next_layer 和 skip_connection）
        # -------------------------------
        x1, s1 = self.c1(x)  # Block 1
        x2, s2 = self.c2(x1) # Block 2
        x3, s3 = self.c3(x2) # Block 3
        x4, s4 = self.c4(x3) # Block 4
        x5, s5 = self.c5(x4) # Block 5（底层，无下采样）

        # -------------------------------
        # 解码器（上采样 + skip connection）
        # -------------------------------
        x = self.u6(x5, s4)
        x = self.u7(x, s3)
        x = self.u8(x, s2)
        x = self.u9(x, s1)

        # -------------------------------
        # 输出卷积
        # -------------------------------
        x = self.relu(self.conv_final1(x))  # 卷积 + ReLU
        out = self.conv_final2(x)           # 1x1 卷积映射到类别数

        return out
````

````python
import torch
from torchsummary import summary

# ===============================
# 定义输入尺寸
# ===============================
img_height = 96       # 输入图像高度 H
img_width = 128       # 输入图像宽度 W
num_channels = 3      # 输入通道数 C，RGB 图像为 3
n_classes = 23        # 输出类别数（像素级分割类别）
n_filters = 32        # U-Net 基础卷积通道数，后续按倍数增加

# ===============================
# 实例化 PyTorch U-Net
# ===============================
# 输入参数：
# - input_channels: 输入图像通道数
# - n_filters: 基础卷积通道数
# - n_classes: 输出类别数
unet = UNet(input_channels=num_channels, n_filters=n_filters, n_classes=n_classes)

# ===============================
# 将模型移动到 GPU（可选）
# ===============================
# 检查是否有可用 GPU，否则使用 CPU
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
unet.to(device)

# ===============================
# 打印模型 summary
# ===============================
# summary 可以显示每层输出尺寸和参数量
# 输入尺寸：(C, H, W)
summary(unet, (num_channels, img_height, img_width))

# ===============================
# 模拟 comparator（可选）
# ===============================
# 如果你有一个预期输出（例如 TensorFlow 模型对应的输出尺寸或参数数量），
# 可以实现一个对比函数 comparator_pytorch。
# 这里提供一个示例函数框架：
def comparator_pytorch(model_summary, expected_output):
    """
    对比 PyTorch 模型 summary 与预期输出
    
    参数：
    - model_summary: torchsummary.summary 返回的信息
    - expected_output: 预期输出信息（例如 TensorFlow 输出）
    
    功能：
    - 打印提示，提醒用户可以实现逐层对比
    """
    print("PyTorch comparator is not fully implemented, please implement layer-wise checks.")

# 使用示例（注释掉，因为 outputs.unet_model_output 需要自己定义）
# comparator_pytorch(summary(unet, (num_channels, img_height, img_width)), outputs.unet_model_output)
````

### 3.6 - 损失函数（Loss Function）

在语义分割（semantic segmentation）任务中，每个像素都有一个整数标签，对应其类别索引（取值范围 `0` 到 `num_classes - 1`）。  
这与分类问题中常用的 **categorical crossentropy** 不同：分类交叉熵要求标签为 **one-hot 编码**。  

为了处理像素级多类别预测，我们使用 **sparse categorical crossentropy**，它直接接收整数标签，而不需要转换为 one-hot 编码。  

在 PyTorch 中，这对应于 `nn.CrossEntropyLoss`：  

- **输入 logits**：模型输出未经过 softmax 的预测张量，形状 `[B, n_classes, H, W]`  
- **目标 masks**：整数类别索引，形状 `[B, H, W]`  
- **计算方式**：对每个像素计算交叉熵损失，取平均作为最终 loss  

优势：

- 可以高效处理大类别数量；
- 避免 one-hot 编码的额外存储开销；
- 与 TensorFlow 中的 `SparseCategoricalCrossentropy(from_logits=True)` 等效。

示例 PyTorch 定义：

```python
import torch.nn as nn

# 定义损失函数
criterion = nn.CrossEntropyLoss()


````python
import torch
import torch.nn as nn
import torch.optim as optim

# ===============================
# 假设 unet 是之前定义的 PyTorch U-Net 模型
# ===============================
# 判断是否有 GPU 可用，否则使用 CPU
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
unet.to(device)  # 将模型移动到指定设备

# -------------------------------
# 优化器
# -------------------------------
# 使用 Adam 优化器更新模型参数，学习率 lr=1e-3
optimizer = optim.Adam(unet.parameters(), lr=1e-3)

# -------------------------------
# 损失函数
# -------------------------------
# 对应 TensorFlow 的 SparseCategoricalCrossentropy(from_logits=True)
# CrossEntropyLoss 输入 logits (未 softmax) [B, n_classes, H, W]
# 目标 masks 是整数类别索引 [B, H, W]，取值范围 0..n_classes-1
criterion = nn.CrossEntropyLoss()
````

````python
# -------------------------------
# 训练示例
# -------------------------------
# 假设 imgs, masks 是一个 batch 数据
# imgs: [B, 3, 96, 128]，浮点张量，RGB 图像
# masks: [B, 96, 128]，整数张量，每个像素是类别索引
imgs = torch.randn(4, 3, 96, 128).to(device)        # batch_size=4, 随机模拟输入
masks = torch.randint(0, 23, (4, 96, 128)).to(device)  # 随机生成 mask

# -------------------------------
# 前向传播
# -------------------------------
# 输出 logits [B, n_classes, H, W]
outputs = unet(imgs)

# -------------------------------
# 计算损失
# -------------------------------
# CrossEntropyLoss 会自动对 logits 做 softmax 并计算交叉熵
loss = criterion(outputs, masks)

# -------------------------------
# 反向传播
# -------------------------------
optimizer.zero_grad()  # 清空梯度
loss.backward()        # 计算梯度
optimizer.step()       # 更新模型参数

# 打印当前 batch 的损失值
print("Loss:", loss.item())
````

### 3.7 - 数据集处理（Dataset Handling）

在下面，你将定义一个函数，用于同时显示输入图像（input image）及其真实掩码（ground truth）。  
**真实掩码** 是模型训练时的目标输出，表示每个像素的正确分类结果。  
你训练得到的模型输出将尽可能地接近这个真实掩码。


````python
# ===============================
# display 函数
# ===============================
def display(display_list):
    """
    可视化输入图像、真实 Mask 和预测 Mask
    
    参数：
        display_list: list，包含要显示的内容，可为：
            - 输入图像 Tensor 或 numpy 数组
            - 真实 Mask Tensor 或 numpy 数组
            - 预测 Mask Tensor 或 numpy 数组
    """
    # 创建画布，宽高 15x15 英寸
    plt.figure(figsize=(15, 15))
    
    # 给每张子图设置标题，最多三个：Input Image / True Mask / Predicted Mask
    title = ['Input Image', 'True Mask', 'Predicted Mask'][:len(display_list)]
    
    # 遍历 display_list 中的每一项
    for i, item in enumerate(display_list):
        # 创建 1 行 len(display_list) 列的子图，第 i+1 个位置
        plt.subplot(1, len(display_list), i+1)
        plt.title(title[i])  # 设置子图标题

        # 如果是 PyTorch Tensor
        if isinstance(item, torch.Tensor):
            # 对 3 维 Tensor（C,H,W）做维度转换为 H,W,C，方便 matplotlib 显示
            if item.dim() == 3:
                img = item.permute(1,2,0).numpy()
            else:  # 2 维 Tensor（H,W）
                img = item.numpy()
            
            # 如果是整数类型 Tensor（通常表示 mask）
            if item.dtype == torch.long:
                plt.imshow(img, cmap='gray')  # 显示灰度图
            else:
                plt.imshow(img)  # 显示彩色图像
        else:
            # 如果不是 Tensor（numpy 数组或其他格式），直接显示
            plt.imshow(item)
        
        plt.axis('off')  # 关闭坐标轴显示
    
    plt.show()  # 显示整张图
````

````python
# ===============================
# 显示同一组图片的原始 & resize
# ===============================

# 从 dataloader_orig 中取出一个 batch（迭代器 next()）
# 这里 batch_size=1，所以取出的各项都是长度为 1 的 Tensor
image_resized, mask_resized, image_orig, mask_orig, img_path, mask_path = next(iter(dataloader_orig))

# 显示原始图像和原始 mask
# 注意：image_orig[0] 取 batch 中的第 0 张图
#       mask_orig[0] 取对应的 mask
display([image_orig[0], mask_orig[0]])  # 显示原始图像及 mask

# 显示 resize 后的图像和 mask
# 注意：image_resized[0] 取 batch 中第 0 张 resize 图
#       mask_resized[0] 取对应的 resize mask
display([image_resized[0], mask_resized[0]])  # 显示 resize 后的图像及 mask
````

## 4 - 训练模型（Train the Model）


````python
# # ===============================
# # 训练循环示例
# # ===============================

# # 1. 选择设备，支持 GPU 优先
# device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

# # 2. 实例化 U-Net 模型，并移动到选定设备
# unet = UNet().to(device)

# # 3. 定义优化器，这里使用 Adam
# optimizer = optim.Adam(unet.parameters(), lr=1e-3)

# # 4. 定义损失函数，适用于多类别语义分割
# #    CrossEntropyLoss 期望：
# #    - 输入 logits: [B, n_classes, H, W]
# #    - 目标 masks: [B, H, W]，整数类别索引
# criterion = nn.CrossEntropyLoss()

# # 5. 训练轮数
# EPOCHS = 2

# # 6. 开始训练循环
# for epoch in range(EPOCHS):
    
#     # 6.1 遍历 dataloader 中的每个 batch
#     # dataloader_resized 返回：
#     # images_resized: resize 后的图像 batch [B, C, H, W]
#     # masks_resized: resize 后的 mask batch [B, H, W]
#     # 后面的 4 个值为原始图像和路径，这里用 _ 占位
#     for images_resized, masks_resized, _, _, _, _ in dataloader_resized:
        
#         # 6.2 将数据移动到 GPU / CPU
#         images_resized = images_resized.to(device)
#         masks_resized = masks_resized.to(device)

#         # 6.3 梯度清零
#         optimizer.zero_grad()

#         # 6.4 前向传播
#         # 输出 shape: [B, n_classes, H, W]
#         outputs = unet(images_resized)

#         # 6.5 计算损失
#         # CrossEntropyLoss 会自动对 logits 做 softmax
#         loss = criterion(outputs, masks_resized)

#         # 6.6 反向传播
#         loss.backward()

#         # 6.7 更新参数
#         optimizer.step()
    
#     # 6.8 每轮打印最后一个 batch 的损失
#     print(f"Epoch {epoch+1}/{EPOCHS} - Loss: {loss.item()}")
````

````python
from tqdm import tqdm
import matplotlib.pyplot as plt

# ===============================
# 记录训练指标
# ===============================
train_acc_history = []  # 每轮训练准确率
train_loss_history = [] # 每轮训练损失

EPOCHS = 5  # 训练轮数
for epoch in range(EPOCHS):
    correct = 0       # 累积正确像素数
    total = 0         # 累积像素总数
    running_loss = 0  # 累积 batch 损失

    # tqdm 显示 batch 进度
    loop = tqdm(dataloader_resized, desc=f"Epoch {epoch+1}/{EPOCHS}", unit="batch")
    for images_resized, masks_resized, _, _, _, _ in loop:
        # 1. 将数据移动到 GPU/CPU
        images_resized = images_resized.to(device)
        masks_resized = masks_resized.to(device)

        # 2. 梯度清零
        optimizer.zero_grad()

        # 3. 前向传播
        # 输出 shape: [B, n_classes, H, W]
        outputs = unet(images_resized)

        # 4. 计算损失
        loss = criterion(outputs, masks_resized)

        # 5. 反向传播
        loss.backward()

        # 6. 更新参数
        optimizer.step()

        # 7. 累积损失
        running_loss += loss.item()

        # 8. 计算 batch 内预测准确率
        # preds: [B,H,W]，取每个像素类别的最大概率
        preds = torch.argmax(outputs, dim=1)
        correct += (preds == masks_resized).sum().item()  # 累积正确像素数
        total += torch.numel(masks_resized)               # 累积总像素数

        # 9. 更新 tqdm 显示信息
        loop.set_postfix(loss=loss.item(), acc=correct/total)

    # 10. 每轮结束后计算平均准确率与平均损失
    epoch_acc = correct / total
    epoch_loss = running_loss / len(dataloader_resized)
    train_acc_history.append(epoch_acc)
    train_loss_history.append(epoch_loss)

# ===============================
# 绘制训练曲线
# ===============================
plt.figure(figsize=(8,4))
plt.plot(train_acc_history, label="Train Accuracy")  # 绘制每轮训练准确率
plt.plot(train_loss_history, label="Train Loss")     # 绘制每轮训练损失
plt.xlabel("Epoch")
plt.ylabel("Value")
plt.title("Training Accuracy & Loss")
plt.legend()
plt.show()
````

<a name='4-1'></a>
### 4.1 - 创建预测掩码（Create Predicted Masks）

现在，定义一个函数，使用 `torch.argmax` 在类别数量的维度上返回最大值的索引，并将预测结果合并为单张图像：


````python
def create_mask(pred_mask):
    """
    将模型输出的 logits/softmax 概率张量转换为单通道灰度预测 mask

    参数:
        pred_mask: torch.Tensor, shape = [B, n_classes, H, W]
                   模型输出，每个像素对每个类别的预测得分

    返回:
        mask: numpy.ndarray, shape = [H, W]
              每个像素的预测类别索引 (整数)
    """
    # 取每个像素预测概率最大的类别，得到 [B,H,W]
    pred_mask = torch.argmax(pred_mask, dim=1)

    # 只返回第一个样本的 mask，并转为 numpy
    mask = pred_mask[0].cpu().numpy()

    return mask
````

<a name='4-3'></a>
### 4.2 - 显示预测结果（Show Predictions）

接下来，将你的预测掩码与真实掩码以及原始输入图像进行对比：


#### 4.2.1 - 灰度图显示

````python
def show_predictions(model, dataloader, device, num_batches=1):
    """
    显示 dataloader 中前 num_batches 个 batch 的图像、真实 mask 和预测 mask

    参数:
        model: PyTorch 模型，输出为 [B, n_classes, H, W]
        dataloader: PyTorch DataLoader，返回 (images_resized, masks_resized, images_orig, masks_orig, ...)
        device: torch.device, 模型和数据所在设备
        num_batches: int, 显示前多少个 batch 的样本
    """
    model.eval()  # 设置模型为评估模式
    with torch.no_grad():  # 禁用梯度计算，节省内存
        count = 0
        for images_resized, masks_resized, images_orig, masks_orig, _, _ in dataloader:
            images_resized = images_resized.to(device)

            # 前向传播，得到预测 logits
            outputs = model(images_resized)  # [B, n_classes, H, W]

            # 将 logits 转为单通道 mask
            pred_masks = create_mask(outputs)  # [H,W] numpy array

            # 取 batch 中第一张图
            image = images_resized[0].cpu()   # [C,H,W] tensor
            mask = masks_resized[0].cpu()     # [H,W] tensor

            # 使用 display 函数显示 输入图像、真实 mask、预测 mask
            display([image, mask, torch.from_numpy(pred_masks)])

            count += 1
            if count >= num_batches:
                break

    model.train()  # 恢复训练模式，防止影响后续训练

# ===============================
# 8. 使用示例
# ===============================
# 假设 dataloader_resized 已经定义，unet 已经训练好
show_predictions(unet, dataloader_resized, device, num_batches=3)
````

#### 4.2.2 - 彩色图显示

````python
# ===============================
# 显示原始真实 & 预测彩色 mask（输入使用原始图片）
# ===============================
import matplotlib.pyplot as plt
import numpy as np
import torch.nn.functional as F

# 定义颜色映射（假设 n_classes <= 23）
N_CLASSES = 23
COLORS = plt.cm.get_cmap("tab20", N_CLASSES)  # tab20 共 20 种颜色，可重复使用

def convert_to_color_mask(mask, n_classes=N_CLASSES):
    """
    将单通道 mask ([H,W]) 转为彩色 mask ([H,W,3])
    
    参数:
        mask: torch.Tensor 或 numpy array, [H,W] int 类别
        n_classes: 类别数
        
    返回:
        color_mask: [H,W,3] uint8 RGB
    """
    mask = mask.cpu().numpy()  # 转为 numpy
    color_mask = COLORS(mask)[:, :, :3]  # RGBA -> RGB
    color_mask = (color_mask * 255).astype(np.uint8)  # 0~1 -> 0~255
    return color_mask

def create_pred_color_mask(pred_mask, n_classes=N_CLASSES):
    """
    将预测 logits 转为彩色 mask
    
    参数:
        pred_mask: [B, n_classes, H, W] logits
        n_classes: 类别数
        
    返回:
        color_mask: [H,W,3] uint8 RGB
    """
    pred_mask = torch.argmax(pred_mask, dim=1)  # [B,H,W]
    return convert_to_color_mask(pred_mask[0], n_classes)

def show_predictions_with_orig_input(model, dataloader, device, n_classes=N_CLASSES, num_batches=1):
    """
    显示每个 batch 的第一张图：
    1. 原始输入图像
    2. 原始真实彩色 mask
    3. 预测彩色 mask（输出 resize 后，可通过上采样到原始尺寸）
    
    参数:
        model: PyTorch 模型
        dataloader: DataLoader, 返回 (resized_img, resized_mask, orig_img, orig_mask, ...)
        device: torch.device
        n_classes: int, 类别数
        num_batches: int, 显示多少个 batch
    """
    model.eval()  # 评估模式
    with torch.no_grad():  # 禁用梯度
        count = 0
        for images_resized, masks_resized, images_orig, masks_orig, _, _ in dataloader:
            images_resized = images_resized.to(device)
            
            # 前向传播
            outputs = model(images_resized)  # [B,n_classes,H,W]

            # 将预测 mask 上采样到原始尺寸 (nearest 适合分割)
            pred_mask_upsampled = F.interpolate(outputs, size=(images_orig.shape[2], images_orig.shape[3]), mode='nearest')
            pred_color_mask = create_pred_color_mask(pred_mask_upsampled, n_classes)

            # 取 batch 第一张图
            image_orig = images_orig[0].cpu()
            mask_orig = masks_orig[0]  # 原始尺寸
            true_color_mask_orig = convert_to_color_mask(mask_orig, n_classes)

            # 显示三张图
            plt.figure(figsize=(18,6))
            
            plt.subplot(1,3,1)
            plt.title("Input Image (Original)")
            plt.imshow(image_orig.permute(1,2,0).numpy())  # [C,H,W] -> [H,W,C]
            plt.axis("off")

            plt.subplot(1,3,2)
            plt.title("True Mask (Original Size)")
            plt.imshow(true_color_mask_orig)
            plt.axis("off")

            plt.subplot(1,3,3)
            plt.title("Predicted Mask (Upsampled to Original Size)")
            plt.imshow(pred_color_mask)
            plt.axis("off")

            plt.show()

            count += 1
            if count >= num_batches:
                break
    model.train()  # 恢复训练模式


# ===============================
# 使用示例
# ===============================
show_predictions_with_orig_input(unet, dataloader_resized, device, n_classes=N_CLASSES, num_batches=3)
````

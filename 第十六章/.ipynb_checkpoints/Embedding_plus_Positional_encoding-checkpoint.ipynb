## 1 - 位置编码 (Positional Encoding)

以下是您在前一次作业中实现的位置编码公式。该编码使用如下公式：

$$
PE_{(pos, 2i)}= sin\left(\frac{pos}{{10000}^{\frac{2i}{d}}}\right)
$$
<br>
$$
PE_{(pos, 2i+1)}= cos\left(\frac{pos}{{10000}^{\frac{2i}{d}}}\right)
$$

<a name='ex-2'></a>

在自然语言处理任务中，将句子转换为 token 后再输入语言模型是标准做法。每个 token 随后会被转换为固定长度的数值向量，称为嵌入 (embedding)，它能够捕捉单词的语义信息。在 Transformer 架构中，一个位置编码向量会被加到嵌入中，以便在模型中传递位置信息。

单靠数值表示很难直观理解这些向量的含义，但可视化可以帮助我们理解单词的语义和位置相似性。正如你在之前的作业中看到的，当嵌入降到二维并绘图时，语义相似的单词会靠得更近，而不相似的单词会分布得更远。位置编码向量也可以进行类似的练习——句子中相近的单词在笛卡尔平面上应当绘得更近，而句子中距离较远的单词应绘得更远。

在本笔记本中，你将创建一系列词嵌入和位置编码向量的可视化，以便直观理解位置编码如何影响词嵌入，并帮助在 Transformer 架构中传递序列信息。


<a name='1-1'></a>
### 1.1 - 位置编码可视化

下面的代码单元包含了你在 Transformer 作业中实现的 `positional_encoding` 函数。干得漂亮！在本笔记本中，你将基于该函数创建更多可视化。


```python
# ----------------------------
# 位置编码函数（Positional Encoding）
# ----------------------------
def positional_encoding(positions, d_model):
    """
    生成 Transformer 模型中使用的位置编码（Positional Encoding）。
    位置编码的作用是为序列中的每个位置提供唯一的向量表示，使模型能够感知词序。
    
    参数:
        positions (int): 序列最大长度，即需要编码的位置数量
        d_model (int): 词向量或隐藏层维度
    
    返回:
        pos_encoding (Tensor): 形状为 (1, positions, d_model) 的位置编码张量
    """
    
    # ----------------------------
    # 1. 创建位置索引向量 pos
    # ----------------------------
    # torch.arange(positions) 生成从 0 到 positions-1 的整数序列，形状为 (positions,)
    # unsqueeze(1) 在第 1 个维度增加维度，使其变为 (positions, 1)
    # 转为浮点类型 float()，便于后续的数学计算
    pos = torch.arange(positions).unsqueeze(1).float()  
    # 示例：positions=5 -> pos=[[0],[1],[2],[3],[4]]
    
    # ----------------------------
    # 2. 创建维度索引向量 i
    # ----------------------------
    # torch.arange(d_model) 生成从 0 到 d_model-1 的整数序列，形状为 (d_model,)
    # unsqueeze(0) 在第 0 个维度增加维度，使其变为 (1, d_model)
    # 转为浮点类型 float()，便于后续计算
    i = torch.arange(d_model).unsqueeze(0).float()  
    # 示例：d_model=6 -> i=[[0,1,2,3,4,5]]
    
    # ----------------------------
    # 3. 计算角度频率矩阵 angle_rads
    # ----------------------------
    # 对每个位置 pos 和每个维度 i 计算角度值
    # 公式：angle_rads[pos, i] = pos / (10000^(2*(i//2)/d_model))
    # i//2 确保每两个维度共享相同频率（偶数维与奇数维配对）
    angle_rads = pos / torch.pow(10000, (2 * (i//2)) / d_model)  
    # 此时 shape=(positions, d_model)
    
    # ----------------------------
    # 4. 偶数维使用正弦函数编码
    # ----------------------------
    # 角度矩阵 angle_rads 的偶数列进行正弦变换
    # 说明：
    # angle_rads: shape = (positions, d_model)，存储每个位置、每个维度的角度值
    # [:, 0::2]：
    #   - ":" 表示选取所有行（即序列中的所有位置）
    #   - "0::2" 表示从第 0 列开始，每隔 2 列取一次，即偶数维度 (0, 2, 4, ...)
    # torch.sin(angle_rads[:, 0::2])：
    #   - 对选中的偶数维列应用 sin 函数
    #   - 将角度值映射到 [-1, 1] 之间
    # 赋值给 angle_rads[:, 0::2]：
    #   - 将原偶数维的角度值替换为正弦后的数值
    angle_rads[:, 0::2] = torch.sin(angle_rads[:, 0::2])
    
    
    # ----------------------------
    # 5. 奇数维使用余弦函数编码
    # ----------------------------
    # 角度矩阵 angle_rads 的奇数列进行余弦变换
    # 说明：
    # [:, 1::2]：
    #   - ":" 表示选取所有行（即序列中的所有位置）
    #   - "1::2" 表示从第 1 列开始，每隔 2 列取一次，即奇数维度 (1, 3, 5, ...)
    # torch.cos(angle_rads[:, 1::2])：
    #   - 对选中的奇数维列应用 cos 函数
    #   - 将角度值映射到 [-1, 1] 之间
    # 赋值给 angle_rads[:, 1::2]：
    #   - 将原奇数维的角度值替换为余弦后的数值
    angle_rads[:, 1::2] = torch.cos(angle_rads[:, 1::2])

    
    # ----------------------------
    # 6. 增加 batch 维度
    # ----------------------------
    # Transformer 的输入通常为 (batch_size, seq_len, d_model)
    # 这里将 shape 从 (positions, d_model) 扩展为 (1, positions, d_model)
    pos_encoding = angle_rads.unsqueeze(0)
    
    # ----------------------------
    # 7. 返回位置编码
    # ----------------------------
    return pos_encoding  # Tensor，形状为 (1, positions, d_model)
```

将嵌入维度 (embedding dimension) 定义为 100。这个数值必须与词嵌入的维度一致。在 [《Attention is All You Need》](https://arxiv.org/abs/1706.03762) 论文中，嵌入维度根据任务不同，范围从 100 到 1024。作者还根据任务定义了最大序列长度，范围从 40 到 512。这里我们将最大序列长度定义为 100，并将最大单词数定义为 64。


```python
# ----------------------------
# 测试位置编码
# ----------------------------

# 设置词向量维度
EMBEDDING_DIM = 100
# MAX_SEQUENCE_LENGTH 表示序列的最大长度（即位置个数）
MAX_SEQUENCE_LENGTH = 100

# 调用 positional_encoding 函数生成位置编码
# 参数说明：
# MAX_SEQUENCE_LENGTH -> positions: 序列长度，也就是需要为多少个 token 生成位置编码
# EMBEDDING_DIM -> d_model: 每个位置编码的向量维度
# 返回值 pos_encoding: tensor，shape = (1, MAX_SEQUENCE_LENGTH, EMBEDDING_DIM)
# 其中 1 表示 batch 维度，方便后续与实际嵌入向量相加
pos_encoding = positional_encoding(MAX_SEQUENCE_LENGTH, EMBEDDING_DIM)

# 打印位置编码的形状
# 输出结果为 torch.Size([1, 100, 100])
# 表示 1 个序列，长度 100，每个位置有 100 维的编码
print("Positional encoding shape:", pos_encoding.shape)
```

你已经在本作业中创建过这个可视化，但让我们更深入地探讨一下。注意矩阵的一些有趣性质——第一个是每个向量的范数（norm）总是一个常数。无论 `pos` 的值是多少，范数始终保持相同，这里为 7.071068。由此可以得出一个结论：两个位置编码向量的点积（dot product）不会受到向量尺度的影响，这对计算相关性（correlation）具有重要意义。


```python
# ----------------------------
# 查看位置编码向量的范数（L2 Norm）
# ----------------------------

# pos_encoding[0, 34, :] 
# 解释：
# 0 -> batch 维度，取第 1 个序列（因为 pos_encoding 的 shape 是 [1, MAX_SEQUENCE_LENGTH, EMBEDDING_DIM]）
# 34 -> 位置索引，取第 35 个位置的编码（索引从 0 开始）
# : -> 取该位置的全部 embedding 维度向量
# 结果是一个 shape 为 [EMBEDDING_DIM] 的 tensor

# torch.norm(...) 
# 计算该位置编码向量的 L2 范数（Euclidean norm）
# 公式: sqrt(sum(x_i^2))，衡量向量长度

# .item() 
# 将单元素 tensor 转换为 Python 数值，方便打印

print("Norm of position 34:", torch.norm(pos_encoding[0,34,:]).item())
# 输出示例：Norm of position 34: 7.071068
# 说明第 35 个位置的编码向量长度约为 sqrt(EMBEDDING_DIM/2)（sin/cos 编码长度规律）
```

另一个有趣的性质是：两个相隔 `k` 个位置的向量之间的差的范数也是一个常数。如果保持 `k` 不变而改变 `pos`，这个差值大致保持相同。这个性质很重要，因为它表明向量差值不依赖于每个编码的位置，而是取决于编码之间的相对间隔。能够将位置编码表示为彼此的线性函数有助于模型通过关注单词的相对位置来学习。

使用向量编码来反映单词位置的差异是难以实现的，特别是因为向量编码的数值必须保持足够小，以免扭曲单词嵌入（word embeddings）。


```python
# ----------------------------
# 计算两个位置编码向量之间的欧氏距离
# ----------------------------

k = 2  # 定义位置间隔 k，表示计算 pos_i 和 pos_(i+k) 的距离

# pos_encoding[0,70,:] 
# 解释：
# 0 -> batch 维度，取第 1 个序列
# 70 -> 位置索引，取第 71 个位置的编码
# : -> 取该位置的全部 embedding 维度向量

# pos_encoding[0,70+k,:] 
# 取第 73 个位置的编码（位置索引 70+k=72）

# pos_encoding[0,70,:] - pos_encoding[0,70+k,:] 
# 对应元素相减，得到两个位置向量的差向量

# torch.norm(...) 
# 计算差向量的 L2 范数（Euclidean distance），即两个向量在高维空间的距离
# 公式: sqrt(sum((x_i - y_i)^2))

# .item() 
# 将单元素 tensor 转换为 Python 数值，方便打印

print("Distance between position 70 and 72:", 
      torch.norm(pos_encoding[0,70,:]-pos_encoding[0,70+k,:]).item())

# 输出示例：
# Distance between position 70 and 72: 1.2345
# 说明在高维空间中，第 71 个位置与第 73 个位置的编码差异大小
```

你已经观察到位置编码向量的一些有趣性质——接下来，你将创建一些可视化，以直观展示这些性质如何影响编码和嵌入（embedding）之间的关系！

### 1.2 - 比较位置编码

#### 1.2.1 - 相关性（Correlation）

位置编码矩阵可以帮助可视化每个位置的向量是如何唯一的。然而，这些向量如何表示句子中单词的相对位置仍不够直观。为了说明这一点，你将计算每个位置向量之间的相关性。一个成功的位置编码器将产生一个完全对称的矩阵，其中最大值位于主对角线上——位置相近的向量应该具有最高的相关性。按照同样的逻辑，随着离主对角线的距离增大，相关性值应当逐渐减小。


```python
# ----------------------------
# 位置编码相关性矩阵可视化
# ----------------------------

# 1. 计算位置编码的相关性矩阵
# pos_encoding[0] -> 第 1 个序列的全部位置编码，形状 (MAX_SEQUENCE_LENGTH, EMBEDDING_DIM)
# pos_encoding[0].T -> 转置，形状 (EMBEDDING_DIM, MAX_SEQUENCE_LENGTH)
# torch.matmul(A, B) -> 矩阵乘法，结果形状 (MAX_SEQUENCE_LENGTH, MAX_SEQUENCE_LENGTH)
# 每个元素 corr[i,j] 表示位置 i 与位置 j 的向量内积（相关性）
corr = torch.matmul(pos_encoding[0], pos_encoding[0].T)

# 2. 创建绘图画布
# figsize=(6,5) -> 设置图像宽 6 英寸，高 5 英寸
plt.figure(figsize=(6,5))

# 3. 设置图标题
plt.title("Positional encoding correlation")

# 4. 绘制伪彩色图
# corr.numpy() -> 将 tensor 转为 numpy 数组
# cmap='RdBu' -> 红蓝渐变色映射相关性大小，红色为正相关，蓝色为负相关
plt.pcolormesh(corr.numpy(), cmap='RdBu')

# 5. 设置 x 轴和 y 轴标签
plt.xlabel('Position')
plt.ylabel('Position')

# 6. 添加颜色条，显示对应的相关性数值
plt.colorbar()

# 7. 显示绘图
plt.show()
```

#### 1.2.2 - 欧几里得距离（Euclidean distance）

你也可以使用欧几里得距离来比较位置编码向量，而不是相关性。在这种情况下，你的可视化将显示一个矩阵，其主对角线为 0，离主对角线越远的元素值越大。


```python
# ----------------------------
# 位置编码欧几里得距离矩阵可视化
# ----------------------------

# 1. 创建一个全零矩阵，存储位置间的欧几里得距离
# 形状为 (MAX_SEQUENCE_LENGTH, MAX_SEQUENCE_LENGTH)
eu = torch.zeros(MAX_SEQUENCE_LENGTH, MAX_SEQUENCE_LENGTH)

# 2. 双重循环计算每一对位置的欧几里得距离
for a in range(MAX_SEQUENCE_LENGTH):              # 遍历第一个位置 a
    for b in range(a+1, MAX_SEQUENCE_LENGTH):     # 遍历第二个位置 b，避免重复计算
        # pos_encoding[0,a,:] -> 位置 a 的向量 (EMBEDDING_DIM,)
        # pos_encoding[0,b,:] -> 位置 b 的向量 (EMBEDDING_DIM,)
        # torch.norm(A-B) -> 计算向量 A-B 的 L2 范数，即欧几里得距离
        eu[a,b] = torch.norm(pos_encoding[0,a,:] - pos_encoding[0,b,:])
        # 利用对称性，将 b,a 的距离赋值为 a,b
        eu[b,a] = eu[a,b]

# 3. 创建绘图画布
plt.figure(figsize=(6,5))  # 宽 6 英寸，高 5 英寸

# 4. 设置图标题
plt.title("Positional encoding Euclidean distance")

# 5. 绘制伪彩色图
# eu.numpy() -> 将 tensor 转为 numpy 数组
# cmap='RdBu' -> 红蓝渐变色映射距离大小
plt.pcolormesh(eu.numpy(), cmap='RdBu')

# 6. 设置 x 轴和 y 轴标签
plt.xlabel('Position')
plt.ylabel('Position')

# 7. 添加颜色条，显示对应的距离数值
plt.colorbar()

# 8. 显示绘图
plt.show()
```

很好！你可以使用这些可视化来检查你创建的任何位置编码。

### 2.1 - 加载预训练嵌入

为了将预训练词嵌入与之前创建的位置编码相结合，首先从 [GloVe](https://nlp.stanford.edu/projects/glove/) 项目中加载一个预训练嵌入。这里你将使用特征维度为 100 的词向量。


```python
# ----------------------------
# 加载 GloVe 词向量
# ----------------------------

# 1️ 定义词向量维度（GloVe 有 50维、100维、200维、300维等版本）
EMBEDDING_DIM = 100  # 使用 100 维词向量表示每个单词的语义信息

# 2️ 创建一个空字典，用于存储“单词 → 向量”的映射关系
embeddings_index = {}

# 3️ 指定 GloVe 文件所在的目录路径（提前下载 GloVe 文件放入该文件夹）
GLOVE_DIR = "glove"  # 假设当前路径下存在名为 glove 的文件夹

# 4️ 打开 GloVe 文件
# 每一行的格式类似：
# the 0.418 0.24968 -0.41242 ...（共100个浮点数）
# 表示单词 “the” 对应的 100 维向量
with open(os.path.join(GLOVE_DIR, 'glove.6B.100d.txt'), 'r', encoding='utf-8') as f:
    # 5️ 按行遍历整个 GloVe 文件
    for line in f:
        # 6️ 以空格为分隔符拆分一行字符串
        # 示例：["the", "0.418", "0.24968", "-0.41242", ...]
        values = line.split()

        # 7️ 第一列是单词本身
        word = values[0]

        # 8️ 其余 100 个值是该单词的词向量
        # 使用 np.asarray() 转换为 NumPy 数组，元素类型为 float32（节省内存）
        coefs = np.asarray(values[1:], dtype='float32')

        # 9️ 将单词与对应的向量存入字典
        # 示例：embeddings_index["the"] = array([...100个浮点数...])
        embeddings_index[word] = coefs

#   打印加载完成后的统计信息，查看共有多少个单词被成功读取
print('Found %s word vectors.' % len(embeddings_index))

# 1️1️ 验证单词 “hi” 的向量是否存在，并输出其维度
# 若输出结果为 (100,) 表示加载成功
print('d_model of "hi":', embeddings_index['hi'].shape)
```

**注意：**  
该词向量（embedding）由 **400,000 个单词** 组成，并且每个单词的向量维度为 **100 个特征**。


考虑以下仅包含两句话的文本。等一下——这些句子没有实际意义！相反，这些句子是经过设计的，具有以下特点：

* 每句话由一些单词组成，每组单词在语义上有一定的相似性。
* 在第一句话中，相似的词是连续排列的，而在第二句话中，词的顺序是随机的。


```python
texts = ['king queen man woman dog wolf football basketball red green yellow',
         'man queen yellow basketball green dog  woman football  king red wolf']
```

首先，运行下面的代码单元，对原始文本进行分词（tokenization）。暂时不用太担心这个步骤具体做了什么——后续的课程会详细讲解。简单总结如下（理解实验大意即可，不是重点）：

* 数组中的每个值表示该句子中每个单词在词典（`word_index`）中的索引。
* 对于长度小于 `MAX_SEQUENCE_LENGTH` 的序列，会用 0 填充以保证长度一致。

再次强调，后续课程会详细讲解，所以现在不用过于纠结这个步骤。


```python
# ----------------------------
# 文本预处理模块
# ----------------------------

# 1. 手动创建一个简单 tokenizer（分词器）
# -----------------------------------
# 假设 'texts' 是一个字符串列表，例如：
# texts = ["the cat sits", "the dog runs"]
# 我们将所有文本拼接成一个字符串，用空格分词，再去重得到所有单词的集合
word_list = list(set(" ".join(texts).split()))

# 2. 创建 word_index：单词 → 索引 的字典
# -----------------------------------
# enumerate() 为每个单词生成唯一编号
# 因为通常 0 被保留为 “填充符 PAD”，所以这里索引从 1 开始
word_index = {word: i + 1 for i, word in enumerate(word_list)}

# 3. 定义最大词汇量（可选）
# -----------------------------------
# 通常我们会设置一个上限，只保留最常用的前 N 个单词
# 这里设置为 64，表示最多保留 64 个不同单词
MAX_NB_WORDS = 64

# 4. 将每个句子转换为数字序列
# -----------------------------------
# 初始化一个空列表，保存每个句子的索引序列
sequences = []

# 遍历每个文本
for text in texts:
    # 将句子中的每个单词替换为对应编号（word_index[w]）
    seq = [word_index[w] for w in text.split()]
    # 转换为 PyTorch 张量（tensor）
    sequences.append(torch.tensor(seq))

# 5. 对所有句子进行 padding（填充到相同长度）
# -----------------------------------
# 使用 PyTorch 的 pad_sequence 函数
# 参数说明：
#   sequences: 一个张量列表（每个句子是一个张量）
#   batch_first=True: 结果的形状为 (batch_size, sequence_length)
#   padding_value=0: 以 0 作为填充值（表示 PAD）
data = pad_sequence(sequences, batch_first=True, padding_value=0)

# 6. 如果句子长度不足 MAX_SEQUENCE_LENGTH，则进行右侧补零
# -----------------------------------
if data.shape[1] < MAX_SEQUENCE_LENGTH:
    # 计算需要补的零的数量
    padding = torch.zeros(data.shape[0], MAX_SEQUENCE_LENGTH - data.shape[1], dtype=torch.long)
    # 拼接在右侧，使每个句子的长度一致
    data = torch.cat([data, padding], dim=1)
else:
    # 如果句子太长，则截断到最大长度
    data = data[:, :MAX_SEQUENCE_LENGTH]

# 7. 打印填充后的结果
# -----------------------------------
print("Padded sequences shape:", data.shape)  # 输出形状 (batch_size, MAX_SEQUENCE_LENGTH)
print(data)  # 打印实际的填充矩阵
```

为了简化模型，你只需要获取文本中出现的单词的词向量（embedding）。在这个例子中，我们只保留句子中出现的 11 个单词。第一个向量将是全零数组，用于表示所有未知单词（UNK）。


```python
# ----------------------------
# 构建 Embedding 矩阵
# ----------------------------

# 1. 创建一个全零矩阵，用于存放每个单词对应的 GloVe 向量
#    shape = (词汇表大小 + 1, 词向量维度)
#    +1 的原因：词索引是从 1 开始的，索引 0 预留给 <PAD> 填充符
embedding_matrix = np.zeros((len(word_index) + 1, EMBEDDING_DIM))

# 2. 遍历词典 word_index，取出每个单词及其对应索引 i
for word, i in word_index.items():
    
    # 3. 从预加载的 GloVe 词典中获取该单词的向量
    vec = embeddings_index.get(word)
    
    # 4. 如果在 GloVe 中找到了对应向量，则存入 embedding_matrix
    #    未找到的单词保持为全零向量（相当于未知词 UNK）
    if vec is not None:
        embedding_matrix[i] = vec

# 5. 打印生成的 embedding_matrix 的维度信息
#    例如 (5000, 100) 表示共 5000 个词，每个词的向量维度是 100
print("Embedding matrix shape:", embedding_matrix.shape)
```

使用从预训练 GloVe 嵌入中提取的权重来创建一个 Embedding 层。


```python
# ----------------------------
# 构建 PyTorch 的 Embedding 层
# ----------------------------

# 使用预训练好的 embedding_matrix 初始化嵌入层
embedding_layer = nn.Embedding.from_pretrained(
    torch.tensor(embedding_matrix, dtype=torch.float32),  # 参数1：词向量矩阵，转为 torch 张量（float32）
    freeze=True                                           # 参数2：是否冻结参数（True 表示不参与训练）
)
```

```python
# 查看嵌入层的权重形状
print("Embedding weight shape:", embedding_layer.weight.shape)

# 随机查看某个单词索引的向量表示
print("Vector for token id 10:", embedding_layer(torch.tensor([10])))
```

将输入的标记化（tokenized）数据通过刚才创建的 Embedding 层转换为词向量表示。  
检查得到的嵌入矩阵的形状，确保最后一个维度对应每个单词的词向量维度（embedding dimension）。


```python
# ----------------------------
# 将文本序列转换为词向量张量
# ----------------------------

# 使用 embedding_layer 将文本序列（单词索引）映射为词向量
embedding = embedding_layer(data)

# 打印嵌入后的张量形状
print("Embedding shape:", embedding.shape)
```

现在，我们将创建一个函数，用于在二维平面上可视化词向量表示。  
由于 Glove 词向量有 100 个维度，需要使用主成分分析（PCA）将其降到 2 维，以便在平面上绘制。


```python
# ----------------------------
# 可视化函数
# ----------------------------

def plot_words(embedding, sequences, sentence):
    """
    功能：
        将指定句子的词向量通过 PCA 降到 2 维，并在平面上以单词文本标注每个点，便于可视化词向量在句子内的分布。

    参数：
        embedding: torch.Tensor
            - 形状通常为 (batch_size, sequence_length, embedding_dim)
            - 这是通过 embedding_layer(data) 得到的输出
        sequences: list 或类似结构
            - 存放每个句子的 token 索引序列，格式示例：[tensor([1,4,5]), tensor([3,2,1,0]), ...]
            - 本函数使用 sequences[sentence] 来获取第 `sentence` 个句子的索引列表
        sentence: int
            - 指定要可视化的句子在 batch 中的索引（例如 0 表示第一个句子）
    返回：
        None（在 notebook 中弹出一张图）
    """

    # ----------------------------
    # 1. 取出目标句子的词向量并转为 numpy 供 sklearn PCA 使用
    # ----------------------------
    # embedding[sentence, 0:len(sequences[sentence]), :] 的含义：
    # - embedding[sentence] -> 取 batch 中第 `sentence` 个样本，形状 (sequence_length, embedding_dim)
    # - 0:len(sequences[sentence]) -> 只取该句子真实长度的 token（避免取到 padding 部分）
    # - ":" -> 取该位置的所有 embedding 维度
    # .detach() -> 如果 embedding 是 requires_grad 的 tensor，detach() 将其从计算图中分离，避免梯度跟踪
    # .numpy() -> 将 torch.Tensor 转为 numpy.ndarray，因为 sklearn 的 PCA 接受 numpy 输入
    #
    # 注意事项：
    # - 确保 embedding[sentence, 0:len(sequences[sentence]), :] 的第一维与 sequences[sentence] 的长度一致
    # - 如果 sequences[sentence] 包含 torch.tensor 元素，len(sequences[sentence]) 返回 token 数量（正确）
    X_pca_train = PCA(n_components=2).fit_transform(
        embedding[sentence, 0:len(sequences[sentence]), :].detach().numpy()
    )
    # 结果 X_pca_train 的形状为 (num_words_in_sentence, 2)，每一行是某个词在 2D 空间的坐标

    # ----------------------------
    # 2. 创建绘图画布和坐标轴
    # ----------------------------
    # figsize=(12,6)：图像大小（英寸单位），可以根据需要调整
    fig, ax = plt.subplots(figsize=(12,6))

    # ----------------------------
    # 3. 绘制散点图
    # ----------------------------
    # X_pca_train[:,0] 是第一主成分（x 轴）
    # X_pca_train[:,1] 是第二主成分（y 轴）
    ax.scatter(X_pca_train[:,0], X_pca_train[:,1])
    # 默认点颜色和大小可以满足教学演示需求，如需美观可加 color/alpha/s参数

    # ----------------------------
    # 4. 准备单词标签列表（从 word_index 中恢复单词）
    # ----------------------------
    # 假设你之前的 word_index 是一个 {word: index} 的字典，index 从 1 开始
    # words = list(word_index.keys()) 只是将键按某一顺序放入列表
    # 警告：字典的键顺序不一定与 index 的顺序一致（Python 3.7+通常保持插入顺序）
    # 在你的代码中后续用 words[index-1] 获取单词，要求 words 列表的顺序与 word_index 的 index 对应
    # 提示：最好使用 index_to_word 映射（index->word）来避免顺序问题
    words = list(word_index.keys())

    # ----------------------------
    # 5. 在每个点旁标注对应单词
    # ----------------------------
    # enumerate(sequences[sentence])：遍历该句子的 token 索引序列
    # i -> 在句子中的顺序位置（0..N-1）
    # index -> token 的词汇表索引（通常从1开始）
    for i, index in enumerate(sequences[sentence]):
        # words[index-1]：
        # - 由于我们上面构造 words 列表并且 word_index 是从 1 开始索引，
        #   因此使用 index-1 来在 words 列表中对应到正确的单词。
        # - 若你的 word_index 是从 0 开始或你使用 index_to_word，请相应修改为 words[index] 或 index_to_word[index]
        #
        # ax.annotate(text, (x, y))：
        # - 在坐标 (x,y) 处添加文本注释
        # - 这里用词作为文本，位置为 X_pca_train 中对应点的坐标
        ax.annotate(words[index-1], (X_pca_train[i,0], X_pca_train[i,1]))

    # ----------------------------
    # 6. 显示图像
    # ----------------------------
    plt.show()

```

很好！现在可以将每个句子的词嵌入在二维平面上进行可视化。  

每个图应显示句子中不同单词的嵌入位置，通过 PCA 降维后呈现。  
这将帮助你直观理解相似词在嵌入空间中彼此接近的关系。


```python
# -------------------------------------------
# 调用函数 plot_words()，可视化指定句子的词向量分布
# -------------------------------------------

# 调用说明：
# plot_words(embedding, sequences, 0)
# 该函数会将指定句子（由参数 sequences[0] 表示）中的每个单词的嵌入向量（embedding）取出，
# 使用 PCA（主成分分析）将高维向量降到二维平面，
# 并用散点图的形式展示每个词在二维空间中的位置。
# 同时，每个点会标注其对应的单词，帮助我们直观理解词语之间的相似性。

plot_words(
    embedding,  # 参数1：embedding —— 词向量张量，形状通常为 (批量大小, 句子长度, 嵌入维度)。
                 # 这里每个词都被映射为一个 EMBEDDING_DIM 维的向量。
    
    sequences,  # 参数2：sequences —— 句子序列（通常由 tokenizer.texts_to_sequences() 生成）。
                 # 它是一个列表，每个元素是一个句子的词索引序列，例如：
                 # [[4, 15, 23, 7], [5, 19, 2, 13, 9]]。
                 # 因此 sequences[0] 表示第一个句子的索引序列。
    
    0           # 参数3：sentence —— 指定要可视化的句子编号。
                 # 在这里传入 0 表示选取第一个句子。
                 # 函数内部会取出 embedding 中对应句子的词向量，然后绘制出来。
)
```

现在绘制第二个句子的词嵌入图。  

请注意，第二个句子包含与第一个句子相同的词，只是顺序不同。  
通过这个图可以看到，单词的顺序不会影响它们的词向量表示，每个词在嵌入空间中的位置保持不变。


```python
# -------------------------------------------
# 可视化第二个句子的词向量分布
# -------------------------------------------

# 调用 plot_words() 函数，将 sequences[1] 对应的句子词向量进行二维可视化
# 并在散点图上标注每个词，观察词向量在空间中的相对位置

plot_words(
    embedding,  # 参数1：embedding —— 词向量张量，形状为 (批量大小, 句子长度, 嵌入维度)
                 # 包含了每个句子中每个词对应的 GloVe 向量表示
    
    sequences,  # 参数2：sequences —— 句子序列列表，每个元素是一个句子的词索引序列
                 # 例如 sequences[1] 表示第二个句子的索引序列
    
    1           # 参数3：sentence —— 指定要可视化的句子编号
                 # 这里传入 1 表示选取第二个句子
                 # 函数内部会提取 embedding 中对应句子的词向量，
                 # 使用 PCA 降到二维后绘制散点图，并标注词文本
)
```

## 3 - 语义与位置编码结合

接下来，你将把之前计算的原始 GloVe 词向量嵌入与位置编码结合起来。在本练习中，你将使用语义嵌入和位置编码的 1:1 权重比例。


```python
# -------------------------------------------
# 将位置编码加入词向量
# -------------------------------------------

# embedding: 原始词向量张量，形状为 (批量大小, 句子长度, EMBEDDING_DIM)
# pos_encoding: 位置编码张量，形状为 (1, MAX_SEQUENCE_LENGTH, EMBEDDING_DIM)
# 注意这里 pos_encoding 可能比句子长度长，函数会按句子长度截取
# 这里 *1.0 是为了确保类型匹配并生成新的张量，不改变原张量
embedding2 = embedding * 1.0 + pos_encoding[:, :, :] * 1.0
# 结果 embedding2 形状与 embedding 相同，表示词向量与对应位置编码叠加后的向量

# -------------------------------------------
# 可视化第一个句子的词向量 + 位置编码
# -------------------------------------------
plot_words(
    embedding2,  # 叠加了位置编码的词向量张量
    sequences,   # 句子索引序列列表
    0            # 可视化第 0 个句子（第一个句子）
)

# -------------------------------------------
# 可视化第二个句子的词向量 + 位置编码
# -------------------------------------------
plot_words(
    embedding2,  # 叠加了位置编码的词向量张量
    sequences,   # 句子索引序列列表
    1            # 可视化第 1 个句子（第二个句子）
)
```

哇，看这些图之间的巨大差异！与原始图相比，这两幅图都发生了显著变化。注意在第二幅图中（对应句子中相似词没有连续出现的情况），像 `red` 和 `wolf` 这样非常不相似的词反而出现在较近的位置。

现在，你可以尝试不同的相对权重，看看它们如何强烈地影响句子中单词的向量表示。


```python
# -------------------------------------------
# 调整词向量和位置编码的权重
# -------------------------------------------

# 设置词向量权重 W1
W1 = 1  # 可调节词向量的重要性，默认 1

# 设置位置编码权重 W2
W2 = 10  # 可调节位置编码的重要性，默认 10

# 将词向量与位置编码按权重叠加
# embedding: 原始词向量张量，形状 (批量大小, 句子长度, EMBEDDING_DIM)
# pos_encoding: 位置编码张量，形状 (1, MAX_SEQUENCE_LENGTH, EMBEDDING_DIM)
# embedding2: 加权叠加后的张量
embedding2 = embedding * W1 + pos_encoding[:, :, :] * W2

# -------------------------------------------
# 可视化第一个句子的加权词向量 + 位置编码
# -------------------------------------------
plot_words(
    embedding2,  # 加权叠加后的词向量张量
    sequences,   # 句子索引序列列表
    0            # 可视化第 0 个句子（第一个句子）
)

# -------------------------------------------
# 可视化第二个句子的加权词向量 + 位置编码
# -------------------------------------------
plot_words(
    embedding2,  # 加权叠加后的词向量张量
    sequences,   # 句子索引序列列表
    1            # 可视化第 1 个句子（第二个句子）
)

# -------------------------------------------
# 参考句子内容（方便理解可视化结果）
# -------------------------------------------
# 第一个句子:
# 'king queen man woman dog wolf football basketball red green yellow'
# 第二个句子:
# 'man queen yellow basketball green dog woman football king red wolf'
```

如果你将 `W1 = 1` 和 `W2 = 10`，你可以看到单词的排列开始呈现顺时针或逆时针的顺序，具体取决于单词在句子中的位置。在这些参数下，位置编码向量已经主导了嵌入表示。

现在尝试反转权重为 `W1 = 10` 和 `W2 = 1`。观察在这些参数下，绘图更接近原始嵌入可视化，单词位置的变化很少。

在前面的 Transformer 作业中，词嵌入会乘以 `sqrt(EMBEDDING_DIM)`。在本例中，相当于使用 `W1 = sqrt(EMBEDDING_DIM) = 10` 和 `W2 = 1`。

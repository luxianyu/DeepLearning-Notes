è¿™ç§ç»“æ„åŒ–çš„æ•°æ®å¯ä»¥å¸®åŠ©é—®ç­”æ¨¡å‹å­¦ä¹ å¦‚ä½•ï¼š
1. ç†è§£ä¸Šä¸‹æ–‡å†…å®¹ï¼›

æ£€æŸ¥æ•´ä¸ªæ•…äº‹æ•°æ®é›†æ˜¯å¦éƒ½ç¬¦åˆè¿™ç§æ ¼å¼ã€‚


```python
# ----------------------------
# ç»Ÿè®¡è®­ç»ƒé›†ä¸­ä¸åŒç±»å‹çš„æ•…äº‹ç±»å‹
# ----------------------------

type_set = set()  # åˆ›å»ºä¸€ä¸ªç©ºé›†åˆï¼Œç”¨äºå­˜å‚¨å”¯ä¸€çš„æ•…äº‹ç±»å‹

for story in babi_dataset['train']:                  # éå†è®­ç»ƒé›†ä¸­çš„æ¯ä¸€æ¡æ ·æœ¬
    story_type = str(story['story']['type'])         # è·å–å½“å‰æ•…äº‹çš„ 'type' å­—æ®µå¹¶è½¬æ¢ä¸ºå­—ç¬¦ä¸²
    if story_type not in type_set:                   # å¦‚æœè¯¥ç±»å‹è¿˜æœªåœ¨é›†åˆä¸­å‡ºç°
        type_set.add(story_type)                     # å°†å…¶åŠ å…¥é›†åˆ

# type_set ä¸­ç°åœ¨å­˜å‚¨äº†è®­ç»ƒé›†ä¸­æ‰€æœ‰å”¯ä¸€çš„æ•…äº‹ç±»å‹
```

ä¸ºäº†è®©æ•°æ®æ›´æ˜“äºå¤„ç†ï¼Œä½ å°†æŠŠæ•°æ®é›†ä»å­—å…¸ç»“æ„â€œå±•å¹³â€ä¸ºè¡¨æ ¼ç»“æ„ã€‚


```python
# å°†å¤šå±‚åµŒå¥—çš„ dataset å±•å¹³ä¸ºå¹³é“ºå½¢å¼ï¼Œæ–¹ä¾¿åç»­å¤„ç†
flattened_babi = babi_dataset.flatten()
```

```python
# è·å– flatten åè®­ç»ƒé›†çš„ç¬¬ä¸€æ¡æ ·æœ¬
next(iter(flattened_babi['train']))
```

ç°åœ¨ï¼Œè®¿é—®æ‰€éœ€ä¿¡æ¯å˜å¾—å®¹æ˜“å¾—å¤šï¼  
ä½ å¯ä»¥è½»æ¾åœ°ä»æ•…äº‹ä¸­æå–å‡ºç­”æ¡ˆï¼ˆanswerï¼‰ã€é—®é¢˜ï¼ˆquestionï¼‰å’Œäº‹å®ï¼ˆfactsï¼‰ï¼Œ  
å¹¶å°†è¿™äº›äº‹å®åˆå¹¶ä¸ºä¸€ä¸ªä½äºâ€œsentencesâ€åˆ—ä¸‹çš„å•æ¡è®°å½•ã€‚


```python
def get_question_and_facts(story):
    """
    åŠŸèƒ½ï¼š
        ä»ä¸€ä¸ª BABI flatten åçš„æ ·æœ¬ä¸­æå–é—®é¢˜ã€ä¸Šä¸‹æ–‡äº‹å®å¥å’Œç­”æ¡ˆ
    å‚æ•°ï¼š
        story: dict
            flatten åçš„å•æ¡ BABI æ ·æœ¬å­—å…¸
            å…¸å‹å­—æ®µåŒ…æ‹¬ï¼š
                - 'story.text': æ‰€æœ‰æ•…äº‹å¥å­çš„åˆ—è¡¨
                - 'story.answer': æ‰€æœ‰å¯¹åº”ç­”æ¡ˆçš„åˆ—è¡¨
    è¿”å›ï¼š
        dic: dict
            åŒ…å«ä¸‰ä¸ªé”®ï¼š
                - 'question' : å½“å‰é—®é¢˜å¥å­
                - 'sentences': ä¸Šä¸‹æ–‡äº‹å®å¥å­ï¼ˆæ‹¼æ¥æˆå­—ç¬¦ä¸²ï¼‰
                - 'answer'   : é—®é¢˜å¯¹åº”çš„ç­”æ¡ˆ
    """

    dic = {}  # åˆ›å»ºä¸€ä¸ªç©ºå­—å…¸ï¼Œç”¨äºå­˜å‚¨ question, sentences, answer

    # æå–ç¬¬3æ¡æ•…äº‹å¥ä½œä¸ºé—®é¢˜ï¼ˆç´¢å¼• 2ï¼‰ï¼Œèµ‹å€¼ç»™å­—å…¸çš„ 'question' é”®
    dic['question'] = story['story.text'][2]

    # å°†å‰ä¸¤æ¡æ•…äº‹å¥ï¼ˆç´¢å¼• 0 å’Œ 1ï¼‰æ‹¼æ¥æˆå­—ç¬¦ä¸²ï¼Œä½œä¸ºä¸Šä¸‹æ–‡äº‹å®å¥
    dic['sentences'] = ' '.join([story['story.text'][0], story['story.text'][1]])

    # æå–ç¬¬3æ¡æ•…äº‹çš„ç­”æ¡ˆï¼ˆç´¢å¼• 2ï¼‰ï¼Œèµ‹å€¼ç»™å­—å…¸çš„ 'answer' é”®
    dic['answer'] = story['story.answer'][2]

    return dic  # è¿”å›åŒ…å« questionã€sentences å’Œ answer çš„å­—å…¸
```

```python
# ---------------------------------------------
# ä½¿ç”¨ map æ–¹æ³•å¯¹ Flattened BABI æ•°æ®é›†è¿›è¡Œå¤„ç†
# ---------------------------------------------

# å¯¹ flattened_babi æ•°æ®é›†åº”ç”¨ get_question_and_facts å‡½æ•°
# processed å°†å­˜å‚¨å¤„ç†åçš„æ•°æ®é›†
# å‚æ•°è¯´æ˜ï¼š
# - flattened_babi: å·²ç»é€šè¿‡ flatten() å±•å¹³çš„ BABI æ•°æ®é›†
# - map(): datasets åº“ä¸­çš„æ–¹æ³•ï¼Œå¯¹æ•°æ®é›†çš„æ¯ä¸€æ¡è®°å½•åº”ç”¨æŒ‡å®šå‡½æ•°
# - get_question_and_facts: è‡ªå®šä¹‰å‡½æ•°ï¼Œå°†æ¯æ¡è®°å½•è½¬æ¢ä¸ºåŒ…å« 'question', 'sentences', 'answer' çš„å­—å…¸

processed = flattened_babi.map(get_question_and_facts)

# å¤„ç†åï¼Œprocessed æ•°æ®é›†æ¯æ¡è®°å½•åŒ…å«ï¼š
# - question: é—®é¢˜æ–‡æœ¬
# - sentences: ä¸é—®é¢˜ç›¸å…³çš„ä¸Šä¸‹æ–‡å¥å­ï¼Œæ‹¼æ¥æˆä¸€ä¸ªå­—ç¬¦ä¸²
# - answer: é—®é¢˜çš„ç­”æ¡ˆ
```

```python
# ---------------------------------------------
# æŸ¥çœ‹å¤„ç†åçš„è®­ç»ƒé›†ä¸­çš„ç¬¬ä¸‰æ¡è®°å½•
# ---------------------------------------------

# processed['train']ï¼šå¤„ç†åçš„è®­ç»ƒé›†
# [2]ï¼šå–è®­ç»ƒé›†ä¸­çš„ç¬¬ 3 æ¡è®°å½•ï¼ˆç´¢å¼•ä» 0 å¼€å§‹ï¼‰
# è¾“å‡ºå°†æ˜¾ç¤ºä¸€ä¸ªå­—å…¸ï¼ŒåŒ…å« 'question', 'sentences', 'answer' ä¸‰ä¸ªé”®åŠå¯¹åº”çš„å€¼
processed['train'][2]
```

```python
# ---------------------------------------------
# æŸ¥çœ‹å¤„ç†åçš„æµ‹è¯•é›†ä¸­çš„ç¬¬ä¸‰æ¡è®°å½•
# ---------------------------------------------

# processed['test']ï¼šå¤„ç†åçš„æµ‹è¯•é›†
# [2]ï¼šå–æµ‹è¯•é›†ä¸­çš„ç¬¬ 3 æ¡è®°å½•ï¼ˆç´¢å¼•ä» 0 å¼€å§‹ï¼‰
# è¾“å‡ºå°†æ˜¾ç¤ºä¸€ä¸ªå­—å…¸ï¼ŒåŒ…å« 'question', 'sentences', 'answer' ä¸‰ä¸ªé”®åŠå¯¹åº”çš„å€¼
processed['test'][2]
```

æŠ½å–å¼é—®ç­”ï¼ˆextractive QAï¼‰çš„ç›®æ ‡æ˜¯ï¼š**æ‰¾åˆ°æ–‡æœ¬ä¸­åŒ…å«é—®é¢˜ç­”æ¡ˆçš„å…·ä½“éƒ¨åˆ†**ã€‚  
æ¢å¥è¯è¯´ï¼Œæ¨¡å‹éœ€è¦åœ¨ä¸Šä¸‹æ–‡ä¸­å®šä½å‡ºç­”æ¡ˆæ‰€åœ¨çš„å­—ç¬¦ä¸²ä½ç½®ã€‚

ä¾‹å¦‚ï¼š  
å¦‚æœæŸä¸ªé—®é¢˜çš„ç­”æ¡ˆæ˜¯ **â€œSeptemberâ€**ï¼Œè€Œä¸Šä¸‹æ–‡å¥å­æ˜¯  
**â€œJane visits Africa in September.â€**ï¼Œ  
é‚£ä¹ˆæˆ‘ä»¬å°±éœ€è¦æ‰¾åˆ°â€œSeptemberâ€åœ¨è¿™å¥è¯ä¸­çš„**èµ·å§‹ç´¢å¼•ï¼ˆstart indexï¼‰**å’Œ**ç»“æŸç´¢å¼•ï¼ˆend indexï¼‰**ã€‚

ä¸‹é¢çš„å‡½æ•°å°†ç”¨äºåœ¨ä½ çš„æ•°æ®é›†ä¸­ï¼Œä¸ºæ¯ä¸ªæ•…äº‹æ‰¾åˆ°ç­”æ¡ˆçš„èµ·å§‹ä¸ç»“æŸå­—ç¬¦ä¸²ç´¢å¼•ã€‚


```python
# -------------------------------------------------------------
# å‡½æ•°åŠŸèƒ½ï¼šè·å–ç­”æ¡ˆåœ¨å¥å­ä¸­çš„èµ·æ­¢ç´¢å¼•
# -------------------------------------------------------------
def get_start_end_idx(story):
    """
    å‚æ•°:
    - story: ä¸€ä¸ªå­—å…¸ï¼ŒåŒ…å«é”® 'sentences' å’Œ 'answer'
      * 'sentences' æ˜¯ç”±å¤šä¸ªå¥å­ç»„æˆçš„ä¸Šä¸‹æ–‡æ–‡æœ¬
      * 'answer' æ˜¯é—®é¢˜çš„ç­”æ¡ˆå­—ç¬¦ä¸²

    è¿”å›:
    - ä¸€ä¸ªå­—å…¸ï¼ŒåŒ…å«ç­”æ¡ˆåœ¨å¥å­ä¸­çš„èµ·å§‹ç´¢å¼•å’Œç»“æŸç´¢å¼•
      * 'str_idx': ç­”æ¡ˆåœ¨å¥å­ä¸­çš„èµ·å§‹ä½ç½®ï¼ˆç¬¬ä¸€ä¸ªå­—ç¬¦ç´¢å¼•ï¼‰
      * 'end_idx': ç­”æ¡ˆåœ¨å¥å­ä¸­çš„ç»“æŸä½ç½®ï¼ˆä¸åŒ…å«è¯¥ç´¢å¼•ï¼‰
    """
    
    # ä½¿ç”¨ str.find() æŸ¥æ‰¾ç­”æ¡ˆå­—ç¬¦ä¸²åœ¨å¥å­ä¸­çš„èµ·å§‹ç´¢å¼•
    str_idx = story['sentences'].find(story['answer'])
    
    # ç»“æŸç´¢å¼• = èµ·å§‹ç´¢å¼• + ç­”æ¡ˆå­—ç¬¦ä¸²é•¿åº¦
    end_idx = str_idx + len(story['answer'])
    
    # è¿”å›å­—å…¸
    return {
        'str_idx': str_idx,
        'end_idx': end_idx
    }
```

```python
# -------------------------------------------------------------
# å°†èµ·æ­¢ç´¢å¼•æ˜ å°„åˆ°æ•´ä¸ªæ•°æ®é›†
# -------------------------------------------------------------
# ä½¿ç”¨ map å‡½æ•°ï¼Œå°† get_start_end_idx å‡½æ•°åº”ç”¨åˆ° processed æ•°æ®é›†çš„æ¯ä¸€ä¸ªæ ·æœ¬
# processed æ•°æ®é›†åŒ…å«é”® 'sentences' å’Œ 'answer'
# map å‡½æ•°ä¼šéå†æ¯ä¸€æ¡æ ·æœ¬ï¼Œå¹¶å°†è¿”å›çš„å­—å…¸ {'str_idx', 'end_idx'} æ·»åŠ åˆ°æ¯æ¡æ ·æœ¬ä¸­
processed = processed.map(get_start_end_idx)
```

```python
# -------------------------------------------------------------
# æŸ¥çœ‹æµ‹è¯•é›†ä¸­ç¬¬ num æ¡æ ·æœ¬çš„ç­”æ¡ˆåŠå…¶åœ¨å¥å­ä¸­çš„ä½ç½®
# -------------------------------------------------------------

num = 187  # è®¾ç½®è¦æŸ¥çœ‹çš„æ ·æœ¬ç´¢å¼•

# æ‰“å°è¯¥æ¡æ ·æœ¬çš„å®Œæ•´ä¿¡æ¯ï¼ˆåŒ…å« 'sentences', 'answer', 'str_idx', 'end_idx' ç­‰é”®ï¼‰
print(processed['test'][num])

# è·å–ç­”æ¡ˆåœ¨å¥å­ä¸­çš„èµ·å§‹å’Œç»“æŸç´¢å¼•
start_idx = processed['test'][num]['str_idx']
end_idx = processed['test'][num]['end_idx']

# æ‰“å°åŸå¥ä¸­å¯¹åº”çš„ç­”æ¡ˆæ–‡æœ¬
print('answer:', processed['test'][num]['sentences'][start_idx:end_idx])
```

ç°åœ¨ï¼Œä½ å·²ç»æ‹¥æœ‰äº†è®­ç»ƒ Transformer æ¨¡å‹è¿›è¡Œé—®ç­”ï¼ˆQuestion Answeringï¼‰æ‰€éœ€çš„æ‰€æœ‰æ•°æ®ï¼  
æ¥ä¸‹æ¥ï¼Œä½ å°†è¿›è¡Œä¸€ä¸ªåœ¨å‘½åå®ä½“è¯†åˆ«å®éªŒä¸­å¯èƒ½å·²ç»é‡åˆ°çš„ä»»åŠ¡ â€”â€” **å¯¹è¾“å…¥è¿›è¡Œåˆ†è¯å¹¶è¿›è¡Œå¯¹é½**ã€‚

ä¸ºäº†å°†æ–‡æœ¬æ•°æ®è¾“å…¥åˆ° Transformer æ¨¡å‹ä¸­ï¼Œä½ éœ€è¦ä½¿ç”¨ [ğŸ¤— Transformer åˆ†è¯å™¨](https://huggingface.co/transformers/main_classes/tokenizer.html) å¯¹è¾“å…¥è¿›è¡Œåˆ†è¯ã€‚  
éå¸¸å…³é”®çš„ä¸€ç‚¹æ˜¯ï¼šä½ ä½¿ç”¨çš„åˆ†è¯å™¨å¿…é¡»ä¸æ‰€ç”¨çš„ Transformer æ¨¡å‹ç±»å‹åŒ¹é…ï¼

åœ¨æœ¬ç»ƒä¹ ä¸­ï¼Œä½ å°†ä½¿ç”¨ ğŸ¤— [DistilBERT å¿«é€Ÿåˆ†è¯å™¨ï¼ˆfast tokenizerï¼‰](https://huggingface.co/transformers/model_doc/distilbert.html)ï¼Œ  
å®ƒä¼šå°†åºåˆ—é•¿åº¦æ ‡å‡†åŒ–ä¸º 512ï¼Œå¹¶ä½¿ç”¨é›¶è¿›è¡Œå¡«å……ã€‚


Transformer æ¨¡å‹é€šå¸¸ä½¿ç”¨å¯ä»¥å°†å•è¯æ‹†åˆ†ä¸ºå­è¯ï¼ˆsubwordsï¼‰çš„åˆ†è¯å™¨è¿›è¡Œè®­ç»ƒã€‚ä¾‹å¦‚ï¼Œå•è¯ "Africa" å¯èƒ½ä¼šè¢«æ‹†åˆ†æˆå¤šä¸ªå­è¯ã€‚  

è¿™å¯èƒ½ä¼šå¯¼è‡´æ•°æ®é›†ä¸­æ ‡ç­¾åˆ—è¡¨ä¸åˆ†è¯å™¨ç”Ÿæˆçš„æ ‡ç­¾åˆ—è¡¨ä¹‹é—´å‡ºç°**é”™ä½**ï¼Œå› ä¸ºåˆ†è¯å™¨å¯èƒ½å°†ä¸€ä¸ªå•è¯æ‹†åˆ†æˆå¤šä¸ªï¼Œæˆ–è€…æ·»åŠ ç‰¹æ®Šæ ‡è®°ã€‚  

åœ¨å¤„ç†ä¹‹å‰ï¼Œ**å¿…é¡»ä½¿ç”¨ `tokenize_and_align()` å‡½æ•°**å°†ç­”æ¡ˆå•è¯çš„èµ·å§‹å’Œç»“æŸç´¢å¼•ä¸å…¶å¯¹åº”çš„å­è¯ token å¯¹é½ã€‚  

åœ¨è¿™ä¸ªä»»åŠ¡ä¸­ï¼Œç”±äºä½ å…³å¿ƒçš„æ˜¯ç­”æ¡ˆçš„èµ·å§‹å’Œç»“æŸç´¢å¼•ï¼Œå› æ­¤éœ€è¦å°†å¥å­ä¸­çš„ç´¢å¼•ä¸å•è¯å¯¹åº”çš„ token ç´¢å¼•è¿›è¡Œå¯¹é½ã€‚


```python
# -------------------------------------------------------------
# åŠ è½½å·²ç»ä¿å­˜å¥½çš„ DistilBERT åˆ†è¯å™¨
# -------------------------------------------------------------

from transformers import DistilBertTokenizerFast  # å¯¼å…¥ HuggingFace Transformers åº“ä¸­çš„ DistilBertTokenizerFast ç±»

# ä»æœ¬åœ°æ–‡ä»¶å¤¹ 'tokenizer/' åŠ è½½åˆ†è¯å™¨
tokenizer = DistilBertTokenizerFast.from_pretrained('tokenizer/')
```

```python
# -------------------------------------------------------------
# å®šä¹‰åˆ†è¯ä¸ç­”æ¡ˆä½ç½®å¯¹é½å‡½æ•°
# -------------------------------------------------------------
def tokenize_align(example):
    """
    å°† BABI æ•°æ®é›†ä¸­çš„æ–‡æœ¬å’Œé—®é¢˜è¿›è¡Œåˆ†è¯ï¼Œå¹¶å°†ç­”æ¡ˆåœ¨æ–‡æœ¬ä¸­çš„å­—ç¬¦ä½ç½®
    è½¬æ¢ä¸º token çº§åˆ«çš„èµ·å§‹å’Œç»“æŸä½ç½®ï¼ˆç”¨äºè®­ç»ƒæŠ½å–å¼é—®ç­”æ¨¡å‹ï¼‰
    
    å‚æ•°ï¼š
    - example: æ•°æ®é›†ä¸­å•æ¡æ ·æœ¬ï¼Œå­—å…¸å½¢å¼ï¼ŒåŒ…å« 'sentences', 'question', 'str_idx', 'end_idx'
    
    è¿”å›ï¼š
    - å­—å…¸ï¼ŒåŒ…å« input_ids, attention_mask, start_positions, end_positions
    """

    # 1. å°†æ–‡æœ¬å’Œé—®é¢˜åŒæ—¶è¿›è¡Œç¼–ç ï¼ˆåˆ†è¯ï¼‰
    encoding = tokenizer(
        example['sentences'],       # ä¸Šä¸‹æ–‡å¥å­
        example['question'],        # é—®é¢˜æ–‡æœ¬
        truncation=True,            # è¶…é•¿æˆªæ–­
        padding=True,               # è¡¥é½åˆ° batch å†…æœ€é•¿é•¿åº¦
        max_length=tokenizer.model_max_length  # æœ€å¤§é•¿åº¦é™åˆ¶
    )

    # 2. å°†ç­”æ¡ˆå­—ç¬¦ä½ç½®æ˜ å°„ä¸º token ç´¢å¼•
    start_positions = encoding.char_to_token(example['str_idx'])      # ç­”æ¡ˆèµ·å§‹ token
    end_positions = encoding.char_to_token(example['end_idx']-1)      # ç­”æ¡ˆç»“æŸ token (åŒ…å«æœ€åä¸€ä¸ªå­—ç¬¦)

    # 3. å¦‚æœç­”æ¡ˆå­—ç¬¦æ˜ å°„ä¸åˆ° tokenï¼Œåˆ™è®¾ç½®ä¸ºæœ€å¤§é•¿åº¦
    if start_positions is None:
        start_positions = tokenizer.model_max_length
    if end_positions is None:
        end_positions = tokenizer.model_max_length

    # 4. è¿”å›ç¼–ç ç»“æœä»¥åŠç­”æ¡ˆçš„ token çº§ä½ç½®
    return {
        'input_ids': encoding['input_ids'],           # token ID åˆ—è¡¨
        'attention_mask': encoding['attention_mask'], # attention mask åˆ—è¡¨
        'start_positions': start_positions,           # ç­”æ¡ˆèµ·å§‹ token ç´¢å¼•
        'end_positions': end_positions                # ç­”æ¡ˆç»“æŸ token ç´¢å¼•
    }
```

```python
# -------------------------------------------------------------
# å°†æ•´ä¸ªæ•°æ®é›†åº”ç”¨ tokenize_align å‡½æ•°
# -------------------------------------------------------------
qa_dataset = processed.map(tokenize_align)

"""
è§£é‡Šï¼š
1. qa_datasetï¼šæ–°çš„æ•°æ®é›†å¯¹è±¡ï¼Œæ¯æ¡æ ·æœ¬å·²ç»è¢«åˆ†è¯å¹¶ä¸”ç­”æ¡ˆä½ç½®å¯¹é½åˆ° tokenã€‚
2. processed.map(tokenize_align)ï¼šå¯¹ processed æ•°æ®é›†ä¸­çš„æ¯æ¡æ ·æœ¬è°ƒç”¨ tokenize_align å‡½æ•°ï¼Œ
   è¿”å›æ–°çš„æ•°æ®é›†ï¼Œå¹¶å°†æ¯æ¡æ ·æœ¬å¢åŠ ä»¥ä¸‹å­—æ®µï¼š
   - input_idsï¼šåˆ†è¯åçš„ token ID åˆ—è¡¨
   - attention_maskï¼šattention mask åˆ—è¡¨
   - start_positionsï¼šç­”æ¡ˆèµ·å§‹ token ç´¢å¼•
   - end_positionsï¼šç­”æ¡ˆç»“æŸ token ç´¢å¼•
3. è¯¥æ•°æ®é›†å¯ä»¥ç›´æ¥ç”¨äº HuggingFace çš„ Trainer æˆ–è‡ªå®šä¹‰ PyTorch DataLoader è®­ç»ƒé—®ç­”æ¨¡å‹ã€‚
"""
```

```python
# -------------------------------------------------------------
# åˆ é™¤ä¸éœ€è¦çš„åŸå§‹åˆ—ï¼Œåªä¿ç•™è®­ç»ƒæ‰€éœ€å­—æ®µ
# -------------------------------------------------------------
qa_dataset = qa_dataset.remove_columns([
    'story.answer',         # åŸå§‹ç­”æ¡ˆæ–‡æœ¬
    'story.id',             # åŸå§‹æ•…äº‹ ID
    'story.supporting_ids', # æ”¯æ’‘å¥å­ ID
    'story.text',           # åŸå§‹æ•…äº‹æ–‡æœ¬
    'story.type'            # é—®é¢˜ç±»å‹
])

"""
è§£é‡Šï¼š
1. qa_dataset.remove_columns([...])ï¼šåˆ é™¤åŸå§‹æ•°æ®é›†ä¸­ä¸å†éœ€è¦çš„åˆ—ï¼Œå‡å°‘å†…å­˜å ç”¨ã€‚
2. åˆ é™¤çš„åˆ—åŒ…æ‹¬ï¼š
   - 'story.answer'ï¼šåŸå§‹ç­”æ¡ˆå­—ç¬¦ä¸²ï¼ˆå·²å¯¹é½åˆ° start/end tokenï¼Œä¸å†éœ€è¦ï¼‰
   - 'story.id'ï¼šæ¯ä¸ªæ•…äº‹çš„å”¯ä¸€æ ‡è¯†
   - 'story.supporting_ids'ï¼šæ”¯æ’‘å¥å­ç´¢å¼•
   - 'story.text'ï¼šå®Œæ•´æ•…äº‹æ–‡æœ¬
   - 'story.type'ï¼šé—®é¢˜ç±»å‹
3. ä¿ç•™çš„å­—æ®µï¼š
   - input_idsï¼šåˆ†è¯åçš„ token ID
   - attention_maskï¼šattention mask
   - start_positionsï¼šç­”æ¡ˆèµ·å§‹ token ç´¢å¼•
   - end_positionsï¼šç­”æ¡ˆç»“æŸ token ç´¢å¼•
   è¿™äº›å­—æ®µå³å¯ç›´æ¥ç”¨äºé—®ç­”æ¨¡å‹è®­ç»ƒã€‚
"""
```

ç°åœ¨ä½ å·²ç»å®Œæˆäº†æ•°æ®çš„åˆ†è¯å’Œå¯¹é½ï¼Œå¯ä»¥å°†å…¶è¾“å…¥åˆ°é¢„è®­ç»ƒçš„ ğŸ¤— Transformer æ¨¡å‹ä¸­ï¼  
ä½ å°†ä½¿ç”¨ä¸€ä¸ª DistilBERT æ¨¡å‹ï¼Œå®ƒä¸ä¹‹å‰ç”¨äºé¢„å¤„ç†æ•°æ®çš„åˆ†è¯å™¨ç›¸åŒ¹é…ã€‚


```python
# ----------------------------
# è®­ç»ƒé›†ä¸æµ‹è¯•é›†åˆ†ç¦»
# ----------------------------

# å°†å¤„ç†å¥½çš„ QA æ•°æ®é›†æ‹†åˆ†ä¸ºè®­ç»ƒé›†å’Œæµ‹è¯•é›†
train_ds = qa_dataset['train']   # è®­ç»ƒé›†
test_ds = qa_dataset['test']     # æµ‹è¯•é›†

"""
è§£é‡Šï¼š
- qa_dataset æ˜¯ç»è¿‡ tokenize_align å’Œåˆ—åˆ é™¤åçš„å®Œæ•´æ•°æ®é›†ã€‚
- ä½¿ç”¨ ['train'] å’Œ ['test'] åˆ†åˆ«è·å–è®­ç»ƒé›†å’Œæµ‹è¯•é›†å­é›†ã€‚
- train_ds ä¸ test_ds å¯ç›´æ¥ç”¨äº DataLoader æˆ– Trainer è¿›è¡Œæ¨¡å‹è®­ç»ƒä¸è¯„ä¼°ã€‚
"""
```

## 2.1 PyTorch å®ç°

[PyTorch](https://pytorch.org/) æ˜¯ç”± Facebook AI Research å®éªŒå®¤å¼€å‘çš„å¼€æºæœºå™¨å­¦ä¹ æ¡†æ¶ï¼Œå¯ç”¨äºè®¡ç®—æœºè§†è§‰å’Œè‡ªç„¶è¯­è¨€å¤„ç†ä»»åŠ¡ã€‚  
å¦‚ä½ æ‰€æ–™ï¼Œå®ƒä¸ bAbI æ•°æ®é›†éå¸¸å…¼å®¹ã€‚


#### è®­ç»ƒé›†ä¸æµ‹è¯•é›†

ç°åœ¨å¯ä»¥å°è¯•é€šè¿‡å¯¼å…¥ PyTorch æ¥åˆ›å»ºè®­ç»ƒé›†å’Œæµ‹è¯•é›†ã€‚


```python
# ----------------------------
# è®¾å®šæ•°æ®é›†æ ¼å¼ä¸º PyTorch tensor å¹¶é€‰æ‹©è¦è¿”å›çš„åˆ—
# ----------------------------

from torch.utils.data import DataLoader

# éœ€è¦è¿”å›çš„åˆ—å
columns_to_return = ['input_ids','attention_mask', 'start_positions', 'end_positions']

# å°†è®­ç»ƒé›†è®¾ç½®ä¸º PyTorch tensor æ ¼å¼
train_ds.set_format(type='pt', columns=columns_to_return)

# å°†æµ‹è¯•é›†è®¾ç½®ä¸º PyTorch tensor æ ¼å¼
test_ds.set_format(type='pt', columns=columns_to_return)

"""
è§£é‡Šï¼š
- set_format(type='pt', columns=...) å°† HuggingFace Dataset çš„æŒ‡å®šåˆ—è½¬æ¢ä¸º PyTorch tensorã€‚
- è¿™æ ·åç»­å¯ä»¥ç›´æ¥ä¼ å…¥ PyTorch DataLoader è¿›è¡Œè®­ç»ƒæˆ–è¯„ä¼°ã€‚
- columns_to_return: 
    - 'input_ids': token ç¼–å·åºåˆ—
    - 'attention_mask': æ³¨æ„åŠ›æ©ç ï¼Œæ ‡è®°å¡«å……ä½ç½®
    - 'start_positions'ã€'end_positions': ç­”æ¡ˆåœ¨è¾“å…¥åºåˆ—ä¸­çš„èµ·æ­¢ token ç´¢å¼•
"""
```

åœ¨ PyTorch å®ç°ä¸­ï¼Œä¸ºäº†è¡¡é‡æ¨¡å‹çš„å‡†ç¡®æ€§ï¼Œä½ éœ€è¦ç¨ä½œè°ƒæ•´ï¼Œ  
ä½¿ç”¨æ•´ä¸ªæµ‹è¯•é›†çš„èµ·å§‹å’Œç»“æŸç´¢å¼•çš„ [F1 åˆ†æ•°](https://scikit-learn.org/stable/modules/generated/sklearn.metrics.f1_score.html) ä½œä¸ºæŸå¤±å‡½æ•°ã€‚


```python
# ----------------------------
# å®šä¹‰è®¡ç®— F1 åˆ†æ•°çš„å‡½æ•°ï¼ˆé’ˆå¯¹é—®ç­”ä»»åŠ¡çš„ start å’Œ end tokenï¼‰
# ----------------------------
from sklearn.metrics import f1_score

def compute_metrics(pred):
    """
    è®¡ç®—é¢„æµ‹ç»“æœçš„ start å’Œ end token çš„ F1 åˆ†æ•°
    
    å‚æ•°:
    - pred: Trainer æˆ–è€…è‡ªå®šä¹‰é¢„æµ‹å¯¹è±¡ï¼ŒåŒ…å« label_ids å’Œ predictions
    
    è¿”å›:
    - å­—å…¸ï¼ŒåŒ…å« 'f1_start' å’Œ 'f1_end'
    """

    # æå–çœŸå®æ ‡ç­¾çš„ start token ç´¢å¼•
    start_labels = pred.label_ids[0]

    # æå–é¢„æµ‹çš„ start token ç´¢å¼•ï¼ˆå–æ¯ä¸ª token æœ€å¤§æ¦‚ç‡å¯¹åº”çš„ç´¢å¼•ï¼‰
    start_preds = pred.predictions[0].argmax(-1)

    # æå–çœŸå®æ ‡ç­¾çš„ end token ç´¢å¼•
    end_labels = pred.label_ids[1]

    # æå–é¢„æµ‹çš„ end token ç´¢å¼•
    end_preds = pred.predictions[1].argmax(-1)

    # è®¡ç®— start token çš„ F1 åˆ†æ•°ï¼ˆmacro å¹³å‡ï¼‰
    f1_start = f1_score(start_labels, start_preds, average='macro')

    # è®¡ç®— end token çš„ F1 åˆ†æ•°ï¼ˆmacro å¹³å‡ï¼‰
    f1_end = f1_score(end_labels, end_preds, average='macro')

    # è¿”å›åŒ…å« start å’Œ end F1 çš„å­—å…¸
    return {
        'f1_start': f1_start,
        'f1_end': f1_end,
    }

"""
è§£é‡Šï¼š
- åœ¨é—®ç­”ä»»åŠ¡ä¸­ï¼Œæ¯ä¸ªæ ·æœ¬çš„ç­”æ¡ˆç”± start å’Œ end token å®šä¹‰ã€‚
- f1_score æ˜¯è¡¡é‡é¢„æµ‹ä¸çœŸå®æ ‡ç­¾åŒ¹é…ç¨‹åº¦çš„æŒ‡æ ‡ï¼Œå€¼è¶Šæ¥è¿‘ 1 è¶Šå¥½ã€‚
- average='macro' è¡¨ç¤ºå¯¹æ‰€æœ‰ç±»åˆ«è®¡ç®— F1 åå–å¹³å‡å€¼ã€‚
"""
```

#### æ¨¡å‹è®­ç»ƒ

ç°åœ¨æ˜¯æ—¶å€™åŠ è½½ä¸€ä¸ªé¢„è®­ç»ƒæ¨¡å‹äº†ã€‚

**æ³¨æ„ï¼š**  
åœ¨ PyTorch å®ç°ä¸­ï¼Œä½ å°†ä½¿ç”¨ **DistilBERT**ï¼Œè€Œä¸æ˜¯ TensorFlow ç‰ˆæœ¬çš„ **TFDistilBERT**ã€‚


```python
# ----------------------------
# è½½å…¥ PyTorch ç‰ˆæœ¬çš„ DistilBERT é—®ç­”æ¨¡å‹
# ----------------------------
from transformers import DistilBertForQuestionAnswering

# ä»æŒ‡å®šè·¯å¾„åŠ è½½å·²ç»è®­ç»ƒå¥½çš„ PyTorch æ¨¡å‹æƒé‡
pytorch_model = DistilBertForQuestionAnswering.from_pretrained("model/pytorch")

"""
è§£é‡Šï¼š
- DistilBertForQuestionAnswering æ˜¯ HuggingFace æä¾›çš„é¢„è®­ç»ƒæ¨¡å‹ç±»ï¼Œç”¨äºé—®ç­”ä»»åŠ¡ã€‚
- from_pretrained æ–¹æ³•å¯ä»¥è½½å…¥å·²ç»ä¿å­˜çš„æ¨¡å‹æƒé‡ï¼Œè·¯å¾„å¯ä»¥æ˜¯æœ¬åœ°ç›®å½•æˆ–è¿œç¨‹æ¨¡å‹åç§°ã€‚
- åŠ è½½åçš„æ¨¡å‹å¯ç›´æ¥ç”¨äº fine-tuning æˆ–é¢„æµ‹ã€‚
"""
```

ä½ å°†ä¸ä½¿ç”¨è‡ªå®šä¹‰è®­ç»ƒå¾ªç¯ï¼Œè€Œæ˜¯ä½¿ç”¨ [ğŸ¤— Trainer](https://huggingface.co/transformers/main_classes/trainer.html)ï¼Œ  
å®ƒåŒ…å«ä¸€ä¸ªåŸºç¡€çš„è®­ç»ƒå¾ªç¯ï¼Œå¹¶ä¸”åœ¨ PyTorch ä¸­å®ç°èµ·æ¥ç›¸å¯¹ç®€å•ã€‚


```python
# ----------------------------
# è®¾ç½®è®­ç»ƒå‚æ•°ä¸ Trainer è®­ç»ƒå™¨
# ----------------------------
from transformers import Trainer, TrainingArguments

# å®šä¹‰è®­ç»ƒå‚æ•°
training_args = TrainingArguments(
    output_dir='results',                 # è®­ç»ƒå®Œæˆåæ¨¡å‹åŠæ—¥å¿—ä¿å­˜çš„ç›®å½•
    overwrite_output_dir=True,            # å¦‚æœç›®å½•å­˜åœ¨ï¼Œè¦†ç›–åŸæœ‰å†…å®¹
    num_train_epochs=3,                   # å¯¹æ•´ä¸ªè®­ç»ƒé›†å¾ªç¯è®­ç»ƒ 3 ä¸ª epoch
    per_device_train_batch_size=8,        # æ¯ä¸ªè®¾å¤‡ï¼ˆGPU/CPUï¼‰æ¯æ¬¡è®­ç»ƒå¤„ç† 8 ä¸ªæ ·æœ¬
    per_device_eval_batch_size=8,         # æ¯ä¸ªè®¾å¤‡è¯„ä¼°æ—¶æ¯æ¬¡å¤„ç† 8 ä¸ªæ ·æœ¬
    warmup_steps=20,                      # è®­ç»ƒåˆæœŸçº¿æ€§å¢åŠ å­¦ä¹ ç‡çš„æ­¥æ•°ï¼ˆé¢„çƒ­ï¼‰
    weight_decay=0.01,                    # æƒé‡è¡°å‡ç³»æ•°ï¼Œç”¨äºé˜²æ­¢è¿‡æ‹Ÿåˆ
    report_to=[],                          # ç¦æ­¢æ—¥å¿—è®°å½•å’Œå›è°ƒï¼ˆå¦‚ TensorBoardï¼‰
)

# åˆå§‹åŒ– Trainer å¯¹è±¡
trainer = Trainer(
    model=pytorch_model,                  # å·²å®ä¾‹åŒ–çš„ PyTorch é—®ç­”æ¨¡å‹
    args=training_args,                   # ä¸Šé¢å®šä¹‰çš„è®­ç»ƒå‚æ•°
    train_dataset=train_ds,               # è®­ç»ƒæ•°æ®é›†
    eval_dataset=test_ds,                 # éªŒè¯/è¯„ä¼°æ•°æ®é›†
    compute_metrics=compute_metrics       # è‡ªå®šä¹‰è¯„ä¼°å‡½æ•°ï¼ˆè®¡ç®— F1 ç­‰æŒ‡æ ‡ï¼‰
)

# å¼€å§‹è®­ç»ƒ
trainer.train()                          # å¯åŠ¨è®­ç»ƒè¿‡ç¨‹ï¼Œä¼šè‡ªåŠ¨æ‰§è¡Œ forward/backward/optimizer step ç­‰æ“ä½œ

"""
é€è¡Œè§£æï¼š
1. output_dir='results'ï¼š
   - è®¾ç½®è®­ç»ƒè¾“å‡ºçš„ç›®å½•ï¼Œæ¨¡å‹æƒé‡ã€é…ç½®æ–‡ä»¶å’Œè®­ç»ƒæ—¥å¿—éƒ½ä¼šä¿å­˜åˆ°è¿™é‡Œã€‚
2. overwrite_output_dir=Trueï¼š
   - å¦‚æœ output_dir å·²å­˜åœ¨ï¼Œè¦†ç›–åŸæœ‰å†…å®¹ï¼Œé¿å…æŠ¥é”™ã€‚
3. num_train_epochs=3ï¼š
   - æ•°æ®é›†å¾ªç¯è®­ç»ƒ 3 æ¬¡ï¼Œå¯ç†è§£ä¸ºæ•´ä¸ªè®­ç»ƒæ•°æ®å®Œæ•´è¿‡ 3 éã€‚
4. per_device_train_batch_size=8ï¼š
   - æ¯ä¸ªè®¾å¤‡ï¼ˆCPU/GPUï¼‰æ¯æ¬¡è®­ç»ƒå¤„ç† 8 æ¡æ ·æœ¬ã€‚
5. per_device_eval_batch_size=8ï¼š
   - æ¯ä¸ªè®¾å¤‡è¯„ä¼°æ—¶æ¯æ¬¡å¤„ç† 8 æ¡æ ·æœ¬ã€‚
6. warmup_steps=20ï¼š
   - è®­ç»ƒå¼€å§‹é˜¶æ®µï¼Œå‰ 20 æ­¥é€æ¸å¢åŠ å­¦ä¹ ç‡ï¼Œæœ‰åŠ©äºæ¨¡å‹ç¨³å®šæ”¶æ•›ã€‚
7. weight_decay=0.01ï¼š
   - æƒé‡è¡°å‡ï¼ˆL2 æ­£åˆ™åŒ–ï¼‰ï¼Œé˜²æ­¢æ¨¡å‹è¿‡æ‹Ÿåˆã€‚
8. report_to=[]ï¼š
   - ç¦æ­¢æ‰€æœ‰æ—¥å¿—å›è°ƒï¼Œä¾‹å¦‚ TensorBoardã€W&B ç­‰ã€‚
9. Trainerï¼š
   - HuggingFace æä¾›çš„è®­ç»ƒå™¨ï¼Œæ•´åˆè®­ç»ƒã€è¯„ä¼°ã€ä¼˜åŒ–å™¨ã€å›è°ƒå’Œæ—¥å¿—ã€‚
10. trainer.train()ï¼š
   - å¯åŠ¨è®­ç»ƒå¾ªç¯ï¼ŒåŒ…æ‹¬å‰å‘è®¡ç®—ã€æŸå¤±è®¡ç®—ã€åå‘ä¼ æ’­ã€ä¼˜åŒ–å™¨æ›´æ–°ã€è¯„ä¼°å’Œæ—¥å¿—è®°å½•ã€‚
"""
```

```python
# -------------------------------------------------------------
# 15. ä½¿ç”¨ Trainer å¯¹æµ‹è¯•é›†è¿›è¡Œè¯„ä¼°
# -------------------------------------------------------------

# è°ƒç”¨ Trainer å¯¹è±¡çš„ evaluate æ–¹æ³•ï¼Œå¯¹ test_ds æ•°æ®é›†è¿›è¡Œè¯„ä¼°
# å‚æ•°ï¼š
#   test_ds: è¦è¯„ä¼°çš„ PyTorch Dataset æˆ– HuggingFace Dataset å¯¹è±¡
# è¿”å›å€¼ï¼š
#   ä¸€ä¸ªå­—å…¸ï¼ŒåŒ…å«è¯„ä¼°æŒ‡æ ‡ï¼Œä¾‹å¦‚ loss æˆ–è‡ªå®šä¹‰çš„ compute_metrics è¿”å›çš„æŒ‡æ ‡
evaluation_results = trainer.evaluate(test_ds)

# è¾“å‡ºè¯„ä¼°ç»“æœ
print("è¯„ä¼°ç»“æœï¼š")
print(evaluation_results)
```

ç°åœ¨æ˜¯æ—¶å€™å‘ä½ çš„ PyTorch æ¨¡å‹æé—®äº†ï¼  
* åœ¨ä½¿ç”¨é—®é¢˜æµ‹è¯•æ¨¡å‹ä¹‹å‰ï¼Œå¯ä»¥å°†æ¨¡å‹å’Œè¾“å…¥å‘é€åˆ° GPUï¼ˆå¦‚æœä½ çš„æœºå™¨æœ‰ GPUï¼‰æˆ– CPUï¼ˆå¦‚æœæ²¡æœ‰ GPUï¼‰ã€‚  
* ç„¶åï¼Œä½ å¯ä»¥å¯¹è¾“å…¥è¿›è¡Œåˆ†è¯ï¼ˆtokenizeï¼‰ã€åˆ›å»º PyTorch å¼ é‡ï¼Œå¹¶å‘é€åˆ°ç›¸åº”çš„è®¾å¤‡ï¼ˆdeviceï¼‰ã€‚  
* å…¶ä½™çš„å¤„ç†æµç¨‹ä¸ä¹‹å‰åœ¨ TensorFlow ä¸­å®ç°çš„æµç¨‹ç›¸å¯¹ç±»ä¼¼ã€‚


```python
# -------------------------------------------------------------
# 16. ä½¿ç”¨è®­ç»ƒå¥½çš„ PyTorch QA æ¨¡å‹è¿›è¡Œå•æ¡é—®ç­”é¢„æµ‹
# -------------------------------------------------------------
import torch

# é€‰æ‹©è®¾å¤‡ï¼šå¦‚æœæœ‰ GPU åˆ™ä½¿ç”¨ CUDAï¼Œå¦åˆ™ä½¿ç”¨ CPU
device = torch.device('cuda') if torch.cuda.is_available() else torch.device('cpu')

# å°†æ¨¡å‹ç§»åŠ¨åˆ°æŒ‡å®šè®¾å¤‡ï¼ˆGPU æˆ– CPUï¼‰
pytorch_model.to(device)

# å®šä¹‰ç¤ºä¾‹é—®é¢˜å’Œä¸Šä¸‹æ–‡æ–‡æœ¬
question = 'What is east of the hallway?'
text = 'The kitchen is east of the hallway. The garden is south of the bedroom.'

# ä½¿ç”¨ tokenizer å¯¹æ–‡æœ¬å’Œé—®é¢˜è¿›è¡Œç¼–ç 
# å‚æ•°è¯´æ˜ï¼š
#   text: ä¸Šä¸‹æ–‡æ–‡æœ¬
#   question: é—®é¢˜æ–‡æœ¬
#   return_tensors='pt': è¿”å› PyTorch tensor æ ¼å¼
input_dict = tokenizer(text, question, return_tensors='pt')

# å°† input_ids å’Œ attention_mask è½¬ç§»åˆ°æŒ‡å®šè®¾å¤‡
input_ids = input_dict['input_ids'].to(device)           # ç¼–ç åçš„ token ID
attention_mask = input_dict['attention_mask'].to(device) # æ³¨æ„åŠ›æ©ç ï¼Œ1è¡¨ç¤ºæœ‰æ•ˆ tokenï¼Œ0è¡¨ç¤º padding

# æ¨¡å‹å‰å‘ä¼ æ’­ï¼Œå¾—åˆ°é¢„æµ‹çš„ start_logits å’Œ end_logits
outputs = pytorch_model(input_ids, attention_mask=attention_mask)

# DistilBERT QA æ¨¡å‹è¾“å‡ºæ˜¯ä¸€ä¸ª tuple:
# outputs[0] -> start_logits: æ¯ä¸ª token ä¸ºç­”æ¡ˆèµ·å§‹ä½ç½®çš„å¾—åˆ†
# outputs[1] -> end_logits: æ¯ä¸ª token ä¸ºç­”æ¡ˆç»“æŸä½ç½®çš„å¾—åˆ†
start_logits = outputs[0]
end_logits = outputs[1]

# å°† input_ids è½¬å›å¯¹åº”çš„ token å­—ç¬¦ä¸²
all_tokens = tokenizer.convert_ids_to_tokens(input_dict["input_ids"].numpy()[0])

# æ‰¾åˆ° start_logits å’Œ end_logits å¾—åˆ†æœ€å¤§çš„ token ç´¢å¼•
# torch.argmax(start_logits, 1)[0] -> ç­”æ¡ˆèµ·å§‹ token ç´¢å¼•
# torch.argmax(end_logits, 1)[0] -> ç­”æ¡ˆç»“æŸ token ç´¢å¼•
answer_tokens = all_tokens[torch.argmax(start_logits, 1)[0] : torch.argmax(end_logits, 1)[0]+1]

# å°† token æ‹¼æ¥ä¸ºå­—ç¬¦ä¸²ï¼Œå¾—åˆ°æœ€ç»ˆç­”æ¡ˆ
answer = ' '.join(answer_tokens)

# æ‰“å°é—®é¢˜ä¸ç­”æ¡ˆï¼ˆé¦–å­—æ¯å¤§å†™ï¼‰
print(question, answer.capitalize())
```

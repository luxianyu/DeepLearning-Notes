````python
import numpy as np
import matplotlib.pyplot as plt
import h5py
import sklearn
import sklearn.datasets
import sklearn.linear_model
import scipy.io

def sigmoid(x):
    """
    Compute the sigmoid of x

    Arguments:
    x -- A scalar or numpy array of any size.

    Return:
    s -- sigmoid(x)
    """
    s = 1/(1+np.exp(-x))
    return s

def relu(x):
    """
    Compute the relu of x

    Arguments:
    x -- A scalar or numpy array of any size.

    Return:
    s -- relu(x)
    """
    s = np.maximum(0,x)
    
    return s

def load_planar_dataset(seed):
    
    np.random.seed(seed)
    
    m = 400 # number of examples
    N = int(m/2) # number of points per class
    D = 2 # dimensionality
    X = np.zeros((m,D)) # data matrix where each row is a single example
    Y = np.zeros((m,1), dtype='uint8') # labels vector (0 for red, 1 for blue)
    a = 4 # maximum ray of the flower

    for j in range(2):
        ix = range(N*j,N*(j+1))
        t = np.linspace(j*3.12,(j+1)*3.12,N) + np.random.randn(N)*0.2 # theta
        r = a*np.sin(4*t) + np.random.randn(N)*0.2 # radius
        X[ix] = np.c_[r*np.sin(t), r*np.cos(t)]
        Y[ix] = j
        
    X = X.T
    Y = Y.T

    return X, Y

def initialize_parameters(layer_dims):
    """
    Arguments:
    layer_dims -- python array (list) containing the dimensions of each layer in our network
    
    Returns:
    parameters -- python dictionary containing your parameters "W1", "b1", ..., "WL", "bL":
                    W1 -- weight matrix of shape (layer_dims[l], layer_dims[l-1])
                    b1 -- bias vector of shape (layer_dims[l], 1)
                    Wl -- weight matrix of shape (layer_dims[l-1], layer_dims[l])
                    bl -- bias vector of shape (1, layer_dims[l])
                    
    Tips:
    - For example: the layer_dims for the "Planar Data classification model" would have been [2,2,1]. 
    This means W1's shape was (2,2), b1 was (1,2), W2 was (2,1) and b2 was (1,1). Now you have to generalize it!
    - In the for loop, use parameters['W' + str(l)] to access Wl, where l is the iterative integer.
    """
    
    np.random.seed(3)
    parameters = {}
    L = len(layer_dims) # number of layers in the network

    for l in range(1, L):
        parameters['W' + str(l)] = np.random.randn(layer_dims[l], layer_dims[l-1]) / np.sqrt(layer_dims[l-1])
        parameters['b' + str(l)] = np.zeros((layer_dims[l], 1))
        
        assert(parameters['W' + str(l)].shape == (layer_dims[l], layer_dims[l-1]))
        assert(parameters['W' + str(l)].shape == layer_dims[l], 1)

        
    return parameters

def forward_propagation(X, parameters):
    """
    Implements the forward propagation (and computes the loss) presented in Figure 2.
    
    Arguments:
    X -- input dataset, of shape (input size, number of examples)
    parameters -- python dictionary containing your parameters "W1", "b1", "W2", "b2", "W3", "b3":
                    W1 -- weight matrix of shape ()
                    b1 -- bias vector of shape ()
                    W2 -- weight matrix of shape ()
                    b2 -- bias vector of shape ()
                    W3 -- weight matrix of shape ()
                    b3 -- bias vector of shape ()
    
    Returns:
    loss -- the loss function (vanilla logistic loss)
    """
        
    # retrieve parameters
    W1 = parameters["W1"]
    b1 = parameters["b1"]
    W2 = parameters["W2"]
    b2 = parameters["b2"]
    W3 = parameters["W3"]
    b3 = parameters["b3"]
    
    # LINEAR -> RELU -> LINEAR -> RELU -> LINEAR -> SIGMOID
    Z1 = np.dot(W1, X) + b1
    A1 = relu(Z1)
    Z2 = np.dot(W2, A1) + b2
    A2 = relu(Z2)
    Z3 = np.dot(W3, A2) + b3
    A3 = sigmoid(Z3)
    
    cache = (Z1, A1, W1, b1, Z2, A2, W2, b2, Z3, A3, W3, b3)
    
    return A3, cache

def backward_propagation(X, Y, cache):
    """
    Implement the backward propagation presented in figure 2.
    
    Arguments:
    X -- input dataset, of shape (input size, number of examples)
    Y -- true "label" vector (containing 0 if cat, 1 if non-cat)
    cache -- cache output from forward_propagation()
    
    Returns:
    gradients -- A dictionary with the gradients with respect to each parameter, activation and pre-activation variables
    """
    m = X.shape[1]
    (Z1, A1, W1, b1, Z2, A2, W2, b2, Z3, A3, W3, b3) = cache
    
    dZ3 = A3 - Y
    dW3 = 1./m * np.dot(dZ3, A2.T)
    db3 = 1./m * np.sum(dZ3, axis=1, keepdims = True)
    
    dA2 = np.dot(W3.T, dZ3)
    dZ2 = np.multiply(dA2, np.int64(A2 > 0))
    dW2 = 1./m * np.dot(dZ2, A1.T)
    db2 = 1./m * np.sum(dZ2, axis=1, keepdims = True)
    
    dA1 = np.dot(W2.T, dZ2)
    dZ1 = np.multiply(dA1, np.int64(A1 > 0))
    dW1 = 1./m * np.dot(dZ1, X.T)
    db1 = 1./m * np.sum(dZ1, axis=1, keepdims = True)
    
    gradients = {"dZ3": dZ3, "dW3": dW3, "db3": db3,
                 "dA2": dA2, "dZ2": dZ2, "dW2": dW2, "db2": db2,
                 "dA1": dA1, "dZ1": dZ1, "dW1": dW1, "db1": db1}
    
    return gradients

def update_parameters(parameters, grads, learning_rate):
    """
    Update parameters using gradient descent
    
    Arguments:
    parameters -- python dictionary containing your parameters:
                    parameters['W' + str(i)] = Wi
                    parameters['b' + str(i)] = bi
    grads -- python dictionary containing your gradients for each parameters:
                    grads['dW' + str(i)] = dWi
                    grads['db' + str(i)] = dbi
    learning_rate -- the learning rate, scalar.
    
    Returns:
    parameters -- python dictionary containing your updated parameters 
    """
    
    n = len(parameters) // 2 # number of layers in the neural networks

    # Update rule for each parameter
    for k in range(n):
        parameters["W" + str(k+1)] = parameters["W" + str(k+1)] - learning_rate * grads["dW" + str(k+1)]
        parameters["b" + str(k+1)] = parameters["b" + str(k+1)] - learning_rate * grads["db" + str(k+1)]
        
    return parameters

def predict(X, y, parameters):
    """
    This function is used to predict the results of a  n-layer neural network.
    
    Arguments:
    X -- data set of examples you would like to label
    parameters -- parameters of the trained model
    
    Returns:
    p -- predictions for the given dataset X
    """
    
    m = X.shape[1]
    #p = np.zeros((1,m), dtype = np.int)
    p = np.zeros((1, m), dtype=int)

    
    # Forward propagation
    a3, caches = forward_propagation(X, parameters)
    
    # convert probas to 0/1 predictions
    for i in range(0, a3.shape[1]):
        if a3[0,i] > 0.5:
            p[0,i] = 1
        else:
            p[0,i] = 0

    # print results

    #print ("predictions: " + str(p[0,:]))
    #print ("true labels: " + str(y[0,:]))
    print("Accuracy: "  + str(np.mean((p[0,:] == y[0,:]))))
    
    return p

def compute_cost(a3, Y):
    """
    Implement the cost function
    
    Arguments:
    a3 -- post-activation, output of forward propagation
    Y -- "true" labels vector, same shape as a3
    
    Returns:
    cost - value of the cost function
    """
    m = Y.shape[1]
    
    logprobs = np.multiply(-np.log(a3),Y) + np.multiply(-np.log(1 - a3), 1 - Y)
    cost = 1./m * np.nansum(logprobs)
    
    return cost

def load_dataset():
    train_dataset = h5py.File('datasets/train_catvnoncat.h5', "r")
    train_set_x_orig = np.array(train_dataset["train_set_x"][:]) # your train set features
    train_set_y_orig = np.array(train_dataset["train_set_y"][:]) # your train set labels

    test_dataset = h5py.File('datasets/test_catvnoncat.h5', "r")
    test_set_x_orig = np.array(test_dataset["test_set_x"][:]) # your test set features
    test_set_y_orig = np.array(test_dataset["test_set_y"][:]) # your test set labels

    classes = np.array(test_dataset["list_classes"][:]) # the list of classes
    
    train_set_y = train_set_y_orig.reshape((1, train_set_y_orig.shape[0]))
    test_set_y = test_set_y_orig.reshape((1, test_set_y_orig.shape[0]))
    
    train_set_x_orig = train_set_x_orig.reshape(train_set_x_orig.shape[0], -1).T
    test_set_x_orig = test_set_x_orig.reshape(test_set_x_orig.shape[0], -1).T
    
    train_set_x = train_set_x_orig/255
    test_set_x = test_set_x_orig/255

    return train_set_x, train_set_y, test_set_x, test_set_y, classes


def predict_dec(parameters, X):
    """
    Used for plotting decision boundary.
    
    Arguments:
    parameters -- python dictionary containing your parameters 
    X -- input data of size (m, K)
    
    Returns
    predictions -- vector of predictions of our model (red: 0 / blue: 1)
    """
    
    # Predict using forward propagation and a classification threshold of 0.5
    a3, cache = forward_propagation(X, parameters)
    predictions = (a3>0.5)
    return predictions

def load_planar_dataset(randomness, seed):
    
    np.random.seed(seed)
    
    m = 50
    N = int(m/2) # number of points per class
    D = 2 # dimensionality
    X = np.zeros((m,D)) # data matrix where each row is a single example
    Y = np.zeros((m,1), dtype='uint8') # labels vector (0 for red, 1 for blue)
    a = 2 # maximum ray of the flower

    for j in range(2):
        
        ix = range(N*j,N*(j+1))
        if j == 0:
            t = np.linspace(j, 4*3.1415*(j+1),N) #+ np.random.randn(N)*randomness # theta
            r = 0.3*np.square(t) + np.random.randn(N)*randomness # radius
        if j == 1:
            t = np.linspace(j, 2*3.1415*(j+1),N) #+ np.random.randn(N)*randomness # theta
            r = 0.2*np.square(t) + np.random.randn(N)*randomness # radius
            
        X[ix] = np.c_[r*np.cos(t), r*np.sin(t)]
        Y[ix] = j
        
    X = X.T
    Y = Y.T

    return X, Y

def plot_decision_boundary(model, X, y):
    # Set min and max values and give it some padding
    x_min, x_max = X[0, :].min() - 1, X[0, :].max() + 1
    y_min, y_max = X[1, :].min() - 1, X[1, :].max() + 1
    h = 0.01
    # Generate a grid of points with distance h between them
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
    # Predict the function value for the whole grid
    Z = model(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    # Plot the contour and training examples
    plt.contourf(xx, yy, Z, cmap=plt.cm.Spectral)
    plt.ylabel('x2')
    plt.xlabel('x1')
    plt.scatter(X[0, :], X[1, :], c=y.ravel(), cmap=plt.cm.Spectral)
    plt.show()
    
def load_2D_dataset():
    data = scipy.io.loadmat('datasets/data.mat')
    train_X = data['X'].T
    train_Y = data['y'].T
    test_X = data['Xval'].T
    test_Y = data['yval'].T

    plt.scatter(train_X[0, :], train_X[1, :], c=train_Y.ravel(), s=40, cmap=plt.cm.Spectral);
    
    return train_X, train_Y, test_X, test_Y
````

## 1 - 无正则化模型

你将使用下面的神经网络（已为你实现）。该模型可以用于：
- *正则化模式* —— 通过将输入参数 `lambd` 设置为非零值来启用。这里用 "`lambd`" 替代 "`lambda`"，因为 "`lambda`" 是 Python 的保留关键字。  
- *Dropout 模式* —— 通过将 `keep_prob` 设置为小于1的值来启用。

你将先尝试无任何正则化的模型。接着，你将实现：
- *L2 正则化* —— 函数 "`compute_cost_with_regularization()`" 和 "`backward_propagation_with_regularization()`"
- *Dropout* —— 函数 "`forward_propagation_with_dropout()`" 和 "`backward_propagation_with_dropout()`"

在每个部分，你将运行带有正确输入的模型，使其调用你实现的函数。请先阅读下面的代码，熟悉模型结构。


````python
def model(X, Y, learning_rate=0.3, num_iterations=30000, print_cost=True, is_plot=True, lambd=0, keep_prob=1):
    """
    实现一个三层的神经网络结构：LINEAR -> RELU -> LINEAR -> RELU -> LINEAR -> SIGMOID
    
    参数：
        X -- 输入数据，形状为 (输入特征数, 样本数)，本例为 (2, m)
        Y -- 标签数据，取值 {0(蓝色), 1(红色)}，形状为 (1, m)
        learning_rate -- 学习率（梯度下降步长），控制每次更新幅度
        num_iterations -- 训练的迭代次数
        print_cost -- 是否在训练过程中打印成本值，每 10000 次迭代打印一次
        is_plot -- 是否绘制成本曲线图
        lambd -- L2 正则化参数，防止过拟合，默认为 0 表示不使用
        keep_prob -- Dropout 保留节点的概率（1 表示不使用 Dropout）
    
    返回：
        parameters -- 学习后的参数字典（包含各层权重和偏置）
    """

    grads = {}                   # 存储反向传播得到的梯度
    costs = []                   # 存储每 1000 次迭代的成本值，用于绘图
    m = X.shape[1]               # 样本数 m
    layers_dims = [X.shape[0],   # 输入层神经元个数（等于输入特征数）
                   20,           # 第一隐藏层神经元个数
                   3,            # 第二隐藏层神经元个数
                   1]            # 输出层神经元个数（sigmoid 输出一个概率）

    # 初始化参数（权重 W 和偏置 b）
    parameters = initialize_parameters(layers_dims)

    # 开始迭代训练
    for i in range(0, num_iterations):

        # -------- 前向传播 --------
        if keep_prob == 1:
            # 不使用 Dropout
            a3, cache = forward_propagation(X, parameters)
        elif keep_prob < 1:
            # 使用 Dropout
            a3, cache = forward_propagation_with_dropout(X, parameters, keep_prob)
        else:
            # 参数错误提示
            print("keep_prob 参数错误！程序退出。")
            exit()

        # -------- 计算成本 --------
        if lambd == 0:
            # 不使用 L2 正则化
            cost = compute_cost(a3, Y)
        else:
            # 使用 L2 正则化
            cost = compute_cost_with_regularization(a3, Y, parameters, lambd)

        # -------- 反向传播 --------
        # 保证不会同时开启 L2 正则化 和 Dropout（本实验不同时使用）
        assert (lambd == 0 or keep_prob == 1)

        if (lambd == 0 and keep_prob == 1):
            # 普通反向传播
            grads = backward_propagation(X, Y, cache)
        elif lambd != 0:
            # 含 L2 正则化的反向传播
            grads = backward_propagation_with_regularization(X, Y, cache, lambd)
        elif keep_prob < 1:
            # 含 Dropout 的反向传播
            grads = backward_propagation_with_dropout(X, Y, cache, keep_prob)

        # -------- 更新参数 --------
        parameters = update_parameters(parameters, grads, learning_rate)

        # -------- 记录成本并按需打印 --------
        if i % 1000 == 0:
            costs.append(cost)  # 每 1000 次记录一次
            if (print_cost and i % 10000 == 0):
                print(f"第 {i} 次迭代，成本值为：{cost}")

    # -------- 绘制成本曲线 --------
    if is_plot:
        plt.plot(costs)
        plt.ylabel('cost')  # y 轴：成本值
        plt.xlabel('iterations (x1,000)')  # x 轴：迭代次数（单位千次）
        plt.title(f"Learning rate = {learning_rate}")
        plt.show()

    # 返回最终训练好的参数
    return parameters
````

让我们先训练一个不带任何正则化的模型，并观察其在训练集和测试集上的准确率。


````python
parameters = model(train_X, train_Y)
print ("On the training set:")
predictions_train = predict(train_X, train_Y, parameters)
print ("On the test set:")
predictions_test = predict(test_X, test_Y, parameters)
````

显然，这个无正则化模型对训练集产生了过拟合，甚至拟合了噪声点！接下来我们来看两种减少过拟合的技术。


## 2 - L2 正则化

避免过拟合的标准方法称为 **L2 正则化**。它通过适当修改损失函数实现，从：
$$J = -\frac{1}{m} \sum\limits_{i = 1}^{m} \large{(}\small  y^{(i)}\log\left(a^{[L](i)}\right) + (1-y^{(i)})\log\left(1- a^{[L](i)}\right) \large{)} \tag{1}$$
变为：
$$J_{regularized} = \small \underbrace{-\frac{1}{m} \sum\limits_{i = 1}^{m} \large{(}\small y^{(i)}\log\left(a^{[L](i)}\right) + (1-y^{(i)})\log\left(1- a^{[L](i)}\right) \large{)} }_\text{交叉熵损失} + \underbrace{\frac{1}{m} \frac{\lambda}{2} \sum\limits_l\sum\limits_k\sum\limits_j W_{k,j}^{[l]2} }_\text{L2正则化损失} \tag{2}$$

我们来修改损失函数并观察效果。

**练习**：实现函数 `compute_cost_with_regularization()`，计算公式（2）中的损失。  
计算 $\sum\limits_k\sum\limits_j W_{k,j}^{[l]2}$ 时，可以使用：
```python
np.sum(np.square(Wl))


注意，你需要对 $W^{[1]}$、$W^{[2]}$ 和 $W^{[3]}$ 分别计算平方和，然后将这三项相加，最后乘以 $\frac{1}{m} \frac{\lambda}{2}$。


````python
def compute_cost_with_regularization(A3, Y, parameters, lambd):
    """
    计算带有 L2 正则化项的成本函数值（公式2）

    参数：
        A3 -- 前向传播得到的预测值（神经网络输出层的激活值），
              维度为 (输出层神经元数量, 样本数量 m)
        Y -- 样本真实标签，维度与 A3 相同
        parameters -- 字典，存储了模型的参数 W1, W2, W3 等
        lambd -- L2 正则化的超参数 λ（控制正则化强度）

    返回：
        cost -- 加入 L2 正则化项后的总成本
    """

    # 1. 获取样本数量 m
    m = Y.shape[1]  # shape[1] 是列数，代表样本数量

    # 2. 从参数字典中取出权重矩阵 W1、W2、W3（偏置不参与 L2 正则化）
    W1 = parameters["W1"]
    W2 = parameters["W2"]
    W3 = parameters["W3"]

    # 3. 计算普通的交叉熵损失（不带正则化）
    cross_entropy_cost = compute_cost(A3, Y)

    # 4. 计算 L2 正则化成本项
    #    np.sum(np.square(W)) 表示求权重矩阵中所有元素的平方和
    #    除以 (2*m) 是公式中的标准做法
    L2_regularization_cost = lambd * (
        np.sum(np.square(W1)) +
        np.sum(np.square(W2)) +
        np.sum(np.square(W3))
    ) / (2 * m)

    # 5. 总成本 = 交叉熵损失 + L2 正则化项
    cost = cross_entropy_cost + L2_regularization_cost

    return cost



````

````python
A3, Y_assess, parameters = compute_cost_with_regularization_test_case()

print("cost = " + str(compute_cost_with_regularization(A3, Y_assess, parameters, lambd = 0.1)))
````

当然，因为你修改了损失函数，反向传播也必须做相应的调整！所有梯度都应基于新的损失函数计算。

**练习**：实现反向传播中与正则化相关的修改。只需要修改 dW1、dW2 和 dW3。  
对于每一个权重梯度，都需要加上正则化项的梯度：
$$
\frac{d}{dW} \left( \frac{1}{2} \frac{\lambda}{m} W^2 \right) = \frac{\lambda}{m} W
$$


````python
def backward_propagation_with_regularization(X, Y, cache, lambd):
    """
    实现带 L2 正则化的反向传播

    参数：
        X -- 输入数据，维度为 (输入层神经元数量, 样本数量 m)
        Y -- 样本真实标签，维度与输出层一致
        cache -- 从 forward_propagation() 得到的缓存值
                 (包括每层的Z、A、W、b)
        lambd -- L2 正则化超参数 λ

    返回：
        gradients -- 字典，存储每个参数的梯度
    """

    # 1. 样本数量 m
    m = X.shape[1]

    # 2. 从 cache 中取出各层的中间值（Z、A、W、b）
    (Z1, A1, W1, b1,
     Z2, A2, W2, b2,
     Z3, A3, W3, b3) = cache

    # 3. 输出层梯度计算
    dZ3 = A3 - Y  # 输出层误差
    dW3 = (1 / m) * np.dot(dZ3, A2.T) + (lambd * W3 / m)  # 加入 L2 正则化梯度
    db3 = (1 / m) * np.sum(dZ3, axis=1, keepdims=True)    # 偏置不加正则化

    # 4. 第二层梯度计算
    dA2 = np.dot(W3.T, dZ3)                  # 误差传回到 A2
    dZ2 = np.multiply(dA2, np.int64(A2 > 0)) # ReLU 的导数
    dW2 = (1 / m) * np.dot(dZ2, A1.T) + (lambd * W2 / m)  # 加 L2 正则化
    db2 = (1 / m) * np.sum(dZ2, axis=1, keepdims=True)

    # 5. 第一层梯度计算
    dA1 = np.dot(W2.T, dZ2)                  # 误差传回到 A1
    dZ1 = np.multiply(dA1, np.int64(A1 > 0)) # ReLU 的导数
    dW1 = (1 / m) * np.dot(dZ1, X.T) + (lambd * W1 / m)   # 加 L2 正则化
    db1 = (1 / m) * np.sum(dZ1, axis=1, keepdims=True)

    # 6. 将所有梯度存储到字典中返回
    gradients = {
        "dZ3": dZ3, "dW3": dW3, "db3": db3,
        "dA2": dA2, "dZ2": dZ2, "dW2": dW2, "db2": db2,
        "dA1": dA1, "dZ1": dZ1, "dW1": dW1, "db1": db1
    }

    return gradients
````

````python
X_assess, Y_assess, cache = backward_propagation_with_regularization_test_case()

grads = backward_propagation_with_regularization(X_assess, Y_assess, cache, lambd = 0.7)
print ("dW1 = "+ str(grads["dW1"]))
print ("dW2 = "+ str(grads["dW2"]))
print ("dW3 = "+ str(grads["dW3"]))
````

- `compute_cost_with_regularization` 代替 `compute_cost`  
- `backward_propagation_with_regularization` 代替 `backward_propagation`  


````python
parameters = model(train_X, train_Y, lambd = 0.7)
print ("On the train set:")
predictions_train = predict(train_X, train_Y, parameters)
print ("On the test set:")
predictions_test = predict(test_X, test_Y, parameters)
````

**观察**：
- 参数 $\lambda$ 是一个超参数，可以通过开发集调优。
- L2 正则化使决策边界更加平滑。如果 $\lambda$ 太大，也可能导致“过度平滑”，产生高偏差模型。

**L2 正则化到底做了什么？**

L2 正则化基于这样一个假设：权重较小的模型比权重较大的模型更简单。因此，通过在损失函数中惩罚权重的平方值，你会促使所有权重变小。权重过大代价太高！这会导致模型更平滑，输出随着输入的变化更缓慢。

<font color='blue'>
**你应该记住的** — 关于 L2 正则化的影响：
- 损失计算：
    - 损失函数中增加了正则化项
- 反向传播：
    - 权重矩阵的梯度中增加了额外项
- 权重趋于变小（“权重衰减”）：
    - 权重被推向更小的数值
</font>


## 3 - Dropout

最后，**Dropout** 是深度学习中特有的一种广泛使用的正则化技术。  
**它在每次迭代中随机关闭（舍弃）部分神经元。**

当你关闭一些神经元时，实际上是在改变模型。Dropout 背后的思想是，每次迭代训练的是一个只使用部分神经元的不同模型。  
因此，使用 Dropout 后，神经元对其他特定神经元的激活变得不那么敏感，因为这些神经元随时可能被关闭。

### 3.1 - 带 Dropout 的前向传播

**练习**：实现带 Dropout 的前向传播。你使用的是一个三层神经网络，将在第一和第二隐藏层添加 Dropout，不对输入层和输出层使用 Dropout。

**说明**：  
你需要在第一和第二层关闭部分神经元。具体做法分四步：

1. 在课程中，我们讨论过创建与 $a^{ } d^{ } ... d^{[1](m)}]$，形状与 $A^{[1]}$ 一致。

2. 根据概率（`1 - keep_prob`）将 $D^{[1]}$ 中的元素置为0，概率（`keep_prob`）置为1。方法是对矩阵 $D^{[1]}$ 进行阈值操作。例如，要将矩阵X中小于0.5的元素置为0，大于等于0.5的元素置为1，可以写成 `X = (X < 0.5)`。注意，0和1分别对应布尔值False和True。

3. 将 $A^{[1]}$ 乘以 $D^{[1]}$，即 $A^{[1]} = A^{[1]} * D^{[1]}$，实现关闭部分神经元。你可以把 $D^{[1]}$ 看作一个“掩码”，乘法过程即屏蔽某些神经元。

4. 将 $A^{[1]}$ 除以 `keep_prob`，确保激活值的期望不变（此技术称为“反向 Dropout”）。



````python
def forward_propagation_with_dropout(X, parameters, keep_prob=0.5):
    """
    实现带 Dropout 的前向传播：
    LINEAR -> RELU + DROPOUT -> LINEAR -> RELU + DROPOUT -> LINEAR -> SIGMOID

    参数：
        X -- 输入数据集，维度为 (输入节点数量, 样本数量)
        parameters -- 字典，存储模型参数 W1,b1,W2,b2,W3,b3
            W1 -- 第一层权重矩阵，维度 (20, 2)
            b1 -- 第一层偏置向量，维度 (20, 1)
            W2 -- 第二层权重矩阵，维度 (3, 20)
            b2 -- 第二层偏置向量，维度 (3, 1)
            W3 -- 第三层权重矩阵，维度 (1, 3)
            b3 -- 第三层偏置向量，维度 (1, 1)
        keep_prob -- 保持神经元活跃的概率，标量

    返回：
        A3 -- 输出层激活值，维度 (1, 样本数量)
        cache -- 保存中间结果的元组，用于反向传播
    """

    # 固定随机种子，保证每次生成的 Dropout mask 一致
    np.random.seed(1)
    
    # --------------------------- 1. 获取参数 ---------------------------
    W1 = parameters["W1"]  # 第一层权重
    b1 = parameters["b1"]  # 第一层偏置
    W2 = parameters["W2"]  # 第二层权重
    b2 = parameters["b2"]  # 第二层偏置
    W3 = parameters["W3"]  # 第三层权重
    b3 = parameters["b3"]  # 第三层偏置
    
    # --------------------------- 2. 第一层 LINEAR -> RELU -> DROPOUT ---------------------------
    Z1 = np.dot(W1, X) + b1   # 线性变换 Z1 = W1*X + b1
    A1 = relu(Z1)             # ReLU 激活 A1 = max(0, Z1)

    # Step 1: 生成随机矩阵 D1 与 A1 相同维度，元素范围 [0,1]
    D1 = np.random.rand(A1.shape[0], A1.shape[1])
    # Step 2: 根据 keep_prob 阈值，将 D1 转换为 0/1 mask
    D1 = D1 < keep_prob
    # Step 3: 对 A1 中的神经元应用 mask（某些神经元关闭）
    A1 = A1 * D1
    # Step 4: 对未关闭的神经元进行缩放，保持期望值不变
    A1 = A1 / keep_prob

    # --------------------------- 3. 第二层 LINEAR -> RELU -> DROPOUT ---------------------------
    Z2 = np.dot(W2, A1) + b2  # 线性变换 Z2 = W2*A1 + b2
    A2 = relu(Z2)              # ReLU 激活 A2 = max(0, Z2)

    # Step 1: 生成随机矩阵 D2 与 A2 相同维度
    D2 = np.random.rand(A2.shape[0], A2.shape[1])
    # Step 2: 根据 keep_prob 转换为 0/1 mask
    D2 = D2 < keep_prob
    # Step 3: 对 A2 中神经元应用 mask
    A2 = A2 * D2
    # Step 4: 对未关闭神经元进行缩放
    A2 = A2 / keep_prob

    # --------------------------- 4. 第三层 LINEAR -> SIGMOID ---------------------------
    Z3 = np.dot(W3, A2) + b3  # 线性变换 Z3 = W3*A2 + b3
    A3 = sigmoid(Z3)           # sigmoid 激活输出层预测值

    # --------------------------- 5. 保存中间变量到 cache ---------------------------
    # 包括每层的 Z、A、W、b 以及 Dropout mask D
    cache = (Z1, D1, A1, W1, b1,
             Z2, D2, A2, W2, b2,
             Z3, A3, W3, b3)

    return A3, cache
````

````python
X_assess, parameters = forward_propagation_with_dropout_test_case()

A3, cache = forward_propagation_with_dropout(X_assess, parameters, keep_prob = 0.7)
print ("A3 = " + str(A3))
````

### 3.2 - 带 Dropout 的反向传播

**练习**：实现带 Dropout 的反向传播。和之前一样，你训练的是一个三层神经网络。在第一和第二隐藏层加入 Dropout，使用缓存中保存的掩码 $D^{[1]}$ 和 $D^{[2]}$。

**说明**：  
带 Dropout 的反向传播其实很简单，需执行以下两步：

1. 前向传播时，你用掩码 $D^{[1]}$ 关闭了部分神经元（应用于 `A1`）。反向传播时，你需要用同样的掩码 $D^{[1]}$ 关闭对应的神经元（应用于 `dA1`）。

2. 前向传播时，你将 `A1` 除以了 `keep_prob`，因此反向传播时也要将 `dA1` 除以 `keep_prob`（数学上，如果 $A^{[1]}$ 被缩放了 `keep_prob`，那么它的导数 $dA^{[1]}$ 也会缩放同样的系数）。


````python
def backward_propagation_with_dropout(X, Y, cache, keep_prob):
    """
    实现带 Dropout 的三层神经网络的反向传播：
    LINEAR -> RELU + DROPOUT -> LINEAR -> RELU + DROPOUT -> LINEAR -> SIGMOID
    
    参数：
        X -- 输入数据，维度为 (输入节点数量, 样本数量)
        Y -- 标签，维度为 (输出节点数量, 样本数量)
        cache -- 前向传播时保存的中间结果，由 forward_propagation_with_dropout 返回
        keep_prob -- 保持神经元活跃的概率，标量

    返回：
        gradients -- 字典，包含每一层参数、激活值和预激活值的梯度
    """

    m = X.shape[1]  # 样本数量

    # --------------------------- 1. 从 cache 中取值 ---------------------------
    # 包括每层的线性输出 Z、激活 A、权重 W、偏置 b 和 Dropout mask D
    Z1, D1, A1, W1, b1, Z2, D2, A2, W2, b2, Z3, A3, W3, b3 = cache

    # --------------------------- 2. 第三层输出层 LINEAR -> SIGMOID ---------------------------
    dZ3 = A3 - Y                         # 输出层误差
    dW3 = 1./m * np.dot(dZ3, A2.T)       # 输出层权重梯度
    db3 = 1./m * np.sum(dZ3, axis=1, keepdims=True)  # 输出层偏置梯度
    dA2 = np.dot(W3.T, dZ3)              # 传播到第二层的梯度

    # --------------------------- 3. 第二层 LINEAR -> RELU + DROPOUT ---------------------------
    # Step 1: 应用 Dropout mask，让被关闭的神经元梯度为0
    dA2 = dA2 * D2
    # Step 2: 对未关闭的神经元梯度进行缩放，保证期望一致
    dA2 = dA2 / keep_prob
    # 计算 ReLU 激活梯度
    dZ2 = np.multiply(dA2, np.int64(A2 > 0))
    dW2 = 1./m * np.dot(dZ2, A1.T)       # 第二层权重梯度
    db2 = 1./m * np.sum(dZ2, axis=1, keepdims=True)  # 第二层偏置梯度

    # --------------------------- 4. 第一层 LINEAR -> RELU + DROPOUT ---------------------------
    dA1 = np.dot(W2.T, dZ2)              # 传播到第一层的梯度
    # Step 1: 应用 Dropout mask，让被关闭的神经元梯度为0
    dA1 = dA1 * D1
    # Step 2: 对未关闭的神经元梯度进行缩放
    dA1 = dA1 / keep_prob
    # 计算 ReLU 激活梯度
    dZ1 = np.multiply(dA1, np.int64(A1 > 0))
    dW1 = 1./m * np.dot(dZ1, X.T)        # 第一层权重梯度
    db1 = 1./m * np.sum(dZ1, axis=1, keepdims=True)  # 第一层偏置梯度

    # --------------------------- 5. 保存梯度到字典 ---------------------------
    gradients = {
        "dZ3": dZ3, "dW3": dW3, "db3": db3,
        "dA2": dA2, "dZ2": dZ2, "dW2": dW2, "db2": db2,
        "dA1": dA1, "dZ1": dZ1, "dW1": dW1, "db1": db1
    }

    return gradients
````

````python
X_assess, Y_assess, cache = backward_propagation_with_dropout_test_case()

gradients = backward_propagation_with_dropout(X_assess, Y_assess, cache, keep_prob = 0.8)

print ("dA1 = " + str(gradients["dA1"]))
print ("dA2 = " + str(gradients["dA2"]))
````

- `forward_propagation_with_dropout` 代替 `forward_propagation`。  
- `backward_propagation_with_dropout` 代替 `backward_propagation`。

**注意**：
- 使用 dropout 时的一个**常见错误**是训练和测试阶段都使用 dropout。实际上，dropout（随机关闭神经元）只应在训练时使用，测试时不要使用。
- 像 [TensorFlow](https://www.tensorflow.org/api_docs/python/tf/nn/dropout)、[PaddlePaddle](http://doc.paddlepaddle.org/release_doc/0.9.0/doc/ui/api/trainer_config_helpers/attrs.html)、[Keras](https://keras.io/layers/core/#dropout) 和 [Caffe](http://caffe.berkeleyvision.org/tutorial/layers/dropout.html) 等深度学习框架都内置了 dropout 层实现。不必担心——你很快就会学习这些框架。

<font color='blu
e'>
**关于 dropout 你应该记住的**：
- Dropout 是一种正则化技术。
- 仅在训练时使用 dropout，测试时不要随机关闭神经元。
- 在前向传播和反向传播时都要应用 dropout。
- 训练时，每个 dropout 层的输出除以 keep_prob，以保持激活值的期望不变。例如，keep_prob 为 0.5 时，平均关闭一半神经元，因此输出缩放为 0.5。除以 0.5 相当于乘以 2，从而保证输出的期望值不变。这个方法对其他 keep_prob 值同样适用。
</font>

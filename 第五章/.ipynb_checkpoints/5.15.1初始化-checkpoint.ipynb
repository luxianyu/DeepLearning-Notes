# 初始化

训练神经网络需要先为权重指定一个初始值。一个精心选择的初始化方法可以帮助模型更好地学习。

如果你已经完成了第一部分《神经网络和深度学习》课程，那么你已经学习了权重初始化，并且效果不错。但对于一个新的神经网络，该如何选择初始化方式呢？本节，你将看到不同的初始化方式会带来怎样不同的结果。

一个精心选择的初始化方法可以：
- 加快梯度下降的收敛速度
- 提高梯度下降收敛到较低训练误差（以及泛化误差）的概率

### 自定义的工具函数

```python
# 可以单独作为 init_utils.py 文件，用#import init_utils导入自定义的工具函数
# sigmoid       —— Sigmoid 激活函数
# relu          —— ReLU 激活函数
# compute_loss  —— 计算损失值
# forward_propagation —— 前向传播函数
# backward_propagation —— 反向传播函数
# update_parameters     —— 参数更新函数（梯度下降）
# predict               —— 用训练好的模型进行预测
# load_dataset          —— 加载数据集
# plot_decision_boundary—— 绘制决策边界
# predict_dec           —— 绘制分类预测的颜色分布图
```

```python
import numpy as np
import matplotlib.pyplot as plt
import h5py
import sklearn
import sklearn.datasets

def sigmoid(x):
    """
    Compute the sigmoid of x

    Arguments:
    x -- A scalar or numpy array of any size.

    Return:
    s -- sigmoid(x)
    """
    s = 1/(1+np.exp(-x))
    return s

def relu(x):
    """
    Compute the relu of x

    Arguments:
    x -- A scalar or numpy array of any size.

    Return:
    s -- relu(x)
    """
    s = np.maximum(0,x)
    
    return s

def forward_propagation(X, parameters):
    """
    Implements the forward propagation (and computes the loss) presented in Figure 2.
    
    Arguments:
    X -- input dataset, of shape (input size, number of examples)
    Y -- true "label" vector (containing 0 if cat, 1 if non-cat)
    parameters -- python dictionary containing your parameters "W1", "b1", "W2", "b2", "W3", "b3":
                    W1 -- weight matrix of shape ()
                    b1 -- bias vector of shape ()
                    W2 -- weight matrix of shape ()
                    b2 -- bias vector of shape ()
                    W3 -- weight matrix of shape ()
                    b3 -- bias vector of shape ()
    
    Returns:
    loss -- the loss function (vanilla logistic loss)
    """
        
    # retrieve parameters
    W1 = parameters["W1"]
    b1 = parameters["b1"]
    W2 = parameters["W2"]
    b2 = parameters["b2"]
    W3 = parameters["W3"]
    b3 = parameters["b3"]
    
    # LINEAR -> RELU -> LINEAR -> RELU -> LINEAR -> SIGMOID
    z1 = np.dot(W1, X) + b1
    a1 = relu(z1)
    z2 = np.dot(W2, a1) + b2
    a2 = relu(z2)
    z3 = np.dot(W3, a2) + b3
    a3 = sigmoid(z3)
    
    cache = (z1, a1, W1, b1, z2, a2, W2, b2, z3, a3, W3, b3)
    
    return a3, cache

def backward_propagation(X, Y, cache):
    """
    Implement the backward propagation presented in figure 2.
    
    Arguments:
    X -- input dataset, of shape (input size, number of examples)
    Y -- true "label" vector (containing 0 if cat, 1 if non-cat)
    cache -- cache output from forward_propagation()
    
    Returns:
    gradients -- A dictionary with the gradients with respect to each parameter, activation and pre-activation variables
    """
    m = X.shape[1]
    (z1, a1, W1, b1, z2, a2, W2, b2, z3, a3, W3, b3) = cache
    
    dz3 = 1./m * (a3 - Y)
    dW3 = np.dot(dz3, a2.T)
    db3 = np.sum(dz3, axis=1, keepdims = True)
    
    da2 = np.dot(W3.T, dz3)
    dz2 = np.multiply(da2, np.int64(a2 > 0))
    dW2 = np.dot(dz2, a1.T)
    db2 = np.sum(dz2, axis=1, keepdims = True)
    
    da1 = np.dot(W2.T, dz2)
    dz1 = np.multiply(da1, np.int64(a1 > 0))
    dW1 = np.dot(dz1, X.T)
    db1 = np.sum(dz1, axis=1, keepdims = True)
    
    gradients = {"dz3": dz3, "dW3": dW3, "db3": db3,
                 "da2": da2, "dz2": dz2, "dW2": dW2, "db2": db2,
                 "da1": da1, "dz1": dz1, "dW1": dW1, "db1": db1}
    
    return gradients

def update_parameters(parameters, grads, learning_rate):
    """
    Update parameters using gradient descent
    
    Arguments:
    parameters -- python dictionary containing your parameters 
    grads -- python dictionary containing your gradients, output of n_model_backward
    
    Returns:
    parameters -- python dictionary containing your updated parameters 
                  parameters['W' + str(i)] = ... 
                  parameters['b' + str(i)] = ...
    """
    
    L = len(parameters) // 2 # number of layers in the neural networks

    # Update rule for each parameter
    for k in range(L):
        parameters["W" + str(k+1)] = parameters["W" + str(k+1)] - learning_rate * grads["dW" + str(k+1)]
        parameters["b" + str(k+1)] = parameters["b" + str(k+1)] - learning_rate * grads["db" + str(k+1)]
        
    return parameters

def compute_loss(a3, Y):
    
    """
    Implement the loss function
    
    Arguments:
    a3 -- post-activation, output of forward propagation
    Y -- "true" labels vector, same shape as a3
    
    Returns:
    loss - value of the loss function
    """
    
    m = Y.shape[1]
    logprobs = np.multiply(-np.log(a3),Y) + np.multiply(-np.log(1 - a3), 1 - Y)
    loss = 1./m * np.nansum(logprobs)
    
    return loss

def load_cat_dataset():
    train_dataset = h5py.File('datasets/train_catvnoncat.h5', "r")
    train_set_x_orig = np.array(train_dataset["train_set_x"][:]) # your train set features
    train_set_y_orig = np.array(train_dataset["train_set_y"][:]) # your train set labels

    test_dataset = h5py.File('datasets/test_catvnoncat.h5', "r")
    test_set_x_orig = np.array(test_dataset["test_set_x"][:]) # your test set features
    test_set_y_orig = np.array(test_dataset["test_set_y"][:]) # your test set labels

    classes = np.array(test_dataset["list_classes"][:]) # the list of classes
    
    train_set_y = train_set_y_orig.reshape((1, train_set_y_orig.shape[0]))
    test_set_y = test_set_y_orig.reshape((1, test_set_y_orig.shape[0]))
    
    train_set_x_orig = train_set_x_orig.reshape(train_set_x_orig.shape[0], -1).T
    test_set_x_orig = test_set_x_orig.reshape(test_set_x_orig.shape[0], -1).T
    
    train_set_x = train_set_x_orig/255
    test_set_x = test_set_x_orig/255

    return train_set_x, train_set_y, test_set_x, test_set_y, classes


def predict(X, y, parameters):
    """
    This function is used to predict the results of a  n-layer neural network.
    
    Arguments:
    X -- data set of examples you would like to label
    parameters -- parameters of the trained model
    
    Returns:
    p -- predictions for the given dataset X
    """
    
    m = X.shape[1]
    p = np.zeros((1,m), dtype = np.int)
    
    # Forward propagation
    a3, caches = forward_propagation(X, parameters)
    
    # convert probas to 0/1 predictions
    for i in range(0, a3.shape[1]):
        if a3[0,i] > 0.5:
            p[0,i] = 1
        else:
            p[0,i] = 0

    # print results
    print("Accuracy: "  + str(np.mean((p[0,:] == y[0,:]))))
    
    return p

def plot_decision_boundary(model, X, y):
    # Set min and max values and give it some padding
    x_min, x_max = X[0, :].min() - 1, X[0, :].max() + 1
    y_min, y_max = X[1, :].min() - 1, X[1, :].max() + 1
    h = 0.01
    # Generate a grid of points with distance h between them
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
    # Predict the function value for the whole grid
    Z = model(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    # Plot the contour and training examples
    plt.contourf(xx, yy, Z, cmap=plt.cm.Spectral)
    plt.ylabel('x2')
    plt.xlabel('x1')
    plt.scatter(X[0, :], X[1, :], c=y, cmap=plt.cm.Spectral)
    plt.show()
    
def predict_dec(parameters, X):
    """
    Used for plotting decision boundary.
    
    Arguments:
    parameters -- python dictionary containing your parameters 
    X -- input data of size (m, K)
    
    Returns
    predictions -- vector of predictions of our model (red: 0 / blue: 1)
    """
    
    # Predict using forward propagation and a classification threshold of 0.5
    a3, cache = forward_propagation(X, parameters)
    predictions = (a3>0.5)
    return predictions

def load_dataset():
    np.random.seed(1)
    train_X, train_Y = sklearn.datasets.make_circles(n_samples=300, noise=.05)
    np.random.seed(2)
    test_X, test_Y = sklearn.datasets.make_circles(n_samples=100, noise=.05)
    # Visualize the data
    plt.scatter(train_X[:, 0], train_X[:, 1], c=train_Y, s=40, cmap=plt.cm.Spectral);
    train_X = train_X.T
    train_Y = train_Y.reshape((1, train_Y.shape[0]))
    test_X = test_X.T
    test_Y = test_Y.reshape((1, test_Y.shape[0]))
    return train_X, train_Y, test_X, test_Y
```

```python
# 加载数据集：这里是一个二维的蓝色/红色点分布在同心圆上的数据集
train_X, train_Y, test_X, test_Y = load_dataset()
# train_X —— 训练集特征矩阵（二维坐标点）
# train_Y —— 训练集标签（蓝点或红点，0 或 1）
# test_X  —— 测试集特征矩阵
# test_Y  —— 测试集标签
```

## 1 - 神经网络模型


你将使用一个三层神经网络。下面是你将要尝试的初始化方法：  
- *全零初始化* —— 在输入参数中设置 `initialization = "zeros"`。  
- *随机初始化* —— 在输入参数中设置 `initialization = "random"`。这会将权重初始化为较大的随机值。  
- *He 初始化* —— 在输入参数中设置 `initialization = "he"`。这种方法会根据 He 等人在2015年的论文，将权重初始化为经过缩放的随机值。  

**说明**：请快速阅读下面的代码，并运行它。在下一部分你将实现这个 `model()` 函数所调用的三种初始化方法。


```python
def model(X, Y, learning_rate=0.01, num_iterations=15000, print_cost=True, initialization="he", is_polt=True):
    """
    实现一个三层的神经网络：LINEAR -> RELU -> LINEAR -> RELU -> LINEAR -> SIGMOID
    
    参数：
        X -- 输入的数据，维度为 (输入特征数, 样本数)
        Y -- 标签数据，取值为 0 或 1，维度为 (1, 样本数)
        learning_rate -- 学习率（梯度下降步长）
        num_iterations -- 梯度下降迭代次数
        print_cost -- 是否打印成本值，每 1000 次迭代打印一次
        initialization -- 参数初始化方式，可选：
                          "zeros"  表示所有权重初始化为 0
                          "random" 表示随机初始化
                          "he"     表示 He 初始化
        is_polt -- 是否绘制成本值下降曲线
    返回：
        parameters -- 学习完成后的网络参数（字典类型）
    """

    grads = {}   # 存储反向传播得到的梯度
    costs = []   # 用于记录每隔 1000 次迭代的成本值
    m = X.shape[1]  # 样本数量（列数）
    layers_dims = [X.shape[0], 10, 5, 1]  # 网络结构：[输入层大小, 隐藏层1大小, 隐藏层2大小, 输出层大小]
    
    # 选择参数初始化方法
    if initialization == "zeros":
        parameters = initialize_parameters_zeros(layers_dims)  # 全零初始化
    elif initialization == "random":
        parameters = initialize_parameters_random(layers_dims) # 随机初始化
    elif initialization == "he":
        parameters = initialize_parameters_he(layers_dims)     # He 初始化
    else:
        print("错误的初始化参数！程序退出")
        exit

    # 开始训练（梯度下降）
    for i in range(0, num_iterations):
        # 前向传播
        a3, cache = init_utils.forward_propagation(X, parameters)
        
        # 计算成本（损失）
        cost = init_utils.compute_loss(a3, Y)
        
        # 反向传播（计算梯度）
        grads = init_utils.backward_propagation(X, Y, cache)
        
        # 更新参数
        parameters = init_utils.update_parameters(parameters, grads, learning_rate)
        
        # 每隔 1000 次迭代记录一次成本值
        if i % 1000 == 0:
            costs.append(cost)
            if print_cost:  # 如果需要打印
                print(f"第{i}次迭代，成本值为：{cost}")

    # 绘制成本下降曲线
    if is_polt:
        plt.plot(costs)
        plt.ylabel('cost')
        plt.xlabel('iterations (per thousands)')
        plt.title(f"Learning rate = {learning_rate}")
        plt.show()
    
    # 返回最终学习到的参数
    return parameters
```

## 2 - 全零初始化

在神经网络中，有两类参数需要初始化：
- 权重矩阵 $(W^{[1]}, W^{[2]}, W^{[3]}, ..., W^{[L-1]}, W^{[L]})$
- 偏置向量 $(b^{[1]}, b^{[2]}, b^{[3]}, ..., b^{[L-1]}, b^{[L]})$

**练习**：请实现以下函数，将所有参数初始化为零。你稍后会发现这种方法效果不好，因为它无法“打破对称性”，但我们先试试，看看会发生什么。使用 `np.zeros((..,..))`，并确保形状正确。


```python
def initialize_parameters_zeros(layers_dims):
    """
    将模型的所有参数全部初始化为 0
    
    参数：
        layers_dims - 列表，记录模型的每一层节点数
                      例如 [输入层节点数, 隐藏层1节点数, 隐藏层2节点数, 输出层节点数]
    返回：
        parameters - 字典，存储每一层的权重矩阵 W 和偏置向量 b
            W1 - 第一层的权重矩阵，形状 (layers_dims[1], layers_dims[0])
            b1 - 第一层的偏置向量，形状 (layers_dims[1], 1)
            ...
            WL - 第 L 层的权重矩阵，形状 (layers_dims[L], layers_dims[L-1])
            bL - 第 L 层的偏置向量，形状 (layers_dims[L], 1)
    """

    parameters = {}  # 创建一个空字典，用来存储每层的参数

    L = len(layers_dims)  # 神经网络的总层数（包括输入层和输出层）

    # 从第 1 层（隐藏层）开始，到最后一层
    for l in range(1, L):
        # 权重矩阵 W 全部初始化为 0，形状为 (当前层节点数, 前一层节点数)
        parameters['W' + str(l)] = np.zeros((layers_dims[l], layers_dims[l-1]))
        # 偏置向量 b 全部初始化为 0，形状为 (当前层节点数, 1)
        parameters['b' + str(l)] = np.zeros((layers_dims[l], 1))

        # 使用断言，确保 W 和 b 的形状符合预期
        assert(parameters['W' + str(l)].shape == (layers_dims[l], layers_dims[l-1]))
        assert(parameters['b' + str(l)].shape == (layers_dims[l], 1))

    return parameters  # 返回初始化好的参数字典
```

表现非常差，损失几乎没有下降，算法的效果不比随机猜测好。为什么会这样？让我们来看一下预测结果和决策边界的细节：


```python
print ("predictions_train = " + str(predictions_train))
print ("predictions_test = " + str(predictions_test))
```

模型对每个样本都预测为0。

一般来说，将所有权重初始化为零会导致网络无法打破对称性。这意味着每一层中的所有神经元都会学习到相同的东西，等于是每层神经元数量都变成了 $n^{[l]}=1$，网络的表达能力不比像逻辑回归这样的线性分类器更强。


<font color='blue'>
**你应该记住的点**：
- 权重 $W^{[l]}$ 应该随机初始化以打破对称性。
- 但是偏置 $b^{[l]}$ 初始化为零是可以的。只要权重 $W^{[l]}$ 是随机初始化的，网络的对称性就能被打破。
</font>


## 3 - 随机初始化

为了打破对称性，我们将权重随机初始化。通过随机初始化，每个神经元就可以学习输入的不同特征。在本练习中，你将看到如果权重被初始化为非常大的随机值，会发生什么情况。

**练习**：请实现以下函数，将权重初始化为较大的随机值（乘以10倍），偏置初始化为零。权重使用 `np.random.randn(.., ..) * 10`，偏置使用 `np.zeros((.., ..))`。  
我们使用了固定的 `np.random.seed(..)`，保证你的“随机”权重和我们的相同，因此不要担心多次运行代码时参数初始值总是相同。


```python
def initialize_parameters_random(layers_dims):
    """
    功能：
        随机初始化神经网络每一层的权重 W 和偏置 b，并将权重放大 10 倍
    
    参数：
        layers_dims - 列表（list），依次存储每一层的节点数量
                      例如 [5, 4, 3] 表示输入层有 5 个节点，隐藏层有 4 个节点，输出层有 3 个节点
    
    返回：
        parameters - 字典（dict），包含每一层的权重 W 和偏置 b
            W1 - 权重矩阵，维度为 (layers_dims[1], layers_dims[0])
            b1 - 偏置向量，维度为 (layers_dims[1], 1)
            ...
            WL - 权重矩阵，维度为 (layers_dims[L], layers_dims[L-1])
            bL - 偏置向量，维度为 (layers_dims[L], 1)
    """
    
    np.random.seed(3)  # 固定随机数种子，保证每次运行生成相同的随机数，方便调试与复现实验
    parameters = {}    # 用于存储所有层的 W 和 b
    L = len(layers_dims)  # 网络总层数（包括输入层和输出层）
    
    # 从第一层（隐藏层）开始初始化，到最后一层（输出层）
    for l in range(1, L):
        # 生成服从标准正态分布 N(0,1) 的随机矩阵，维度为 (当前层节点数, 前一层节点数)
        # 乘以 10 是为了故意放大权重，后续会导致梯度爆炸，作为学习初始化方法时的反例
        parameters['W' + str(l)] = np.random.randn(layers_dims[l], layers_dims[l-1]) * 10
        
        # 偏置向量 b 初始化为全零，维度为 (当前层节点数, 1)
        parameters['b' + str(l)] = np.zeros((layers_dims[l], 1))

        # 断言（assert）用于调试：确保 W 的形状正确
        assert(parameters["W" + str(l)].shape == (layers_dims[l], layers_dims[l-1]))
        # 断言（assert）确保 b 的形状正确
        assert(parameters["b" + str(l)].shape == (layers_dims[l], 1))
    
    return parameters  # 返回包含所有 W 和 b 的字典
```

```python
parameters = initialize_parameters_random([3, 2, 1])
print("W1 = " + str(parameters["W1"]))
print("b1 = " + str(parameters["b1"]))
print("W2 = " + str(parameters["W2"]))
print("b2 = " + str(parameters["b2"]))
```

运行以下代码，使用随机初始化对模型进行15,000次迭代的训练。


```python
parameters = model(train_X, train_Y, initialization = "random")
print ("On the train set:")
predictions_train = predict(train_X, train_Y, parameters)
print ("On the test set:")
predictions_test = predict(test_X, test_Y, parameters)
```

无论如何，看起来你已经成功打破了对称性，结果比之前好了很多。模型不再全部输出0了。

```python
plt.title("Model with large random initialization")
axes = plt.gca()
axes.set_xlim([-1.5,1.5])
axes.set_ylim([-1.5,1.5])
plot_decision_boundary(lambda x: predict_dec(parameters, x.T), train_X, train_Y)
```

**观察**：
- 损失起初非常高。这是因为使用较大随机权重时，最后一层的激活函数（sigmoid）对某些样本输出接近0或1的值，当预测错误时，这些样本的损失非常高。确实，当 $\log(a^{[3]}) = \log(0)$ 时，损失会趋向无穷大。
- 不合适的初始化可能导致梯度消失或爆炸，也会减慢优化算法的速度。
- 如果训练时间更长，网络表现会更好，但用过大的随机数初始化会使优化变慢。

<font color='blue'>
**总结**：
- 将权重初始化为非常大的随机值效果不好。
- 希望用较小的随机值初始化效果会更好。关键问题是：这些随机值应该有多小？让我们在下一部分继续探讨！
</font>


## 4 - He 初始化

最后，尝试使用“He 初始化”，该方法以 He 等人（2015年）论文的第一作者命名。（如果你听说过“Xavier 初始化”，它们很相似，区别是 Xavier 初始化的权重 $W^{[l]}$ 缩放因子为 `sqrt(1./layers_dims[l-1])`，而 He 初始化则用 `sqrt(2./layers_dims[l-1])`。）

**练习**：请实现以下函数，使用 He 初始化方法初始化参数。

**提示**：这个函数和之前的 `initialize_parameters_random(...)` 很相似。唯一的区别是，你不再将 `np.random.randn(.., ..)` 乘以10，而是乘以 $\sqrt{\frac{2}{\text{上一层的维度}}}$，这是 He 初始化对使用 ReLU 激活层的建议。


```python
def initialize_parameters_he(layers_dims):
    """
    功能：
        使用 He 初始化方法来设置神经网络每一层的权重 W 和偏置 b
        He 初始化在 ReLU 及其变种激活函数中表现优秀，能有效缓解梯度消失/爆炸问题
    
    参数：
        layers_dims - 列表（list），依次存储每一层的节点数量
                      例如 [5, 4, 3] 表示输入层有 5 个节点，隐藏层有 4 个节点，输出层有 3 个节点
    
    返回：
        parameters - 字典（dict），包含每一层的权重 W 和偏置 b
            W1 - 权重矩阵，维度为 (layers_dims[1], layers_dims[0])
            b1 - 偏置向量，维度为 (layers_dims[1], 1)
            ...
            WL - 权重矩阵，维度为 (layers_dims[L], layers_dims[L-1])
            bL - 偏置向量，维度为 (layers_dims[L], 1)
    """
    
    np.random.seed(3)      # 固定随机数种子，保证实验可复现
    parameters = {}        # 用于存储所有 W 和 b
    L = len(layers_dims)   # 网络总层数（包括输入层和输出层）
    
    for l in range(1, L):
        # np.random.randn(...) 生成服从标准正态分布 N(0,1) 的随机矩阵
        # 乘以 sqrt(2 / 前一层节点数) 实现 He 初始化
        # He 初始化核心思想：保持前向传播和反向传播的方差稳定
        parameters['W' + str(l)] = (
            np.random.randn(layers_dims[l], layers_dims[l - 1]) * np.sqrt(2 / layers_dims[l - 1])
        )
        
        # 偏置 b 初始化为全零
        parameters['b' + str(l)] = np.zeros((layers_dims[l], 1))
        
        # 断言检查权重形状是否正确
        assert(parameters["W" + str(l)].shape == (layers_dims[l], layers_dims[l - 1]))
        # 断言检查偏置形状是否正确
        assert(parameters["b" + str(l)].shape == (layers_dims[l], 1))
        
    return parameters  # 返回包含所有 W 和 b 的字典
```

```python
parameters = initialize_parameters_he([2, 4, 1])
print("W1 = " + str(parameters["W1"]))
print("b1 = " + str(parameters["b1"]))
print("W2 = " + str(parameters["W2"]))
print("b2 = " + str(parameters["b2"]))
```

运行以下代码，使用 He 初始化对模型进行15,000次迭代的训练。


```python
parameters = model(train_X, train_Y, initialization = "he")
print ("On the train set:")
predictions_train = predict(train_X, train_Y, parameters)
print ("On the test set:")
predictions_test = predict(test_X, test_Y, parameters)
```

```python
plt.title("Model with He initialization")
axes = plt.gca()
axes.set_xlim([-1.5,1.5])
axes.set_ylim([-1.5,1.5])
plot_decision_boundary(lambda x: predict_dec(parameters, x.T), train_X, train_Y)
```

**观察**：
- 使用 He 初始化的模型能够在较少的迭代次数内很好地将蓝点和红点分开。

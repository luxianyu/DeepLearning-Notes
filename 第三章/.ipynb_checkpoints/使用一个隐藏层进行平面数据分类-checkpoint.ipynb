## 4 - 神经网络模型

逻辑回归在“花朵数据集”上表现不佳。为取得更好的效果，我们将训练一个带有单隐藏层的神经网络。

**这是我们的模型**：
<img src="images/classification_kiank.png" style="width:600px;height:300px;">

**数学表达式**：

对于一个样本 $x^{(i)}$：
$$z^{[1] (i)} =  W^{[1]} x^{(i)} + b^{[1] (i)}\tag{1}$$ 
$$a^{[1] (i)} = \tanh(z^{[1] (i)})\tag{2}$$
$$z^{[2] (i)} = W^{[2]} a^{[1] (i)} + b^{[2] (i)}\tag{3}$$
$$\hat{y}^{(i)} = a^{[2] (i)} = \sigma(z^{ [2] (i)})\tag{4}$$
$$y^{(i)}_{prediction} = \begin{cases} 1 & \mbox{如果 } a^{[2](i)} > 0.5 \\ 0 & \mbox{否则} \end{cases}\tag{5}$$

针对所有样本的预测结果，你也可以计算成本函数 $J$：
$$J = - \frac{1}{m} \sum\limits_{i = 0}^{m} \large\left(\small y^{(i)}\log\left(a^{[2] (i)}\right) + (1-y^{(i)})\log\left(1- a^{[2] (i)}\right)  \large  \right) \small \tag{6}$$

**提醒**：构建神经网络的一般步骤是：
1. 定义神经网络结构（输入单元数、隐藏单元数等）
2. 初始化模型参数
3. 循环执行：
    - 实现前向传播
    - 计算损失
    - 实现反向传播获取梯度
    - 更新参数（梯度下降）

通常你会写一些辅助函数完成上述步骤，然后把它们合并到一个函数中，称为 `nn_model()`。一旦构建了 `nn_model()` 并学得了正确的参数，就可以用来对新数据进行预测。


### 4.1 - 定义神经网络结构 ####

**练习**：定义三个变量：
- n_x：输入层的大小
- n_h：隐藏层的大小（设为4）
- n_y：输出层的大小

**提示**：利用 `X` 和 `Y` 的形状来确定 `n_x` 和 `n_y`。隐藏层大小固定写成4。


````python
def layer_sizes(X, Y):
    """
    参数说明：
     X -- 输入数据集，维度为 (输入特征数, 样本数)
     Y -- 标签，维度为 (输出类别数, 样本数)
    
    返回值：
     n_x -- 输入层神经元数量（等于输入特征数）
     n_h -- 隐藏层神经元数量（这里固定为4）
     n_y -- 输出层神经元数量（等于输出类别数）
    """
    
    # 输入层大小，取 X 的第一维度，即特征数量
    n_x = X.shape[0]
    
    # 隐藏层大小，这里人为设定为4个神经元（可调节超参数）
    n_h = 4
    
    # 输出层大小，取 Y 的第一维度，即输出类别数量
    n_y = Y.shape[0]
    
    # 返回输入层、隐藏层、输出层的神经元数量，供后续网络构建使用
    return (n_x, n_h, n_y)
````

### 4.2 - 初始化模型参数 ####

**练习**：实现函数 `initialize_parameters()`。

**说明**：
- 确保参数的尺寸正确。如有需要，可参考上方神经网络结构图。
- 权重矩阵使用随机值初始化：
  - 使用 `np.random.randn(a,b) * 0.01` 来随机初始化形状为 (a,b) 的矩阵。
- 偏置向量初始化为全零：
  - 使用 `np.zeros((a,b))` 来初始化形状为 (a,b) 的零矩阵。


````python
def initialize_parameters(n_x, n_h, n_y):
    """
    参数说明：
        n_x -- 输入层神经元数量（特征数）
        n_h -- 隐藏层神经元数量
        n_y -- 输出层神经元数量（类别数）
    
    返回值：
        parameters -- 包含神经网络参数的字典：
            W1 -- 第一层权重矩阵，形状为 (n_h, n_x)
            b1 -- 第一层偏置向量，形状为 (n_h, 1)
            W2 -- 第二层权重矩阵，形状为 (n_y, n_h)
            b2 -- 第二层偏置向量，形状为 (n_y, 1)
    """
    
    # 固定随机种子，保证每次初始化的权重相同，方便调试和复现
    np.random.seed(2)
    
    # 初始化第一层权重，使用标准正态分布随机数乘以0.01缩小初始值，防止梯度爆炸
    W1 = np.random.randn(n_h, n_x) * 0.01
    
    # 第一层偏置初始化为全零，形状为 (n_h, 1)
    b1 = np.zeros((n_h, 1))
    
    # 初始化第二层权重，形状为 (n_y, n_h)，同样乘以0.01缩放
    W2 = np.random.randn(n_y, n_h) * 0.01
    
    # 第二层偏置初始化为全零，形状为 (n_y, 1)
    b2 = np.zeros((n_y, 1))
    
    # 断言检查，确保各参数的维度符合预期，避免后续计算出错
    assert(W1.shape == (n_h, n_x))
    assert(b1.shape == (n_h, 1))
    assert(W2.shape == (n_y, n_h))
    assert(b2.shape == (n_y, 1))
    
    # 将所有参数打包进字典，方便统一管理和传递
    parameters = {
        "W1": W1,
        "b1": b1,
        "W2": W2,
        "b2": b2
    }
    
    # 返回初始化好的参数字典
    return parameters
````

### 4.3 - 训练循环 ####
### 4.3.1 前向传播 ###


**问题**：实现函数 `forward_propagation()`。

**说明**：
- 请参考上方神经网络的数学表达式。
- 你可以使用内置的 `sigmoid()` 函数（已导入）。
- 你可以使用 numpy 库中的 `np.tanh()` 函数。
- 你需要实现的步骤：
  1. 从字典 `parameters`（即 `initialize_parameters()` 的输出）中取出各参数，使用 `parameters[".."]`。
  2. 实现前向传播，计算 $Z^{[1]}, A^{[1]}, Z^{[2]}$ 和 $A^{[2]}$（即所有训练样本的预测结果向量）。
- 前向传播中计算的值需要在反向传播中使用，这些值将存储在 `cache` 中，并作为输入传递给反向传播函数。


````python
def forward_propagation(X, parameters):
    """
    参数说明：
         X -- 输入数据，维度为 (n_x, m)，n_x 是特征数量，m 是样本数量
         parameters -- 初始化参数函数 initialize_parameters 返回的字典，包含 W1, b1, W2, b2
    
    返回值：
         A2 -- 输出层激活值，经过 sigmoid 函数，表示预测的概率，形状为 (1, m)
         cache -- 包含中间计算值的字典，包含 Z1, A1, Z2, A2，用于后续反向传播
    """
    
    # 从参数字典中取出权重和偏置
    W1 = parameters["W1"]  # 第一层权重，形状 (n_h, n_x)
    b1 = parameters["b1"]  # 第一层偏置，形状 (n_h, 1)
    W2 = parameters["W2"]  # 第二层权重，形状 (n_y, n_h)，这里 n_y=1
    b2 = parameters["b2"]  # 第二层偏置，形状 (n_y, 1)
    
    # 前向传播第一步：计算第一层线性组合 Z1 = W1·X + b1，形状 (n_h, m)
    Z1 = np.dot(W1, X) + b1
    
    # 激活函数：使用 tanh 函数对 Z1 激活，得到隐藏层激活值 A1，形状 (n_h, m)
    A1 = np.tanh(Z1)
    
    # 前向传播第二步：计算第二层线性组合 Z2 = W2·A1 + b2，形状 (n_y, m)
    Z2 = np.dot(W2, A1) + b2
    
    # 激活函数：使用 sigmoid 函数对 Z2 激活，得到输出层激活值 A2，表示概率，形状 (1, m)
    A2 = sigmoid(Z2)
    
    # 断言检查，确保输出 A2 形状正确，即 1 行 m 列
    assert(A2.shape == (1, X.shape[1]))
    
    # 缓存中间变量，供反向传播时使用
    cache = {
        "Z1": Z1,
        "A1": A1,
        "Z2": Z2,
        "A2": A2
    }
    
    # 返回输出层激活值和缓存
    return A2, cache
````

````python
# 从测试用例中获取输入数据和参数，用于验证前向传播函数的正确性
X_assess, parameters = forward_propagation_test_case()

# 调用前向传播函数，计算输出激活值和缓存
A2, cache = forward_propagation(X_assess, parameters)

# 计算并打印 cache 中 Z1、A1、Z2、A2 四个矩阵的均值，方便和预期结果对比验证
print(np.mean(cache['Z1']), np.mean(cache['A1']), np.mean(cache['Z2']), np.mean(cache['A2']))

# 打印输出层激活值 A2 的形状，确认其维度是否符合预期 (1, m)
print(A2.shape)

# 打印输入数据 X_assess 的形状，确认输入格式正确
print(X_assess.shape)
````

现在你已经计算出了 $A^{[2]}$（Python 变量名为 "`A2`"），它包含了每个样本的 $a^{[2](i)}$，接下来你可以计算代价函数，公式如下：

$$J = - \frac{1}{m} \sum\limits_{i = 0}^{m} \large{(} \small y^{(i)}\log\left(a^{[2] (i)}\right) + (1-y^{(i)})\log\left(1- a^{[2] (i)}\right) \large{)} \small\tag{7}$$


### 4.3.2 损失计算 ###

**练习**：实现函数 `compute_cost()` 来计算代价 $J$ 的值。

**说明**：
- 计算交叉熵损失有多种方法。为了帮助你，我们给出了计算
$- \sum\limits_{i=0}^{m}  y^{(i)}\log(a^{[2](i)})$ 的一种实现方法：
```python
logprobs = np.multiply(np.log(A2), Y)
cost = - np.sum(logprobs)                # 不需要使用 for 循环！

```

(你可以选择先用 `np.multiply()` 再用 `np.sum()` ，也可以直接用 `np.dot()`实现。).


````python
def compute_cost(A2, Y, parameters):
    """
    计算交叉熵成本（Cross-Entropy Cost），对应公式（7）
    
    参数说明：
         A2 -- 输出层激活值，使用 sigmoid 函数计算，形状为 (1, m)，表示预测概率
         Y -- 真实标签向量，形状为 (1, m)，每个元素是 0 或 1
         parameters -- 包含网络参数的字典（未在函数内使用，可用于扩展）
    
    返回值：
         cost -- 交叉熵损失的标量值，浮点数
    """
    
    # 样本数量 m
    m = Y.shape[1] 

    # 计算每个样本的交叉熵损失（向量形式）
    logprobs = Y * np.log(A2) + (1 - Y) * np.log(1 - A2)
    
    # 计算整体交叉熵损失，取负平均
    cost = -1/m * np.sum(logprobs)
    
    # np.squeeze 用于将数组压缩成标量（例如将 [[17]] 转为 17）
    cost = np.squeeze(cost)
                               
    # 断言确保 cost 是浮点数，方便后续使用
    assert(isinstance(cost, float))
    
    # 返回计算得到的交叉熵损失
    return cost
````

````python
A2, Y_assess, parameters = compute_cost_test_case()

print("cost = " + str(compute_cost(A2, Y_assess, parameters)))
````

### 4.3.3 反向传播 ###

**问题**：实现函数 `backward_propagation()`。

**说明**：
反向传播通常是深度学习中最难（数学性最强）的一部分。下面是关于反向传播的幻灯片截图。你将使用右侧的六个公式，因为你需要实现向量化的计算。

<img src="images/grad_summary.png" style="width:600px;height:300px;">

<!--
$\frac{\partial \mathcal{J} }{ \partial z_{2}^{(i)} } = \frac{1}{m} (a^{[2](i)} - y^{(i)})$

$\frac{\partial \mathcal{J} }{ \partial W_2 } = \frac{\partial \mathcal{J} }{ \partial z_{2}^{(i)} } a^{[1] (i) T} $

$\frac{\partial \mathcal{J} }{ \partial b_2 } = \sum_i{\frac{\partial \mathcal{J} }{ \partial z_{2}^{(i)}}}$

$\frac{\partial \mathcal{J} }{ \partial z_{1}^{(i)} } =  W_2^T \frac{\partial \mathcal{J} }{ \partial z_{2}^{(i)} } * ( 1 - a^{[1] (i) 2}) $

$\frac{\partial \mathcal{J} }{ \partial W_1 } = \frac{\partial \mathcal{J} }{ \partial z_{1}^{(i)} }  X^T $

$\frac{\partial \mathcal{J} _i }{ \partial b_1 } = \sum_i{\frac{\partial \mathcal{J} }{ \partial z_{1}^{(i)}}}$

- 注意，符号 $*$ 表示逐元素相乘。
- 以下符号是深度学习中常用的梯度表示：
    - dW1 = $\frac{\partial \mathcal{J} }{ \partial W_1 }$
    - db1 = $\frac{\partial \mathcal{J} }{ \partial b_1 }$
    - dW2 = $\frac{\partial \mathcal{J} }{ \partial W_2 }$
    - db2 = $\frac{\partial \mathcal{J} }{ \partial b_2 }$
    
!-->

- 小贴士：
    - 计算 dZ1 时，你需要计算激活函数 $g^{[1]'}(Z^{[1]})$。由于 $g^{[1]}(.)$ 是 tanh 激活函数，且如果 $a = g^{[1]}(z)$，则有 $g^{[1]'}(z) = 1 - a^2$。所以你可以用 `(1 - np.power(A1, 2))` 来计算 $g^{[1]'}(Z^{[1]})$。


````python
def backward_propagation(parameters, cache, X, Y):
    """
    使用链式法则实现神经网络的反向传播，计算参数梯度。
    
    参数说明：
     parameters -- 包含权重和偏置的字典，包括 W1, b1, W2, b2
     cache -- 前向传播缓存字典，包含 Z1, A1, Z2, A2
     X -- 输入数据，维度为 (2, m)，2 是输入特征数，m 是样本数量
     Y -- 真实标签，维度为 (1, m)
    
    返回值：
     grads -- 包含参数梯度的字典，包括 dW1, db1, dW2, db2
    """
    
    m = X.shape[1]  # 样本数量
    
    # 从参数字典中取出权重 W1 和 W2
    W1 = parameters["W1"]  # 形状 (n_h, n_x)
    W2 = parameters["W2"]  # 形状 (n_y, n_h)
        
    # 从缓存中取出激活值 A1 和 A2
    A1 = cache["A1"]  # 隐藏层激活，形状 (n_h, m)
    A2 = cache["A2"]  # 输出层激活，形状 (1, m)
    
    # 计算输出层误差项 dZ2 = A2 - Y，形状 (1, m)
    dZ2 = A2 - Y
    
    # 计算第二层权重梯度 dW2，形状 (1, n_h)
    dW2 = (1 / m) * np.dot(dZ2, A1.T)
    
    # 计算第二层偏置梯度 db2，形状 (1, 1)
    db2 = (1 / m) * np.sum(dZ2, axis=1, keepdims=True)
    
    # 计算隐藏层误差项 dZ1，使用 tanh 激活函数的导数，形状 (n_h, m)
    dZ1 = np.dot(W2.T, dZ2) * (1 - np.power(A1, 2))
    
    # 计算第一层权重梯度 dW1，形状 (n_h, n_x)
    dW1 = (1 / m) * np.dot(dZ1, X.T)
    
    # 计算第一层偏置梯度 db1，形状 (n_h, 1)
    db1 = (1 / m) * np.sum(dZ1, axis=1, keepdims=True)
    
    # 将所有梯度存入字典，方便后续参数更新
    grads = {
        "dW1": dW1,
        "db1": db1,
        "dW2": dW2,
        "db2": db2
    }
    
    return grads
````

### 4.3.4 参数更新 ###


**问题**：实现参数更新规则，使用梯度下降法。你需要用梯度 `(dW1, db1, dW2, db2)` 来更新参数 `(W1, b1, W2, b2)`。

**梯度下降的一般更新公式**：  
$$ \theta = \theta - \alpha \frac{\partial J }{ \partial \theta } $$  
其中，$\alpha$ 是学习率，$\theta$ 表示参数。

**示意图**：  
良好学习率下的梯度下降算法（收敛）与不良学习率下的梯度下降算法（发散）。图片由 Adam Harley 提供。

<img src="images/sgd.gif" style="width:400px;height:400px;"> <img src="images/sgd_bad.gif" style="width:400px;height:400;">


````python
def update_parameters(parameters, grads, learning_rate=1.2):
    """
    使用梯度下降法根据计算的梯度更新参数
    
    参数说明：
     parameters -- 包含当前参数的字典，包含 W1, b1, W2, b2
     grads -- 包含对应参数梯度的字典，包含 dW1, db1, dW2, db2
     learning_rate -- 学习率，控制更新步长，默认值为1.2
    
    返回值：
     parameters -- 更新后的参数字典
    """
    
    # 从参数字典中取出当前权重和偏置
    W1 = parameters["W1"]
    b1 = parameters["b1"]
    W2 = parameters["W2"]
    b2 = parameters["b2"]
    
    # 从梯度字典中取出对应的梯度
    dW1 = grads["dW1"]
    db1 = grads["db1"]
    dW2 = grads["dW2"]
    db2 = grads["db2"]
    
    # 根据梯度下降更新规则，更新参数
    W1 = W1 - learning_rate * dW1  # 更新第一层权重
    b1 = b1 - learning_rate * db1  # 更新第一层偏置
    W2 = W2 - learning_rate * dW2  # 更新第二层权重
    b2 = b2 - learning_rate * db2  # 更新第二层偏置
    
    # 将更新后的参数放入字典返回
    parameters = {
        "W1": W1,
        "b1": b1,
        "W2": W2,
        "b2": b2
    }
    
    return parameters
````

### 4.4 - 将第4.1、4.2和4.3部分整合到 `nn_model()` 函数中 ####

**问题**：在 `nn_model()` 中构建你的神经网络模型。

**说明**：神经网络模型需要按照正确顺序调用之前实现的各个函数。


````python
def nn_model(X, Y, n_h, num_iterations=10000, print_cost=False):
    """
    构建并训练一个两层神经网络模型
    
    参数说明：
        X -- 输入数据，维度为 (2, 样本数)
        Y -- 真实标签，维度为 (1, 样本数)
        n_h -- 隐藏层神经元数量
        num_iterations -- 梯度下降的迭代次数，默认10000
        print_cost -- 是否打印成本，每1000次迭代打印一次，默认不打印
    
    返回值：
        parameters -- 训练好的参数字典，可用于预测
    """    
    
    # 设置随机种子，确保结果可复现
    np.random.seed(3)
    
    # 获取输入层和输出层大小
    n_x = layer_sizes(X, Y)[0]  # 输入层大小
    n_y = layer_sizes(X, Y)[2]  # 输出层大小
    
    # 初始化网络参数（权重和偏置）
    parameters = initialize_parameters(n_x, n_h, n_y)
    W1 = parameters["W1"]
    b1 = parameters["b1"]
    W2 = parameters["W2"]
    b2 = parameters["b2"]
    
    # 开始梯度下降迭代训练
    for i in range(num_iterations):
         
        # 前向传播，计算输出和缓存
        A2, cache = forward_propagation(X, parameters)
        
        # 计算当前的交叉熵成本
        cost = compute_cost(A2, Y, parameters)
        
        # 反向传播，计算梯度
        grads = backward_propagation(parameters, cache, X, Y)
        
        # 更新参数
        parameters = update_parameters(parameters, grads)
                  
        # 每1000次迭代打印一次成本（如果设置了 print_cost=True）
        if print_cost and i % 1000 == 0:
            print("Cost after iteration %i: %f" % (i, cost))

    # 返回训练好的参数
    return parameters
````

### 4.5 预测

**问题**：通过构建 `predict()` 函数来使用你的模型进行预测。  
使用前向传播来预测结果。

**提醒**：预测规则如下：  
predictions = $y_{prediction} = \mathbb 1 \text{{activation > 0.5}} = \begin{cases}
      1 & \text{如果}\ activation > 0.5 \\
      0 & \text{否则}
    \end{cases}$  

举个例子：  
如果你想根据某个阈值将矩阵 `X` 的元素设置为0或1，可以这样写：  
```python
X_new = (X > threshold)


````python
def predict(parameters, X):
    """
    使用训练好的参数对输入数据 X 进行预测分类

    参数说明：
    parameters -- 包含神经网络参数的字典，包括 W1, b1, W2, b2
    X -- 输入数据，维度为 (n_x, m)，n_x 是特征数，m 是样本数

    返回值：
    predictions -- 模型的预测结果（向量），每个元素是 0 或 1：
                   0 表示分类为红色，1 表示分类为蓝色
    """
    
    # 使用前向传播计算每个样本的预测概率 A2（形状为 (1, m)）
    A2, cache = forward_propagation(X, parameters)
    
    # 根据阈值 0.5 进行分类判断，概率 ≥ 0.5 划为 1，否则划为 0
    predictions = np.round(A2)
    
    return predictions
````

### 4.6 - 调整隐藏层大小 ###

运行以下代码，可能会花费 1-2 分钟的时间。你将观察到模型在不同隐藏层大小下的表现差异。


````python
# 这段代码将测试不同隐藏层大小的神经网络对数据的分类能力，并可视化决策边界

# 设置整张图的画布尺寸为宽度 16 英寸，高度 32 英寸
plt.figure(figsize=(16, 32))  

# 定义一个列表，包含不同的隐藏层神经元数量
hidden_layer_sizes = [1, 2, 3, 4, 5, 10, 20]  

# 遍历上述每种隐藏层大小
for i, n_h in enumerate(hidden_layer_sizes):
    
    # 在 5 行 2 列的图像网格中创建第 i+1 个子图
    plt.subplot(5, 2, i + 1)  
    
    # 设置当前子图的标题，显示隐藏层的大小
    plt.title('Hidden Layer of size %d' % n_h)  
    
    # 使用当前隐藏层大小训练一个神经网络，迭代次数为 5000 次
    parameters = nn_model(X, Y, n_h, num_iterations=5000)
    
    # 绘制该模型对应的决策边界
    # 使用 lambda 函数包装 predict，以便对每个点进行预测
    plot_decision_boundary(lambda x: predict(parameters, x.T), X, Y[0, :])
    
    # 用训练好的模型对训练集进行预测
    predictions = predict(parameters, X)
    
    # 计算预测准确率
    # np.dot(Y, predictions.T) 表示预测为1且标签为1的数量
    # np.dot(1-Y, 1-predictions.T) 表示预测为0且标签为0的数量
    # 相加后除以总样本数，再乘100即为准确率百分比
    accuracy = float((np.dot(Y, predictions.T) + np.dot(1 - Y, 1 - predictions.T)) / float(Y.size) * 100)
    
    # 打印当前隐藏层大小下的准确率
    print("Accuracy for {} hidden units: {} %".format(n_h, accuracy))
````

**解读**：  
- 较大的模型（隐藏单元更多）能够更好地拟合训练集，但当模型变得过大时，会出现过拟合现象。  
- 最佳的隐藏层大小似乎在 n_h = 5 附近。这个数值能够较好地拟合数据，同时避免明显的过拟合。  
- 你后续还会学习正则化技术，它能让你使用非常大的模型（例如 n_h = 50）而不至于过拟合。

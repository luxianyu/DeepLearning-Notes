### 2 数据加载函数

````python
# ==============================
# 数据加载函数
# ==============================
def load_dataset():
    """
    功能：
        从 HDF5 文件中加载训练集和测试集的数据和标签
        
    返回：
        train_set_x_orig -- 训练集特征，shape = (m_train, 64, 64, 3)，每张图片为 64x64 RGB 图像
        train_set_y_orig -- 训练集标签，shape = (m_train,)，整数表示类别
        test_set_x_orig  -- 测试集特征，shape = (m_test, 64, 64, 3)
        test_set_y_orig  -- 测试集标签，shape = (m_test,)
        classes          -- 标签类别数组，shape = (n_classes,)
    """

    # ------------------------------
    # 加载训练集 HDF5 文件
    # ------------------------------
    train_dataset = h5py.File('datasets/train_happy.h5', "r")
    # train_dataset: HDF5 文件对象
    # 参数 "datasets/train_signs.h5" 表示文件路径
    # "r" 表示只读模式，不能修改文件内容

    train_set_x_orig = np.array(train_dataset["train_set_x"][:])
    # train_dataset["train_set_x"]: HDF5 文件中的训练特征数据集
    # [:] 表示读取整个数据集
    # np.array(...) 将 HDF5 数据转为 numpy 数组，方便后续处理
    # 结果 shape = (m_train, 64, 64, 3)

    train_set_y_orig = np.array(train_dataset["train_set_y"][:])
    # train_dataset["train_set_y"]: HDF5 文件中的训练标签数据集
    # np.array(...) 转为 numpy 数组
    # 结果 shape = (m_train,)，每个元素为整数类别

    # ------------------------------
    # 加载测试集 HDF5 文件
    # ------------------------------
    test_dataset = h5py.File('datasets/test_happy.h5', "r")
    # test_dataset: HDF5 文件对象，测试集数据
    # "r" 表示只读模式

    test_set_x_orig = np.array(test_dataset["test_set_x"][:])
    # 测试集特征数据，shape = (m_test, 64, 64, 3)

    test_set_y_orig = np.array(test_dataset["test_set_y"][:])
    # 测试集标签数据，shape = (m_test,)

    classes = np.array(test_dataset["list_classes"][:])
    # list_classes: 类别列表，如 [0,1,2,...,5]
    # np.array(...) 转为 numpy 数组，方便索引和分类

    # ------------------------------
    # 返回加载的数据
    # ------------------------------
    return train_set_x_orig, train_set_y_orig, test_set_x_orig, test_set_y_orig, classes
````

### 3 自定义Dataset和DataLoader 

````python
# ==============================
# 自定义 Dataset类
# ==============================
class SignsDataset(Dataset):
    """
    功能：
        将 NumPy 数据集封装为 PyTorch Dataset，用于 DataLoader 迭代
        支持图像归一化、通道转换以及可选 transform
    """

    def __init__(self, X_np, Y_np, transform=None):
        """
        初始化 Dataset

        参数：
        X_np     -- NumPy 图像数组，shape = (m,64,64,3)，dtype uint8 或 float32
        Y_np     -- NumPy 标签数组，shape = (m,) 或 (m,1)
        transform-- 可选图像变换（如 torchvision.transforms）
        """

        # ------------------------------
        # 特征数据转换为 float32 类型
        # ------------------------------
        self.X = X_np.astype(np.float32)
        # astype(np.float32)：确保数据类型为 float32，以便后续 PyTorch 计算

        # 如果最大值 > 2.0，说明是原始 0-255 图像，需要归一化到 [0,1]
        if self.X.max() > 2.0:
            self.X = self.X / 255.0

        # ------------------------------
        # PyTorch 要求图像通道在前（C,H,W）
        # ------------------------------
        self.X = np.transpose(self.X, (0,3,1,2))
        # np.transpose：调整数组维度顺序
        # (0,3,1,2) 说明：
        # 0: 样本数 m 不变
        # 3: 通道数 C 从最后一维移到第二维
        # 1: 高度 H
        # 2: 宽度 W
        # 最终 shape = (m,3,64,64)

        # ------------------------------
        # 标签处理
        # ------------------------------
        self.y = np.array(Y_np).reshape(-1).astype(np.int64)
        # reshape(-1)：确保标签为一维向量
        # int64：PyTorch 分类任务要求标签为 long 类型

        # ------------------------------
        # 图像变换（可选）
        # ------------------------------
        self.transform = transform
        # 如果传入 transform（如随机裁剪、旋转等），后续 __getitem__ 会应用

    # ------------------------------
    # 返回样本总数
    # ------------------------------
    def __len__(self):
        return self.X.shape[0]
        # Dataset 长度就是样本数 m

    # ------------------------------
    # 返回单个样本和标签
    # ------------------------------
    def __getitem__(self, idx):
        """
        功能：
            根据索引 idx 返回单个样本图像和标签

        参数：
        idx -- 样本索引，整数

        返回：
        img   -- torch.Tensor 图像，shape = (3,64,64)，dtype float32
        label -- torch.Tensor 标签，dtype long
        """

        # 取第 idx 张图像
        img = self.X[idx]       

        # 取对应标签
        label = self.y[idx]     

        # 如果定义了 transform，应用图像变换
        if self.transform:
            img = self.transform(img)

        # 转换为 PyTorch Tensor
        return torch.from_numpy(img), torch.tensor(label, dtype=torch.long)
    
        # ------------------------------
        # torch.from_numpy(img)
        # 功能：将 NumPy 数组 img 转换为 PyTorch Tensor
        # 输入 img：
        #   - 形状为 (C, H, W)，通道数在前
        #   - 数据类型通常为 float32
        # 输出：
        #   - PyTorch Tensor，保持原数组数据
        #   - 与原 NumPy 数组共享内存，修改 Tensor 会影响原数组
        # 用途：
        #   - PyTorch 模型训练需要 Tensor 输入，而不是 NumPy 数组
        #   - 这里 img 是单张图像，用于 Dataset 返回一个样本
        
        # torch.tensor(label, dtype=torch.long)
        # 功能：将单个标签 label 转换为 PyTorch Tensor
        # 参数：
        #   - label：样本的类别标签（整数）
        #   - dtype=torch.long：指定数据类型为 long (int64)，PyTorch 分类任务要求
        # 输出：
        #   - PyTorch Tensor，形状为标量 ( ) 或 (1,)
        #   - 数据类型为 torch.int64
        # 用途：
        #   - 作为目标标签，输入到分类损失函数 nn.CrossEntropyLoss 时必须是 long 类型
````

````python
# ==============================
# 创建 Dataset 和 DataLoader
# ==============================

# 定义类别数
num_classes = 6
# 原始脚本中 X_train/X_test 已除以 255，这里在 SignsDataset 内部已做归一化

# 使用自定义 SignsDataset 封装训练集和测试集
train_dataset = SignsDataset(X_train_orig, Y_train_orig)
# train_dataset: 封装训练集的 Dataset 对象，可用于 DataLoader 迭代
test_dataset  = SignsDataset(X_test_orig,  Y_test_orig)
# test_dataset: 封装测试集的 Dataset 对象

# 定义每个 batch 的大小
batch_size = 64

# ------------------------------
# 设置随机种子保证 DataLoader shuffle 可复现
# ------------------------------
g = torch.Generator()   # 创建一个随机数生成器
g.manual_seed(3)        # 设置固定种子 3，确保每次 shuffle 顺序相同

# 创建训练集 DataLoader
train_loader = DataLoader(
    train_dataset,      # 数据集对象
    batch_size=batch_size,  # 每个 batch 的样本数
    shuffle=True,       # 打乱样本顺序
    generator=g         # 使用固定种子生成器，保证可复现
)

# 创建测试集 DataLoader
test_loader  = DataLoader(
    test_dataset,       # 数据集对象
    batch_size=batch_size,  # 每个 batch 的样本数
    shuffle=False       # 测试集不打乱
)

````

### 4 定义模型

````python
# ==============================
# PyTorch 模型定义
# ==============================

def HappyModel(input_shape):
    """
    功能：
        构建一个用于笑容检测的简单卷积神经网络（PyTorch 实现）
    
    参数：
        input_shape -- 输入图像的形状，格式为 (C, H, W)
                       C = 通道数（如 RGB=3），
                       H = 图像高度，
                       W = 图像宽度
    
    返回：
        model -- 一个 nn.Sequential 定义的模型
    """
    # 解包输入图像的形状
    C, H, W = input_shape

    # 使用 nn.Sequential 顺序构建网络
    model = nn.Sequential(
        # ------------------------------
        # 卷积层 (Conv2d)
        # ------------------------------
        nn.Conv2d(
            in_channels=C,    # 输入通道数，等于输入图像的通道数 C
            out_channels=32,  # 卷积核个数，即输出特征图的通道数
            kernel_size=7,    # 卷积核大小 7x7
            stride=1,         # 步幅为 1
            padding=3         # 填充 3，使输出大小保持接近输入（SAME 效果）
        ),

        # ------------------------------
        # 批归一化 (BatchNorm2d)
        # ------------------------------
        nn.BatchNorm2d(
            num_features=32   # 特征图通道数 = conv 输出通道数
        ),

        # ------------------------------
        # 激活函数 (ReLU)
        # ------------------------------
        nn.ReLU(),

        # ------------------------------
        # 最大池化层 (MaxPool2d)
        # ------------------------------
        nn.MaxPool2d(
            kernel_size=2,    # 池化窗口大小 2x2
            stride=2          # 步幅 2，特征图宽高缩小一半
        ),

        # ------------------------------
        # 展平层 (Flatten)
        # ------------------------------
        nn.Flatten(),  # 将张量展平为 (batch_size, 特征数)

        # ------------------------------
        # 全连接层 (Linear)
        # ------------------------------
        nn.Linear(
            in_features=32 * (H // 2) * (W // 2),  # 输入特征数 = 池化后特征图的元素总数
            out_features=1                         # 输出一个数（用于二分类）
        ),

        # ------------------------------
        # Sigmoid 激活函数
        # ------------------------------
        nn.Sigmoid()   # 将输出压缩到 [0,1]，作为笑容概率
    )

    return model
````

### 5 创建模型实体

````python
# ==============================
# 创建模型实体
# ==============================

# 选择运行设备：
# torch.device() 是 PyTorch 中用来指定张量或模型运行设备的函数
# "cuda" 表示使用 GPU (如果本机安装了支持 CUDA 的显卡和驱动)
# "cpu"  表示使用 CPU 进行计算
# torch.cuda.is_available() 会返回布尔值，判断当前环境是否支持 CUDA
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# 调用我们前面定义的 HappyModel 来创建一个模型实例
# 传入输入的形状 (3,64,64)，
#   - 3 表示图像有 3 个通道（RGB 彩色图像）
#   - 64 表示输入图像的高度为 64 像素
#   - 64 表示输入图像的宽度为 64 像素
happy_model = HappyModel((3,64,64)).to(device)  

# .to(device) 的作用：
#   - 将模型的所有参数和缓存迁移到指定的计算设备上（CPU 或 GPU）
#   - 这样之后在训练或推理时，数据和模型就在同一设备上，不会发生设备冲突
````

````python
# ==============================
# 模型结构展示
# ==============================

# print(happy_model) 用于打印模型的结构信息
# 当模型是用 nn.Sequential 定义时，print 输出会按照顺序列出每一层
# 例如：
# Sequential(
#   (0): Conv2d(3, 32, kernel_size=(7, 7), stride=(1, 1), padding=(3, 3))
#   (1): BatchNorm2d(32, eps=1e-05, momentum=0.1, affine=True, track_running_stats=True)
#   (2): ReLU()
#   (3): MaxPool2d(kernel_size=2, stride=2, padding=0, dilation=1, ceil_mode=False)
#   (4): Flatten(start_dim=1, end_dim=-1)
#   (5): Linear(in_features=65536, out_features=1, bias=True)
#   (6): Sigmoid()
# )
# 
# 各层含义：
# (0) 卷积层：输入通道数=3，输出通道数=32，卷积核大小=7x7，步长=1，填充=3
# (1) 批量归一化层：作用于 32 个特征通道
# (2) ReLU 激活函数：增加非线性
# (3) 最大池化层：核大小 2x2，步长=2，用于下采样
# (4) Flatten：把多维特征图展平为一维向量
# (5) 全连接层：输入 65536 个特征，输出 1 个数值
# (6) Sigmoid：将输出压缩到 (0,1) 之间，作为概率
print(happy_model)
````

### 6 损失函数和优化器

````python
# ==============================
# 损失函数和优化器
# ==============================

# 定义损失函数：二元交叉熵（Binary Cross Entropy Loss）
# nn.BCELoss() 要求：
#   - 模型的输出值在 [0,1] 之间（通过 Sigmoid 层保证）
#   - 标签 (Y) 为 0 或 1
# 公式:  L = - [ y*log(p) + (1-y)*log(1-p) ]
#       y 为真实标签, p 为预测概率
criterion = nn.BCELoss()                  

# 定义优化器：Adam 优化器
# 参数：
#   happy_model.parameters() -> 告诉优化器需要更新哪些参数（模型的所有可训练参数）
#   lr=0.001 -> 学习率，控制每一步参数更新的幅度
# Adam 特点：
#   - 结合了 Momentum（动量）和 RMSProp（自适应学习率）
#   - 在训练深度神经网络时收敛更快且效果稳定
optimizer = torch.optim.Adam(happy_model.parameters(), lr=0.001)
````

### 7 训练模型

````python
# ==============================
# 训练模型
# ==============================

# 定义训练的轮数（整个训练集会被完整地使用 num_epochs 次）
num_epochs = 40

# 外层循环：按轮次进行训练
for epoch in range(num_epochs):
    happy_model.train()      # 将模型设置为训练模式（启用 Dropout/BatchNorm 的训练行为）
    running_loss = 0.0       # 累积当前轮的总损失
    correct = 0              # 累积预测正确的样本数
    total = 0                # 累积样本总数

    # 内层循环：按 batch 遍历训练数据
    for X_batch, Y_batch in train_loader:
        # 将数据和标签放到 GPU 或 CPU（根据 device 自动选择）
        # Y_batch 需要转成 float 并在最后增加一个维度 (N,1)，以便与模型 Sigmoid 输出对齐
        X_batch, Y_batch = X_batch.to(device), Y_batch.to(device).float().unsqueeze(1)

        # 1. 梯度清零
        optimizer.zero_grad()

        # 2. 前向传播：得到预测输出
        outputs = happy_model(X_batch)

        # 3. 计算损失函数（与真实标签对比）
        loss = criterion(outputs, Y_batch)

        # 4. 反向传播：计算梯度
        loss.backward()

        # 5. 更新参数：Adam 优化器根据梯度调整模型参数
        optimizer.step()

        # 6. 累积损失
        running_loss += loss.item() * X_batch.size(0)  # loss.item() 是 batch 平均损失，需乘 batch 大小

        # 7. 计算准确率
        predicted = (outputs >= 0.5).float()           # 预测概率 >=0.5 判为正类(1)，否则为负类(0)
        correct += (predicted == Y_batch).sum().item() # 累加预测正确的个数
        total += Y_batch.size(0)                       # 累加样本数

    # 每个 epoch 的平均损失和准确率
    epoch_loss = running_loss / total
    epoch_acc  = correct / total

    # 打印本轮训练的结果
    print(f"Epoch [{epoch+1}/{num_epochs}]  Loss: {epoch_loss:.4f}  Accuracy: {epoch_acc:.4f}")
````

### 8 评估模型

````python
# ==============================
# 评估模型
# ==============================

# 将模型切换到评估模式（禁用 Dropout/BatchNorm 的训练行为）
happy_model.eval()

# 初始化测试集的损失、正确预测数和总样本数
test_loss = 0.0
correct = 0
total = 0

# 在评估时关闭梯度计算，提高推理速度并节省显存
with torch.no_grad():
    # 遍历测试集的所有 batch
    for X_batch, Y_batch in test_loader:
        # 将输入数据和标签放到 GPU 或 CPU
        # Y_batch 转成 float 并扩展维度 (N,1)，保持与模型输出一致
        X_batch, Y_batch = X_batch.to(device), Y_batch.to(device).float().unsqueeze(1)

        # 前向传播，得到预测输出
        outputs = happy_model(X_batch)

        # 计算测试集的损失
        loss = criterion(outputs, Y_batch)
        # 累积当前 batch 的总损失（loss.item() 是平均损失，需乘 batch 大小）
        test_loss += loss.item() * X_batch.size(0)

        # 将预测结果二值化：>=0.5 判为 1，否则判为 0
        predicted = (outputs >= 0.5).float()
        # 累加预测正确的个数
        correct += (predicted == Y_batch).sum().item()
        # 累加样本总数
        total += Y_batch.size(0)

# 计算测试集平均损失
test_loss /= total
# 计算测试集准确率
test_acc = correct / total

# 打印最终的测试误差和准确度
print("误差值 =", test_loss)
print("准确度 =", test_acc)
````
